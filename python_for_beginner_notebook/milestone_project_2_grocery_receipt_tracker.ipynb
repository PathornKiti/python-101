# Part 0: Create the demo receipt file

Run this cell first.

The file uses this simple format:

```text
item_name,category,price
```

Example:

```text
milk,dairy,42
```

````python
from pathlib import Path

BASE = Path("milestone_project_2_files")
BASE.mkdir(exist_ok=True)

receipt_path = BASE / "grocery_receipt.txt"

receipt_path.write_text(
    " milk , dairy , 42\n"
    "bread,bakery,35\n"
    "apple,fruit,20\n"
    "banana,fruit,20\n"
    "milk,dairy,42\n"
    "orange juice,drink,78\n"
    "potato chips,snack,55\n"
    "chocolate bar,snack,47\n"
    "toothpaste,personal care,89\n",
    encoding="utf-8"
)

print("Created file:", receipt_path)
````

# Part 1: Read the file line by line

Goal:

- open the file,
- loop through each line,
- remove the newline character,
- print each raw line.

````python
with open(receipt_path, "r", encoding="utf-8") as f:
    for line in f:
        clean_line = line.rstrip("\n")
        print(repr(clean_line))
````

# Part 2: Clean strings and split each line

Goal:

For each line:

1. remove extra spaces,
2. split the line by comma,
3. clean item name and category,
4. convert price from string to integer.

Example:

```python
" milk , dairy , 42"
```

should become:

```python
item = "Milk"
category = "Dairy"
price = 42
```

````python
sample_line = " milk , dairy , 42"

parts = sample_line.split(",")

item = parts[0].strip().title()
category = parts[1].strip().title()
price = int(parts[2].strip())

print(item)
print(category)
print(price)
````

# Part 3: Store grocery records in a list of tuples

Each record will be a tuple:

```python
(item, category, price)
```

Example:

```python
("Milk", "Dairy", 42)
```

All records will be stored in a list named `records`.

````python
records = []

with open(receipt_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.rstrip("\n")
        parts = line.split(",")

        item = parts[0].strip().title()
        category = parts[1].strip().title()
        price = int(parts[2].strip())

        record = (item, category, price)
        records.append(record)

print(records)
````

# Part 4: Basic list summary

Use list skills to answer:

1. How many records are there?
2. What is the first record?
3. What is the last record?
4. What are the first 3 records?

````python
print("Number of records:", len(records))
print("First record:", records[0])
print("Last record:", records[-1])
print("First 3 records:", records[:3])
````

# Part 5: Loop through tuples

Because each record is a tuple, we can unpack it:

```python
item, category, price = record
```

Goal:

Print each item in a readable sentence.

````python
for record in records:
    item, category, price = record
    print(item + " is in " + category + " and costs " + str(price) + " baht.")
````

# Part 6: Calculate total, average, highest, and lowest price

Use:

- list
- loop
- `sum()`
- `max()`
- `min()`

````python
prices = []

for record in records:
    item, category, price = record
    prices.append(price)

total_spending = sum(prices)
average_price = total_spending / len(prices)
highest_price = max(prices)
lowest_price = min(prices)

print("Total spending:", total_spending, "baht")
print("Average price:", average_price, "baht")
print("Highest price:", highest_price, "baht")
print("Lowest price:", lowest_price, "baht")
````

# Part 7: Find the most expensive item

Goal:

Use a loop to find the item with the highest price.

````python
most_expensive_item = ""
most_expensive_price = 0

for record in records:
    item, category, price = record

    if price > most_expensive_price:
        most_expensive_item = item
        most_expensive_price = price

print("Most expensive item:", most_expensive_item)
print("Price:", most_expensive_price, "baht")
````

# Part 8: Use a set to find unique categories

A set removes duplicates automatically.

Goal:

Create a set of categories.

````python
categories = set()

for record in records:
    item, category, price = record
    categories.add(category)

print("Unique categories:")
for category in sorted(categories):
    print(category)
````

# Part 9: Use a set to find duplicate item names

Goal:

Find items that appear more than once.

We will use two sets:

- `seen_items`: items we have already seen
- `duplicate_items`: items that appear again

````python
seen_items = set()
duplicate_items = set()

for record in records:
    item, category, price = record

    if item in seen_items:
        duplicate_items.add(item)
    else:
        seen_items.add(item)

print("Duplicate item names:")
for item in sorted(duplicate_items):
    print(item)
````

# Part 10: Use a dictionary to calculate spending by category

Goal:

Create a dictionary like this:

```python
{
    "Dairy": 84,
    "Fruit": 40
}
```

The key is the category name.

The value is the total spending in that category.

````python
category_totals = {}

for record in records:
    item, category, price = record

    if category not in category_totals:
        category_totals[category] = 0

    category_totals[category] = category_totals[category] + price

print(category_totals)
````

# Part 11: Print the category summary nicely

Use dictionary looping and sorting.

````python
print("Spending by category:")

for category in sorted(category_totals):
    print(category + ":", category_totals[category], "baht")
````

# Part 12: Use a dictionary to count each item

Goal:

Create a dictionary like this:

```python
{
    "Milk": 2,
    "Bread": 1
}
```

````python
item_counts = {}

for record in records:
    item, category, price = record

    if item not in item_counts:
        item_counts[item] = 0

    item_counts[item] = item_counts[item] + 1

print(item_counts)
````

# Part 13: Create the final report text

Use strings and `join()` to build a readable report.

````python
report_lines = []

report_lines.append("DAILY GROCERY RECEIPT SUMMARY")
report_lines.append("-" * 36)
report_lines.append("Total records: " + str(len(records)))
report_lines.append("Unique item names: " + str(len(seen_items)))
report_lines.append("Total spending: " + str(total_spending) + " baht")
report_lines.append("Average item price: " + str(round(average_price, 2)) + " baht")
report_lines.append("Most expensive item: " + most_expensive_item + " - " + str(most_expensive_price) + " baht")
report_lines.append("")
report_lines.append("Spending by category:")

for category in sorted(category_totals):
    report_lines.append(category + ": " + str(category_totals[category]) + " baht")

report_lines.append("")
report_lines.append("Duplicate item names:")

if len(duplicate_items) == 0:
    report_lines.append("No duplicate item")
else:
    for item in sorted(duplicate_items):
        report_lines.append(item)

report = "\n".join(report_lines)

print(report)
````

# Part 14: Save the report to a new file

Goal:

Write the final report to `grocery_summary.txt`.

````python
summary_path = BASE / "grocery_summary.txt"

with open(summary_path, "w", encoding="utf-8") as f:
    f.write(report)

print("Saved summary to:", summary_path)
````

# Part 15: Read the saved summary file

Goal:

Check that the file was saved correctly.

````python
with open(summary_path, "r", encoding="utf-8") as f:
    for line in f:
        print(line.rstrip("\n"))
````
