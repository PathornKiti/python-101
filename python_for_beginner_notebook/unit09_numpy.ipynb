# Sample 1 — Python list vs NumPy array

A Python list is flexible and can store many kinds of values.

A NumPy array is designed for numerical work.

The biggest beginner difference is how operators behave.

| Operation | Python list | NumPy array |
|---|---|---|
| `+` | joins lists | adds values position by position |
| `* 2` | repeats the list | multiplies each value by 2 |

Daily-life idea: suppose we have daily expenses. NumPy can add or multiply all values at once.


````python
import numpy as np

list_a = [1, 2, 3]
list_b = [10, 20, 30]

array_a = np.array([1, 2, 3])
array_b = np.array([10, 20, 30])

print("Python list + list:", list_a + list_b)
print("NumPy array + array:", array_a + array_b)
print()
print("Python list * 2:", list_a * 2)
print("NumPy array * 2:", array_a * 2)
````


# Sample 2 — Creating arrays and inspecting them

Common NumPy vocabulary:

| Word | Meaning | Example |
|---|---|---|
| `ndarray` | NumPy array object | `np.array([1, 2, 3])` |
| `shape` | size of each dimension | `(7,)`, `(3, 4)` |
| `ndim` | number of dimensions | `1`, `2` |
| `size` | total number of values | `12` |
| `dtype` | data type stored in the array | `int64`, `float64` |

For a 2D array, think about a table:

- `shape = (rows, columns)`
- `axis=0` means down the rows, summarize each column
- `axis=1` means across the columns, summarize each row


````python
import numpy as np

weekly_steps = np.array([5000, 7000, 6500, 8000, 9000, 4000, 3000])

# Rows = students, columns = subjects
scores = np.array([
    [80, 75, 90],
    [60, 85, 70],
    [95, 90, 88]
])

print("weekly_steps:", weekly_steps)
print("shape:", weekly_steps.shape)
print("ndim:", weekly_steps.ndim)
print("size:", weekly_steps.size)
print("dtype:", weekly_steps.dtype)

print("\nscores table:")
print(scores)
print("shape:", scores.shape)
print("ndim:", scores.ndim)
print("size:", scores.size)
print("dtype:", scores.dtype)
````


# Sample 3 — Creating useful arrays

Common beginner methods:

| Method | Meaning | Example |
|---|---|---|
| `np.array(list)` | create from an existing list | `np.array([1, 2, 3])` |
| `np.zeros(shape)` | create zeros | `np.zeros((2, 3))` |
| `np.ones(shape)` | create ones | `np.ones((3, 2))` |
| `np.arange(start, stop, step)` | create values with a step | `np.arange(5, 31, 5)` |
| `np.linspace(start, stop, count)` | create evenly spaced values | `np.linspace(0, 100, 5)` |
| `np.eye(n)` | create identity matrix | `np.eye(3)` |

Note: `stop` in `np.arange(start, stop, step)` is not included.


````python
import numpy as np

print("zeros:")
print(np.zeros((2, 3), dtype=int))

print("\nones:")
print(np.ones((3, 2), dtype=int))

print("\narange:")
print(np.arange(5, 31, 5))

print("\nlinspace:")
print(np.linspace(0, 100, 5))

print("\nidentity matrix:")
print(np.eye(3, dtype=int))
````


# Sample 4 — Element-wise operations

Element-wise means NumPy applies the operation to each value.

Daily-life examples:

- add bonus points to every student score,
- multiply every price by tax,
- calculate item total from price and quantity.


````python
import numpy as np

# Example 1: add bonus points to all scores
scores = np.array([65, 70, 85, 90])
bonus_scores = scores + 10

print("original scores:", scores)
print("after bonus:", bonus_scores)

# Example 2: add 7% VAT to item prices
prices = np.array([100, 250, 80, 40])
prices_with_vat = prices * 1.07

print("\nprices:", prices)
print("prices with VAT:", prices_with_vat)

# Example 3: price x quantity
price = np.array([100, 200, 50])
quantity = np.array([2, 1, 4])
item_total = price * quantity

print("\nitem total:", item_total)
print("cart total:", item_total.sum())
````


# Sample 5 — Broadcasting

Broadcasting means NumPy automatically expands a smaller value or smaller array so that an operation is possible.

The easiest case is:

```python
array + single_number
```

The single number is applied to every element.

A useful 2D case:

- `scores` has shape `(3, 3)`
- `subject_bonus` has shape `(3,)`

NumPy treats `subject_bonus` like one row and adds it to every student.


````python
import numpy as np

scores = np.array([
    [80, 70, 90],
    [60, 85, 75],
    [90, 88, 92]
])

# Bonus for math, English, science
subject_bonus = np.array([5, 0, 3])

new_scores = scores + subject_bonus

print("Original scores:")
print(scores)
print("\nSubject bonus:", subject_bonus)
print("\nNew scores:")
print(new_scores)
````


# Sample 6 — Indexing and slicing

Indexing selects one value.

Slicing selects a range of values.

Python index starts at `0`.

| Expression | Meaning |
|---|---|
| `arr[0]` | first value |
| `arr[-1]` | last value |
| `arr[:3]` | first three values |
| `arr[2:5]` | values from index 2 to 4 |
| `matrix[row, col]` | one value in a 2D array |
| `matrix[row, :]` | all columns of one row |
| `matrix[:, col]` | all rows of one column |


````python
import numpy as np

steps = np.array([5000, 7000, 6500, 8000, 9000, 4000, 3000])

print("Monday steps:", steps[0])
print("Sunday steps:", steps[-1])
print("First three days:", steps[:3])
print("Weekend:", steps[5:7])

scores = np.array([
    [80, 70, 90],
    [60, 85, 75],
    [90, 88, 92]
])

print("\nStudent 0, subject 1:", scores[0, 1])
print("All subjects of student 1:", scores[1, :])
print("Subject 2 for all students:", scores[:, 2])
````


# Sample 7 — Boolean masks

A Boolean mask is an array of `True` and `False` values.

It lets us filter only the values that match a condition.

Daily-life examples:

- find expenses greater than 100 baht,
- find scores that pass,
- find days where steps reach the target.


````python
import numpy as np

expenses = np.array([80, 250, 120, 40, 500, 60])

big_expense_mask = expenses >= 100
big_expenses = expenses[big_expense_mask]

print("expenses:", expenses)
print("mask:", big_expense_mask)
print("big expenses:", big_expenses)
print("number of big expenses:", big_expense_mask.sum())
print("total big expenses:", big_expenses.sum())

steps = np.array([5000, 7000, 6500, 8000, 9000, 4000, 3000])
goal_days = steps >= 7000

print("\nsteps on goal days:", steps[goal_days])
print("number of goal days:", goal_days.sum())
````


# Sample 8 — Aggregation and `axis`

Aggregation summarizes many values into fewer values.

Common aggregation methods:

| Method | Meaning |
|---|---|
| `.sum()` | total |
| `.mean()` | average |
| `.min()` | minimum |
| `.max()` | maximum |
| `.argmax()` | index of maximum value |
| `.argmin()` | index of minimum value |

## Axis idea for 2D arrays

For a table:

- `axis=0` summarizes **down the rows**, so the result is by column.
- `axis=1` summarizes **across the columns**, so the result is by row.

Daily-life example:

Rows = students.  
Columns = math, English, science.


````python
import numpy as np

scores = np.array([
    [80, 70, 90],
    [60, 85, 75],
    [90, 88, 92]
])

print("All scores average:", scores.mean())
print("Subject averages:", scores.mean(axis=0))  # each column
print("Student averages:", scores.mean(axis=1))  # each row

student_average = scores.mean(axis=1)
print("Best student index:", student_average.argmax())
````


# Sample 9 — Reshape and transpose

`reshape` changes the shape of an array without changing the values.

Important rule: the total number of values must stay the same.

Example:

```python
np.arange(1, 7).reshape(2, 3)
```

This has 6 values and becomes 2 rows × 3 columns.

`array.T` transposes a 2D array, meaning rows become columns and columns become rows.


````python
import numpy as np

numbers = np.arange(1, 7)
print("Original:")
print(numbers)

matrix = numbers.reshape(2, 3)
print("\nReshaped to 2 rows x 3 columns:")
print(matrix)

print("\nTranspose:")
print(matrix.T)
````


# Sample 10 — Dot product and weighted score

Dot product is useful for weighted score calculations.

Daily-life example:

| Component | Score | Weight |
|---|---:|---:|
| Homework | 80 | 0.30 |
| Midterm | 70 | 0.30 |
| Final | 90 | 0.40 |

Formula:

```text
final_score = homework * 0.30 + midterm * 0.30 + final * 0.40
```

In NumPy:

```python
final_score = scores @ weights
```

or

```python
final_score = np.dot(scores, weights)
```


````python
import numpy as np

component_scores = np.array([80, 70, 90])
weights = np.array([0.30, 0.30, 0.40])

final_score = component_scores @ weights
print("single student final score:", final_score)

# Multiple students
scores = np.array([
    [80, 70, 90],
    [60, 85, 75],
    [90, 88, 92]
])

final_scores = scores @ weights
print("multiple students final scores:", final_scores)
````

This sample combines many NumPy skills.

Rows = days.  
Columns = latte, americano, cocoa.

Prices:


# Unit 09 NumPy Cheat Sheet

## Import and create arrays

| Goal | Code pattern | Example |
|---|---|---|
| Import NumPy | `import numpy as np` | `import numpy as np` |
| Create array from list | `np.array(list_data)` | `np.array([1, 2, 3])` |
| Create zeros | `np.zeros(shape)` | `np.zeros((2, 3))` |
| Create integer zeros | `np.zeros(shape, dtype=int)` | `np.zeros((2, 3), dtype=int)` |
| Create ones | `np.ones(shape)` | `np.ones((2, 3))` |
| Create range | `np.arange(start, stop, step)` | `np.arange(5, 31, 5)` |
| Create equally spaced values | `np.linspace(start, stop, count)` | `np.linspace(0, 100, 5)` |
| Create identity matrix | `np.eye(n)` | `np.eye(3)` |

## Array information

| Goal | Code pattern | Meaning |
|---|---|---|
| Shape | `arr.shape` | Number of rows/columns or dimensions |
| Number of dimensions | `arr.ndim` | 1D, 2D, etc. |
| Total number of values | `arr.size` | Count of all elements |
| Data type | `arr.dtype` | Type of stored values |

## Element-wise operations and broadcasting

| Goal | Code pattern | Example |
|---|---|---|
| Add value to all elements | `arr + number` | `scores + 10` |
| Multiply all elements | `arr * number` | `prices * 1.07` |
| Add two arrays | `arr1 + arr2` | `score1 + score2` |
| Multiply two arrays | `arr1 * arr2` | `prices * quantities` |
| Add row bonus to 2D table | `table + row_array` | `scores + subject_bonus` |

## Indexing and slicing

| Goal | Code pattern | Example |
|---|---|---|
| First value | `arr[0]` | `steps[0]` |
| Last value | `arr[-1]` | `steps[-1]` |
| First three values | `arr[:3]` | `steps[:3]` |
| Index 2 to 4 | `arr[2:5]` | `steps[2:5]` |
| One value in 2D | `matrix[row, col]` | `scores[0, 1]` |
| One full row | `matrix[row, :]` | `scores[1, :]` |
| One full column | `matrix[:, col]` | `scores[:, 0]` |

## Boolean masks

| Goal | Code pattern | Example |
|---|---|---|
| Create condition mask | `mask = arr >= value` | `steps >= 7000` |
| Filter values | `arr[mask]` | `steps[steps >= 7000]` |
| Count `True` values | `mask.sum()` | `(steps >= 7000).sum()` |
| Multiple conditions | `(arr >= a) & (arr <= b)` | `(scores >= 50) & (scores <= 100)` |

## Aggregation and axis

| Goal | Code pattern | Example |
|---|---|---|
| Total | `arr.sum()` | `expenses.sum()` |
| Average | `arr.mean()` | `scores.mean()` |
| Minimum | `arr.min()` | `scores.min()` |
| Maximum | `arr.max()` | `scores.max()` |
| Index of maximum | `arr.argmax()` | `daily_total.argmax()` |
| Column totals | `table.sum(axis=0)` | `spending.sum(axis=0)` |
| Row totals | `table.sum(axis=1)` | `spending.sum(axis=1)` |
| Column averages | `table.mean(axis=0)` | `scores.mean(axis=0)` |
| Row averages | `table.mean(axis=1)` | `scores.mean(axis=1)` |

## Reshape, transpose, and weighted score

| Goal | Code pattern | Example |
|---|---|---|
| Reshape | `arr.reshape(rows, cols)` | `np.arange(1, 13).reshape(3, 4)` |
| Transpose | `matrix.T` | `seat_map.T` |
| Dot product | `arr1 @ arr2` | `scores @ weights` |
| Dot product alternative | `np.dot(arr1, arr2)` | `np.dot(scores, weights)` |

## Common beginner reminders

| Reminder | Example |
|---|---|
| Python indexes start at `0`. | The first element is `arr[0]`. |
| In slicing, the stop index is not included. | `arr[:3]` gets index `0`, `1`, `2`. |
| `axis=0` summarizes down rows, by column. | `scores.mean(axis=0)` gives subject averages. |
| `axis=1` summarizes across columns, by row. | `scores.mean(axis=1)` gives student averages. |
| `reshape` must keep the same total size. | 12 values can become `(3, 4)` or `(4, 3)`. |
| Boolean mask must have the same length as the array being filtered. | `arr[arr >= 10]` works. |
