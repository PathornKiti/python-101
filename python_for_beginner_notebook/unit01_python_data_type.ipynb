## 1. Basic data types

Unit 01 focuses on three core data types:

- `int`: whole numbers such as `-10`, `0`, `25`
- `float`: real numbers such as `10.0`, `3.14`, `1.23e5`
- `str`: text such as `"hello"` or `'hello'`

A few important reminders:

- `12` is an integer, but `12.0` is a floating-point number
- `"12"` is text, not a number
- scientific notation like `1e3` means `1000.0`
- Python does **not** allow an integer literal with a leading zero like `020`


````python
examples = [42, -10, 10.0, 1.23e5, "12", "Let's go shopping"]

for value in examples:
    print(f"value = {value!r:>20}   type = {type(value).__name__}")
````

Use `type(...)` when you want to inspect what Python is storing.


````python
a = 12
b = 12.0
c = "12"

print(type(a).__name__, a)
print(type(b).__name__, b)
print(type(c).__name__, c)
print("1e3 =", 1e3, "and its type is", type(1e3).__name__)
````

## 2. Variables and naming rules

A variable is a name that refers to a value.


- use letters, digits, or `_`
- do not start with a digit

## 3. Arithmetic operators

The main arithmetic operators in Unit 01 are:

- `+` addition
- `-` subtraction
- `*` multiplication
- `**` exponentiation
- `/` true division
- `//` floor division
- `%` remainder

When an integer interacts with a float, the result is usually a float.


````python
x = 11
y = 4

print("x + y  =", x + y)
print("x - y  =", x - y)
print("x * y  =", x * y)
print("x / y  =", x / y)
print("x // y =", x // y)
print("x % y  =", x % y)
print("x ** y =", x ** y)
print("2 + 1.0 =", 2 + 1.0, "with type", type(2 + 1.0).__name__)
````

## 4. Precedence and floor division

A reliable beginner memory rule is:

1. parentheses
2. exponentiation `**`
3. multiplication, division, floor division, remainder
4. addition and subtraction

Two classic Unit 01 examples are:

$$
3 + 97 // 2^3 \% 8
$$

and

$$
12 // 3 / 2 + 2^{3^2}
$$

Be especially careful with floor division on negative numbers.


````python
pairs = [(1, 2), (-1, 2), (11, 10), (-11, 10), (11, -10)]

for a, b in pairs:
    print(f"{a:>3} // {b:>3} = {a // b:>3}")

expr_1 = 3 + 97 // 2**3 % 8
expr_2 = 12 // 3 / 2 + 2**3**2

print("\n3 + 97 // 2**3 % 8 =", expr_1)
print("12 // 3 / 2 + 2**3**2 =", expr_2)
````

### Parentheses prevent mistakes

Compare these:

- `a / 2 * b` means `(a / 2) * b`
- `a / (2 * b)` means something different
- `x**1/3` means `(x**1)/3`, **not** the cube root of `x`
- for a cube root, write `x**(1/3)`


````python
a = 24
b = 3
x = 27

print("(a / 2) * b =", a / 2 * b)
print("a / (2 * b) =", a / (2 * b))
print("x**1/3      =", x**1/3)
print("x**(1/3)    =", x**(1/3))
````

## 5. Strings in expressions

Strings behave differently from numbers.

- `"12" + "23"` joins text
- `"12" * 3` repeats text
- `"12" + 3` causes an error


````python
print('"12" + "23" =', "12" + "23")
print('"12" * 3    =', "12" * 3)

try:
    print("12" + 3)
except TypeError as e:
    print("Error:", e)
````

If a line contains mixed types, read strings first and convert afterward.

Also, when values are separated by spaces, prefer `input().split()` instead of `input().split(" ")`.


````python
raw_line = "  alpha beta gamma   "
print("strip example ->", raw_line.strip())

line1 = "red blue green"
a, b, c = line1.split()
print("strings ->", a, b, c)

line2 = "10 20"
x, y = [int(e) for e in line2.split()]
print("ints ->", x, y, "sum =", x + y)

line3 = "1.5 2.5 3.5"
p, q, r = [float(e) for e in line3.split()]
print("floats ->", p, q, r, "average =", (p + q + r) / 3)

line4 = "2.75 4"
f, n = line4.split()
f = float(f)
n = int(n)
print("mixed ->", f, n)
````

## 10. Common beginner mistakes

Below are classic Unit 01 mistakes:

1. forgetting to convert input before calculating
2. forgetting `*` in multiplication
3. confusing operator precedence
4. forgetting `import math`
5. mismatching parentheses
6. mixing numbers and strings directly
7. reusing built-in names such as `int`
8. thinking `++k` or `--k` works like in C


````python
# Mistake 1: input-like text used as a number
try:
    x = "5"
    y = x**2 + 7
except TypeError as e:
    print("Mistake 1:", e)

# Mistake 2: mixing string and number in output
try:
    area = math.pi * 2**2
    print("area = " + area)
except TypeError as e:
    print("Mistake 2:", e)

# Mistake 3: using math without import (simulated)
try:
    namespace = {}
    exec("y = math.sqrt(9)", namespace)
except Exception as e:
    print("Mistake 3:", type(e).__name__ + ":", e)

# Mistake 4: reusing int
shadowed = {"int": 27}
exec("print('shadowed int =', int)", shadowed)

# Mistake 5: ++k does not increment in Python
k = 10
print("++k gives", ++k, "and k is still", k)

# Mistake 6: case sensitivity
count = 0
Count = count + 1
print("count =", count, "| Count =", Count)
````

## 11. Worked example: Triangle area from two sides and the included angle

Suppose a triangle has side lengths `a` and `b`, and the angle between them is `C` degrees.

The area formula is:

$$
\text{area} = \frac{1}{2}ab\sin(C)
$$

But in Python, `math.sin(...)` expects **radians**, not degrees.

So the safe workflow is:

1. read `a`, `b`, and `C_deg`
2. convert the angle to radians
3. compute the area
4. print the result in the required format


````python
import math

a = 10.0
b = 10.0
C_deg = 90.0

C_rad = math.radians(C_deg)
area = 0.5 * a * b * math.sin(C_rad)

print("area =", area, "(sq cm)")
````

### Why this example matters

This one example teaches many Unit 01 ideas at once:

- input conversion
- correct multiplication syntax
- importing `math`
- degree-to-radian conversion
- exact output formatting

## 13. Worked example: Triangle 2 with the law of cosines

If two sides `a` and `b` and the included angle `C` are known, the third side `c` is:

$$
c^2 = a^2 + b^2 - 2ab\cos(C)
$$

So:

$$
c = \sqrt{a^2 + b^2 - 2ab\cos(C)}
$$

Again, convert degrees to radians before calling `math.cos(...)`.


````python
import math

a = 3.0
b = 4.0
C_deg = 90.0
C_rad = math.radians(C_deg)

c = math.sqrt(a**2 + b**2 - 2*a*b*math.cos(C_rad))
print("c =", c, "cm.")
````

## 14. Worked example: ISBN-10 check digit

For a 10-digit ISBN with digits `n1` through `n10`, the checksum rule is

$$
10n_1 + 9n_2 + 8n_3 + 7n_4 + 6n_5 + 5n_6 + 4n_7 + 3n_8 + 2n_9 + n_{10}
$$

and that total must be divisible by `11`.

If the first 9 digits are known, you can compute the final digit.


````python
isbn9 = "020131452"

weighted_sum = sum((10 - i) * int(d) for i, d in enumerate(isbn9))
check_digit = (-weighted_sum) % 11

print("first 9 digits =", isbn9)
print("check digit    =", check_digit)
print("full ISBN      =", isbn9 + str(check_digit))
````
