## Step 2: Add spending level

Now classify the spending level.

Use this simple rule:

- spending is less than or equal to 70% of budget → low spending
- spending is more than 70% but not over budget → normal spending
- spending is over budget → over budget

```python
daily_budget = 150
spending_today = 120

if spending_today <= daily_budget * 0.70:
    print("Low spending. Good saving day.")
elif spending_today <= daily_budget:
    print("Normal spending. Still okay.")
else:
    print("Over budget. Be careful tomorrow.")
```

## Reference solution

Run this cell and enter your own spending.  
This solution does not use lists.

```python
print("Daily Allowance Checker")
print("-----------------------")

name = input("Enter your name: ")
daily_budget = float(input("Enter your daily budget: "))
number_of_days = int(input("How many days do you want to check? "))

while number_of_days <= 0:
    print("Number of days must be more than 0.")
    number_of_days = int(input("How many days do you want to check? "))

total_spending = 0
over_budget_days = 0
highest_spending = 0
highest_day = 0

for day in range(1, number_of_days + 1):
    spending = float(input("Enter spending for day " + str(day) + ": "))
    
    total_spending = total_spending + spending
    
    if spending > daily_budget:
        over_budget_days = over_budget_days + 1
    
    if spending > highest_spending:
        highest_spending = spending
        highest_day = day

average_spending = total_spending / number_of_days
total_budget = daily_budget * number_of_days
money_left = total_budget - total_spending

print()
print("Summary for", name)
print("Total budget:", total_budget)
print("Total spending:", total_spending)
print("Money left:", money_left)
print("Average spending per day:", average_spending)
print("Over-budget days:", over_budget_days)
print("Highest spending:", highest_spending)
print("Highest spending day:", highest_day)

print()
print("Final advice:")

if average_spending <= daily_budget * 0.70 and over_budget_days == 0:
    print("Great job. Your spending is very safe.")
elif average_spending <= daily_budget and over_budget_days <= 1:
    print("Your spending is okay, but keep watching it.")
else:
    print("Your spending is risky. Try to reduce small daily costs.")
```
