## 1. Basic repetition patterns

Unit 03 begins with the main repetition shapes:

```python
while True:
    P1
    if stop_condition:
        break
    P2
```

```python
while condition:
    P
```

```python
for k in range(n):
    P
```

```python
for p in range(m):
    for k in range(n):
        P
```

Choose:
- `while` when the stopping condition depends on what happens during the loop
- `for` when you already know the pattern of values

## 2. A `while` loop must change its condition

A very important Unit 03 rule:

Inside a `while` loop, there should be some statement that helps make the condition become false later.

Example:

```python
i = 0
while i < 10:
    ...
    i += 1
```

Without a change like `i += 1`, the loop may never end.

````python
i = 0
history = []

while i < 8:
    history.append(i)
    i += 2

print(history)
assert history == [0, 2, 4, 6]
````

## 3. `for` loops and `range(...)`

The most common forms are:

```python
range(n)              # 0, 1, 2, ..., n-1
range(start, stop)    # start, ..., stop-1
range(start, stop, step)
```

Examples:
- `range(5)` gives `0, 1, 2, 3, 4`
- `range(2, 7)` gives `2, 3, 4, 5, 6`
- `range(10, 1, -3)` gives `10, 7, 4`

````python
print(list(range(5)))
print(list(range(2, 7)))
print(list(range(10, 1, -3)))

total = 0
for k in range(1, 6):
    total += k
print("sum 1..5 =", total)

assert list(range(5)) == [0, 1, 2, 3, 4]
assert list(range(2, 7)) == [2, 3, 4, 5, 6]
assert list(range(10, 1, -3)) == [10, 7, 4]
assert total == 15
````

## 4. `for-else` and `while-else`

Python allows `else` on loops.

The `else` block runs only when the loop finishes **normally**.
It does **not** run if the loop stops because of `break`.

This is useful for search problems:
- if you `break`, you found the target
- if the loop ends naturally, the target was not found

````python
def first_empty_locker(lockers):
    for i in range(len(lockers)):
        if lockers[i] == "empty":
            return i
    else:
        return None


print("first empty locker =", first_empty_locker(["full", "full", "empty", "full"]))
print("first empty locker =", first_empty_locker(["full", "full", "full"]))


def weeks_until_savings_target(target, weekly_saving):
    week = 0
    total = 0

    while total < target:
        week += 1
        total += weekly_saving
    else:
        return week


print("weeks until 1,000 baht =", weeks_until_savings_target(1000, 150))

assert first_empty_locker(["full", "full", "empty", "full"]) == 2
assert first_empty_locker(["full", "full", "full"]) is None
assert weeks_until_savings_target(1000, 150) == 7
````

## 5. Nested loops

A nested loop means one loop is inside another loop.

Typical uses:
- coordinate pairs
- small tables
- checking all combinations
- double sums
- searching over two or more variables

````python
table = []
for i in range(1, 4):
    row = []
    for j in range(1, 5):
        row.append(i * j)
    table.append(row)

for row in table:
    print(row)

assert table == [
    [1, 2, 3, 4],
    [2, 4, 6, 8],
    [3, 6, 9, 12],
]
````

## 6. Common loop patterns

Unit 03 repeatedly uses these ideas:

- **accumulate** a sum
- **count** how many values satisfy a condition
- **search** until something is found
- **track min/max**
- **sentinel loop**: continue until a special value appears

These patterns appear again and again in later chapters.

````python
values = [8, -3, 12, 0, -10, 5]

# accumulate
s = 0
for x in values:
    s += x

# count
neg = 0
for x in values:
    if x < 0:
        neg += 1

# min/max
mn = values[0]
mx = values[0]
for x in values:
    if x < mn:
        mn = x
    if x > mx:
        mx = x

print("sum =", s)
print("negative count =", neg)
print("min =", mn, "max =", mx)

assert s == 12
assert neg == 2
assert mn == -10
assert mx == 12
````

## 7. Common mistakes

Common Unit 03 mistakes include:

- forgetting to change the `while` condition
- off-by-one errors in `range(...)`
- using `break` when you actually want to continue the search differently
- mixing up `<` and `<=`
- resetting an accumulator inside the loop by mistake
- putting `else` under `if` when you meant loop-`else`

Another frequent issue:
the variable in a `for` loop is controlled by `range(...)`, not by manual reassignment.

````python
# Good: accumulator starts once before the loop.
values = [1, 2, 3, 4]
s = 0
for x in values:
    s += x
print("correct sum =", s)

# Bad pattern shown safely:
partials = []
for x in values:
    wrong_s = 0   # reset every round: this is usually a mistake
    wrong_s += x
    partials.append(wrong_s)
print("reset-every-round pattern =", partials)

assert s == 10
assert partials == [1, 2, 3, 4]
````

## 8. Changing the loop variable inside `for`

Another specific Unit 03 point:

If you write

```python
for k in range(...):
    ...
    k += 2
```

the `k += 2` does **not** change the next value that comes from `range(...)`.

If you truly want full control of the next value, use `while`.

````python
for_result = []
for k in range(1, 4):
    for_result.append(k)
    k += 2
    for_result.append(k)

while_result = []
k = 1
while k < 4:
    while_result.append(k)
    k += 2
    while_result.append(k)

print("for  :", for_result)
print("while:", while_result)

assert for_result == [1, 3, 2, 4, 3, 5]
assert while_result == [1, 3, 3, 5]
````

## 8.1 Seamless conversion between `for` and `while`

A strong Unit 03 student should be able to translate a **counting loop** back and forth between `for` and `while` without changing the result.

### Core pattern: `for` to `while`

A loop of the form

```python
for k in range(start, stop, step):
    BODY
```

can be rewritten as

```python
k = start

if step > 0:
    while k < stop:
        BODY
        k += step
else:
    while k > stop:
        BODY
        k += step
```

### Core pattern: `while` to `for`

A loop of the form

```python
k = start
while k < stop:
    BODY
    k += 1
```

can usually become

```python
for k in range(start, stop):
    BODY
```

and similarly for other fixed steps.

### When conversion is natural

- Use **`for`** when the loop follows a clear counting pattern.
- Use **`while`** when the number of repetitions is not known in advance.
- A **sentinel loop** such as “read until `-1`” is usually more natural with `while`.
- A loop over a fixed numeric progression is usually more natural with `for`.

### Conversion checklist

When you rewrite one style into the other, keep these parts identical:

1. **initial value**
2. **stopping rule**
3. **update rule**
4. **body of work**
5. **final result**


````python
def sum_range_for(a: int, b: int) -> int:
    s = 0
    for k in range(a, b + 1):
        s += k
    return s

def sum_range_while(a: int, b: int) -> int:
    s = 0
    k = a
    while k <= b:
        s += k
        k += 1
    return s

def stepped_values_for(start: int, stop: int, step: int):
    out = []
    for k in range(start, stop, step):
        out.append(k)
    return out

def stepped_values_while(start: int, stop: int, step: int):
    out = []
    k = start

    if step > 0:
        while k < stop:
            out.append(k)
            k += step
    else:
        while k > stop:
            out.append(k)
            k += step
    return out

def first_multiple_of_17_for(a: int, b: int):
    for k in range(a, b + 1):
        if k % 17 == 0:
            return k
    return None

def first_multiple_of_17_while(a: int, b: int):
    k = a
    while k <= b:
        if k % 17 == 0:
            return k
        k += 1
    return None

print(sum_range_for(3, 7), sum_range_while(3, 7))
print(stepped_values_for(2, 15, 4))
print(stepped_values_while(2, 15, 4))
print(first_multiple_of_17_for(30, 50), first_multiple_of_17_while(30, 50))

assert sum_range_for(3, 7) == sum_range_while(3, 7) == 25
assert stepped_values_for(2, 15, 4) == stepped_values_while(2, 15, 4) == [2, 6, 10, 14]
assert stepped_values_for(20, 3, -5) == stepped_values_while(20, 3, -5) == [20, 15, 10, 5]
assert first_multiple_of_17_for(30, 50) == first_multiple_of_17_while(30, 50) == 34
assert first_multiple_of_17_for(35, 50) == first_multiple_of_17_while(35, 50) == None
````

### 9.1 Saving money until a target

You save the same amount every week.

**Goal:** count how many weeks are needed until the saved money reaches the target.

Formula idea:

$$
\text{total saved} = \text{total saved} + \text{weekly saving}
$$

Stop when:

$$
\text{total saved} \ge \text{target}
$$


````python
def weeks_to_reach_target(target, weekly_saving):
    week = 0
    total = 0

    while total < target:
        week += 1
        total += weekly_saving

    return week, total


weeks, total = weeks_to_reach_target(1000, 150)
print("weeks =", weeks)
print("total saved =", total)

assert weeks_to_reach_target(1000, 150) == (7, 1050)
assert weeks_to_reach_target(500, 100) == (5, 500)
````


### 9.2 Total and average daily spending

You record daily spending for several days.

**Goal:** compute total spending and average spending.

Formula idea:

$$
\text{total} = x_1 + x_2 + \cdots + x_n
$$

$$
\text{average} = \frac{\text{total}}{n}
$$


````python
def total_and_average_spending(spending):
    total = 0

    for amount in spending:
        total += amount

    average = total / len(spending)
    return total, average


sample_spending = [120, 80, 150, 90, 60]
print(total_and_average_spending(sample_spending))

assert total_and_average_spending(sample_spending) == (500, 100)
````


### 9.3 Count days that meet a step goal

You have a list of daily step counts.

**Goal:** count how many days reached at least the step goal.

Counting idea:

$$
\text{count} = \text{count} + 1
$$

only when:

$$
\text{steps} \ge \text{goal}
$$


````python
def count_step_goal_days(step_counts, goal):
    count = 0

    for steps in step_counts:
        if steps >= goal:
            count += 1

    return count


steps = [6500, 10000, 8500, 12000, 4000, 11000]
print(count_step_goal_days(steps, 10000))

assert count_step_goal_days(steps, 10000) == 3
````


### 9.4 Average quiz score until a stop value

A teacher enters quiz scores one by one.

The stop value is `-1`.

**Goal:** calculate the average of all scores before `-1`.

Formula idea:

$$
\text{average} = \frac{\text{sum of scores}}{\text{number of scores}}
$$

If no score was entered before `-1`, return `"No Data"`.


````python
def average_until_stop(scores):
    total = 0
    count = 0
    i = 0

    while i < len(scores):
        score = scores[i]

        if score == -1:
            break

        total += score
        count += 1
        i += 1

    if count == 0:
        return "No Data"

    return total / count


print(average_until_stop([80, 90, 70, -1, 100]))
print(average_until_stop([-1, 80, 90]))

assert average_until_stop([80, 90, 70, -1, 100]) == 80
assert average_until_stop([-1, 80, 90]) == "No Data"
````


### 9.5 Shopping within a budget

You buy items in order from a shopping list.

**Goal:** stop buying when the next item would exceed the budget.

Repeated update idea:

$$
\text{spent} = \text{spent} + \text{item price}
$$

Stop when:

$$
\text{spent} + \text{item price} > \text{budget}
$$


````python
def shop_until_budget(prices, budget):
    spent = 0
    bought = 0

    for price in prices:
        if spent + price > budget:
            break

        spent += price
        bought += 1

    remaining = budget - spent
    return bought, spent, remaining


prices = [35, 50, 120, 25, 80]
print(shop_until_budget(prices, 200))

assert shop_until_budget([35, 50, 90, 25, 80], 200) == (4, 200, 0)
assert shop_until_budget([35, 50, 120, 25, 80], 200) == (2, 85, 115)
````


### 9.6 Weekly study tracker with nested loops

A student records study hours for several weeks.

Each week has several days.

**Goal:** use nested loops to calculate:

- total study hours
- number of days with at least 2 study hours

Formula idea:

$$
\text{total hours} = \text{total hours} + \text{hours for one day}
$$

For nested loops:

- outer loop = week
- inner loop = day inside that week


````python
def study_tracker(weekly_hours):
    total_hours = 0
    strong_days = 0

    for week in weekly_hours:
        for hours in week:
            total_hours += hours

            if hours >= 2:
                strong_days += 1

    return total_hours, strong_days


hours = [
    [1, 2, 0, 3, 2],
    [2, 2, 1, 0, 4],
]
print(study_tracker(hours))

assert study_tracker(hours) == (17, 6)
````


## 10. Worked example: phone battery simulation

A phone starts with a battery percentage.

Every hour, the battery decreases by the same percentage.

**Goal:** count how many hours until the battery becomes less than or equal to the warning level.

Formula idea:

$$
\text{battery} = \text{battery} - \text{drain per hour}
$$

Stop when:

$$
\text{battery} \le \text{warning level}
$$


````python
def hours_until_low_battery(start_battery, drain_per_hour, warning_level):
    hours = 0
    battery = start_battery

    while battery > warning_level:
        hours += 1
        battery -= drain_per_hour

    return hours, battery


print(hours_until_low_battery(80, 12, 20))

assert hours_until_low_battery(80, 12, 20) == (5, 20)
assert hours_until_low_battery(75, 10, 20) == (6, 15)
````


## 13. Unit 03 Repetition Cheat Sheet

| Concept | Python pattern | Use when | Daily-life example | Common mistake |
|---|---|---|---|---|
| `while condition` | `while condition:` | You repeat until a condition becomes false. | Save money until reaching a target. | Forgetting to update the variable inside the loop. |
| `while True` + `break` | `while True:` then `if stop: break` | You do not know when the input will stop. | Enter spending until `0`. | Forgetting the `break`, causing an infinite loop. |
| `for k in range(n)` | `for k in range(n):` | You know exactly how many times to repeat. | Add transport cost for `n` days. | Thinking `range(n)` includes `n`; it stops at `n - 1`. |
| `range(start, stop)` | `range(a, b)` | You want values from `a` to `b - 1`. | Week numbers before the final week. | Forgetting that `stop` is excluded. |
| `range(start, stop, step)` | `range(a, b, step)` | You want to skip by a fixed amount. | Count every 2nd day. | Using the wrong sign for `step`. |
| Accumulator | `total = total + x` | You need a total. | Total daily spending. | Resetting `total = 0` inside the loop. |
| Counter | `count = count + 1` | You need to count how many times something happens. | Count days with at least 10,000 steps. | Counting every item instead of only matching items. |
| Search loop | `if found: break` | You want the first matching item. | First rainy day. | Continuing after already finding the answer. |
| `for-else` | `for ...: ... break` / `else:` | You want to handle “not found”. | No rainy day in the forecast. | Thinking loop `else` means the same as `if-else`. |
| Nested loops | Loop inside another loop. | Data has two dimensions. | Weeks and days. | Updating the wrong variable in the wrong loop. |
| Convert `for` to `while` | initialize → condition → update | You want to understand loop mechanics. | Weekly allowance from week `a` to `b`. | Forgetting the update step. |
| Off-by-one check | Check first and last repeated value. | Any loop using `range`. | Day 1 to Day 7. | Printing one too many or one too few values. |

### Quick loop templates

**Template 1: Accumulate total**

```python
total = 0

    total += x
```

**Template 2: Count matching values**

```python

        count += 1
```

**Template 3: Search for first match**

    # body
