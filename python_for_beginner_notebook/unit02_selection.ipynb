## 1. The basic selection patterns

Unit 02 starts with four core shapes:

```python
if C1:
    P1
```

```python
if C1:
    pass
else:
    P1
```

```python
if not C1:
    P1
```

```python
if C1:
    P1
else:
    P2
```

```python
if C1:
    P1
elif C2:
    P2
elif C3:
    P3
else:
    P4
```

A condition should evaluate to `True` or `False`.


````python

x = 7

if x > 0:
    print("positive")

if x % 2 == 0:
    print("even")
else:
    print("odd")

score = 74
if score >= 80:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "F"

print("grade =", grade)
````

## 3. Boolean operators: `and`, `or`, `not`

Selection in Unit 02 depends heavily on combining conditions.

- `A and B` is true only when both are true
- `A or B` is true when at least one is true
- `not A` reverses a truth value


````python

x = 6
y = 3

print("x > 0 and y > 0  ->", x > 0 and y > 0)
print("x < 0 or y > 0   ->", x < 0 or y > 0)
print("not(x == 6)      ->", not(x == 6))
````

## 4. Equivalent conditions you should recognize

Unit 02 expects you to see that these mean the same thing:

```python
not(x == 0)        ==        x != 0
not(x == 2 or x == 4) ==     x != 2 and x != 4
not(x < 2 and y >= 4) ==     x >= 2 or y < 4
3 <= x and x < 9   ==        3 <= x < 9
a < b and b < c and c < d and d <= e  ==  a < b < c < d <= e
c == 'a' or c == 'e' or c == 'i' or c == 'o' or c == 'u'
==
c in ('a', 'e', 'i', 'o', 'u')
==
c in 'aeiou'
```

The shorter form is often clearer.


````python

x = 5
y = 1
c = "e"

print(not(x == 0), x != 0)
print(not(x == 2 or x == 4), x != 2 and x != 4)
print(not(x < 2 and y >= 4), x >= 2 or y < 4)

x = 7
print(3 <= x and x < 9, 3 <= x < 9)

print(c == "a" or c == "e" or c == "i" or c == "o" or c == "u")
print(c in ("a", "e", "i", "o", "u"))
print(c in "aeiou")
````

## 7. String comparison in conditions

A Unit 02 warning is that strings compare **lexicographically** (dictionary order), not numerically.

Examples:
- `'0' < '9' < 'A' < 'Z' < 'a' < 'z'`
- `'abcdegh' < 'b'` is `True`
- `'1234' < '9'` is also `True`

That last one surprises many beginners.

Why? Because Python compares strings from left to right:
- `'1'` is smaller than `'9'`
- so `'1234' < '9'` is true


````python

x = "1234"
y = "9"

print(x < y)   # string comparison

x_num = 1234
y_num = 9
print(x_num < y_num)  # numeric comparison
````

## 8. Cleaner conditional code

Unit 02 also highlights two good code-improvement ideas.

### Idea A: move common statements out

Instead of writing the same line in both branches:

```python
if d > 0:
    a = 9
    c += d - 5
    e = c
else:
    a = 9
    c -= d + 7
    e = c
```

you can write:

```python
a = 9
if d > 0:
    c += d - 5
else:
    c -= d + 7
e = c
```

### Idea B: reduce repeated comparisons in ranges

Instead of:

```python
if s >= 80:
    g = 'A'
elif 70 <= s < 80:
    g = 'B'
elif 60 <= s < 70:
    g = 'C'
elif 50 <= s < 60:
    g = 'D'
else:
    g = 'F'
```

you can write:

```python
if s >= 80:
    g = 'A'
elif s >= 70:
    g = 'B'
elif s >= 60:
    g = 'C'
elif s >= 50:
    g = 'D'
else:
    g = 'F'
```

This works because earlier branches already ruled out the larger values.


````python

def grade_v1(s):
    if s >= 80:
        return "A"
    elif 70 <= s < 80:
        return "B"
    elif 60 <= s < 70:
        return "C"
    elif 50 <= s < 60:
        return "D"
    else:
        return "F"

def grade_v2(s):
    if s >= 80:
        return "A"
    elif s >= 70:
        return "B"
    elif s >= 60:
        return "C"
    elif s >= 50:
        return "D"
    else:
        return "F"

for s in [49, 50, 61, 70, 80, 93]:
    print(s, grade_v1(s), grade_v2(s))
````

## 10. Worked example: median of 3 integers

A clean strategy is to check which number lies in the middle.

If `a` is between `b` and `c`, then `a` is the median.  
If `b` is between `a` and `c`, then `b` is the median.  
Otherwise `c` must be the median.


````python

def median_of_three(a, b, c):
    if (b <= a <= c) or (c <= a <= b):
        return a
    elif (a <= b <= c) or (c <= b <= a):
        return b
    else:
        return c

tests = [
    (3, 8, 5),
    (10, 1, 7),
    (4, 4, 9),
    (-2, -5, -3),
]

for t in tests:
    print(t, "->", median_of_three(*t))
````

## 11. Worked example: quadrant / axis / origin

For a point `(x, y)`:

- `(0, 0)` is the origin
- `y == 0` means on the x-axis
- `x == 0` means on the y-axis
- otherwise determine the quadrant from the signs of `x` and `y`


````python

def locate_point(x, y):
    if x == 0 and y == 0:
        return "origin"
    elif y == 0:
        return "x-axis"
    elif x == 0:
        return "y-axis"
    elif x > 0 and y > 0:
        return "Q1"
    elif x < 0 and y > 0:
        return "Q2"
    elif x < 0 and y < 0:
        return "Q3"
    else:
        return "Q4"

for point in [(0, 0), (3, 0), (0, -2), (4, 7), (-5, 8), (-1, -9), (2, -3)]:
    print(point, "->", locate_point(*point))
````

## 12. Worked example: two circles

Suppose each circle is given by `(x, y, r)`.

Let `d` be the distance between centers.

- if `d == r1 + r2`, the circles **touch**
- if `d < r1 + r2`, they **overlap**
- if `d > r1 + r2`, they are **free**

For beginner Unit 02 style problems, this outside-touch/overlap/free classification is enough.


````python

import math

def circle_relation(x1, y1, r1, x2, y2, r2):
    d = math.hypot(x1 - x2, y1 - y2)
    if math.isclose(d, r1 + r2):
        return "touch"
    elif d < r1 + r2:
        return "overlap"
    else:
        return "free"

examples = [
    (0, 0, 2, 4, 0, 2),
    (0, 0, 3, 2, 0, 2),
    (0, 0, 1, 5, 0, 1),
]

for e in examples:
    print(e, "->", circle_relation(*e))
````

## 13. Worked example: special lottery range problem

This is the big worked example in Unit 02.

You buy every ticket from number `n1` through `n2` inclusive.
The draw results are:

- first prize: `p1` worth 10,000
- last 2 digits: `p2` worth 25
- last 3 digits: `p3` worth 100

The important bug to avoid is **double-counting** when `n1` and `n2`
sit inside the same 100-block or the same 1000-block.

A robust way is:

1. find the **first** number in `[n1, n2]` whose suffix matches
2. if that first number is outside the interval, the count is `0`
3. otherwise count how many times we can add `100` or `1000`


````python
def count_suffix_matches(n1, n2, suffix, mod):
    """Count how many integers in [n1, n2] end with the given suffix.

    mod = 100  -> compare last 2 digits
    mod = 1000 -> compare last 3 digits
    """
    first = n1 + ((suffix - n1) % mod)
    if first > n2:
        return 0
    return 1 + (n2 - first) // mod


def lottery_prize_total(p1, p2, p3, n1, n2):
    s = 0

    if n1 <= p1 <= n2:
        s += 10000

    s += 25 * count_suffix_matches(n1, n2, p2, 100)
    s += 100 * count_suffix_matches(n1, n2, p3, 1000)

    return s


samples = [
    (1234, 11, 811, 1000, 1250),
    (99999, 99, 999, 99950, 99999),
    (19999, 13, 1, 9015, 13000),
]

for row in samples:
    print(row, "->", lottery_prize_total(*row))
````

### Why the previous lottery version failed

The previous notebook version counted the left boundary and right boundary
with separate `if` statements.

That can accidentally count the **same matching block twice**.

Example:
- `n1 = 1000`
- `n2 = 1250`
- `p2 = 11`

There is only one ticket ending in `11` inside the range: `1011`.

But these two conditions are both true:

```python
n1 % 100 <= p2
p2 <= n2 % 100
```

so the old code added the 2-digit prize **twice**.


## 14. Worked example: Days in Month

The book uses **B.E.** year input, so convert to **C.E.** first:

$$
\text{CE} = \text{BE} - 543
$$

February has 29 days exactly when the C.E. year is a leap year:

- divisible by 4 and not divisible by 100, or
- divisible by 400


````python

def days_in_month(month, year_be):
    year_ce = year_be - 543

    if month in (1, 3, 5, 7, 8, 10, 12):
        return 31
    elif month in (4, 6, 9, 11):
        return 30
    elif month == 2:
        leap = (year_ce % 4 == 0 and year_ce % 100 != 0) or (year_ce % 400 == 0)
        if leap:
            return 29
        else:
            return 28
    else:
        return None

for month, year_be in [(10, 2557), (2, 2557), (4, 2556), (2, 2547)]:
    print(month, year_be, "->", days_in_month(month, year_be))
````

## 16. Worked example: grade cutting

This version validates the score range **first**.

That matters because `if` / `elif` stops at the first true condition.
So a value like `120` must be rejected before checking the grade bands.


````python
def cut_grade(x):
    if x < 0 or x > 100:
        return "ERROR"
    elif x >= 80:
        return "A"
    elif x >= 75:
        return "B+"
    elif x >= 70:
        return "B"
    elif x >= 65:
        return "C+"
    elif x >= 60:
        return "C"
    elif x >= 55:
        return "D+"
    elif x >= 50:
        return "D"
    else:
        return "F"

for x in [87.25, 69.95, 120, -3, 50, 75]:
    print(x, "->", cut_grade(x))
````

## 16. Worked example: parking fee

Rules:

1. at most 15 minutes -> free
2. over 15 minutes but not more than 3 hours -> 10 baht per started hour
3. from hour 4 through hour 6 -> 20 baht per started hour
4. over 6 hours -> flat 200 baht

A convenient method is:

- convert entry and exit times to minutes
- compute total parked minutes
- convert to **started hours** when needed


````python

import math

def parking_fee(h_in, m_in, h_out, m_out):
    start = h_in * 60 + m_in
    end = h_out * 60 + m_out
    parked = end - start

    if parked <= 15:
        return 0

    hours_started = math.ceil(parked / 60)

    if hours_started <= 3:
        return hours_started * 10
    elif hours_started <= 6:
        return 3 * 10 + (hours_started - 3) * 20
    else:
        return 200

examples = [
    (7, 0, 7, 15),
    (7, 0, 7, 16),
    (7, 30, 10, 30),
    (7, 30, 10, 31),
    (7, 30, 13, 31),
]

for ex in examples:
    print(ex, "->", parking_fee(*ex))
````
