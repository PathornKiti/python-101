## 1. The basic selection patterns

Unit 02 starts with four core shapes:

```python
if C1:
    P1
```

```python
if C1:
    pass
else:
    P1
```

```python
if not C1:
    P1
```

```python
if C1:
    P1
else:
    P2
```

```python
if C1:
    P1
elif C2:
    P2
elif C3:
    P3
else:
    P4
```

A condition should evaluate to `True` or `False`.

## 2. Comparison operators

These are the main comparison operators you use inside conditions:

- `==` equal to
- `!=` not equal to
- `<`
- `<=`
- `>`
- `>=`

A comparison produces a Boolean value.


````python

a = 5
b = 9

print(a == b)
print(a != b)
print(a < b)
print(a <= b)
print(a > b)
print(a >= b)
````

## 3. Boolean operators: `and`, `or`, `not`

Selection in Unit 02 depends heavily on combining conditions.

- `A and B` is true only when both are true
- `A or B` is true when at least one is true
- `not A` reverses a truth value


````python

x = 6
y = 3

print("x > 0 and y > 0  ->", x > 0 and y > 0)
print("x < 0 or y > 0   ->", x < 0 or y > 0)
print("not(x == 6)      ->", not(x == 6))
````

## 4. Equivalent conditions you should recognize

Unit 02 expects you to see that these mean the same thing:

```python
not(x == 0)        ==        x != 0
not(x == 2 or x == 4) ==     x != 2 and x != 4
not(x < 2 and y >= 4) ==     x >= 2 or y < 4
3 <= x and x < 9   ==        3 <= x < 9
a < b and b < c and c < d and d <= e  ==  a < b < c < d <= e
c == 'a' or c == 'e' or c == 'i' or c == 'o' or c == 'u'
==
c in ('a', 'e', 'i', 'o', 'u')
==
c in 'aeiou'
```

The shorter form is often clearer.


````python

x = 5
y = 1
c = "e"

print(not(x == 0), x != 0)
print(not(x == 2 or x == 4), x != 2 and x != 4)
print(not(x < 2 and y >= 4), x >= 2 or y < 4)

x = 7
print(3 <= x and x < 9, 3 <= x < 9)

print(c == "a" or c == "e" or c == "i" or c == "o" or c == "u")
print(c in ("a", "e", "i", "o", "u"))
print(c in "aeiou")
````

## 5. Nested `if`

A nested `if` is an `if` placed inside another branch.

Use nesting when the second decision only makes sense after the first one.


````python

x = 12

if x >= 0:
    print("non-negative")
    if x % 2 == 0:
        print("and even")
    else:
        print("and odd")
else:
    print("negative")
````

## 7. String comparison in conditions

A Unit 02 warning is that strings compare **lexicographically** (dictionary order), not numerically.

Examples:
- `'0' < '9' < 'A' < 'Z' < 'a' < 'z'`
- `'abcdegh' < 'b'` is `True`
- `'1234' < '9'` is also `True`

That last one surprises many beginners.

Why? Because Python compares strings from left to right:
- `'1'` is smaller than `'9'`
- so `'1234' < '9'` is true


````python

x = "1234"
y = "9"

print(x < y)   # string comparison

x_num = 1234
y_num = 9
print(x_num < y_num)  # numeric comparison
````

## 8. Cleaner conditional code

Unit 02 also highlights two good code-improvement ideas.

### Idea A: move common statements out

Instead of writing the same line in both branches:

```python
if d > 0:
    a = 9
    c += d - 5
    e = c
else:
    a = 9
    c -= d + 7
    e = c
```

you can write:

```python
a = 9
if d > 0:
    c += d - 5
else:
    c -= d + 7
e = c
```

### Idea B: reduce repeated comparisons in ranges

Instead of:

```python
if s >= 80:
    g = 'A'
elif 70 <= s < 80:
    g = 'B'
elif 60 <= s < 70:
    g = 'C'
elif 50 <= s < 60:
    g = 'D'
else:
    g = 'F'
```

you can write:

```python
if s >= 80:
    g = 'A'
elif s >= 70:
    g = 'B'
elif s >= 60:
    g = 'C'
elif s >= 50:
    g = 'D'
else:
    g = 'F'
```

This works because earlier branches already ruled out the larger values.


````python

def grade_v1(s):
    if s >= 80:
        return "A"
    elif 70 <= s < 80:
        return "B"
    elif 60 <= s < 70:
        return "C"
    elif 50 <= s < 60:
        return "D"
    else:
        return "F"

def grade_v2(s):
    if s >= 80:
        return "A"
    elif s >= 70:
        return "B"
    elif s >= 60:
        return "C"
    elif s >= 50:
        return "D"
    else:
        return "F"

for s in [49, 50, 61, 70, 80, 93]:
    print(s, grade_v1(s), grade_v2(s))
````

## 9. Common Unit 02 thinking patterns

A lot of Unit 02 problems follow one of these beginner-friendly patterns:

1. **Validate input first**  
   Example: a score must be between `0` and `100`.

2. **Classify a value into a range**  
   Example: grade cut, shirt size, ticket price, delivery fee.

3. **Use combined conditions**  
   Example: free delivery if `order_amount >= 300` and `distance <= 5`.

4. **Check special cases first**  
   Example: invalid input, free ticket, senior discount, free parking.

5. **Compare multiple values**  
   Example: find the middle value from three commute times.

6. **Compute first, then classify**  
   Example: calculate parking duration first, then decide the parking fee.


## 10. Worked example: median of 3 integers

A clean strategy is to check which number lies in the middle.

If `a` is between `b` and `c`, then `a` is the median.  
If `b` is between `a` and `c`, then `b` is the median.  
Otherwise `c` must be the median.


````python

def median_of_three(a, b, c):
    if (b <= a <= c) or (c <= a <= b):
        return a
    elif (a <= b <= c) or (c <= b <= a):
        return b
    else:
        return c

tests = [
    (3, 8, 5),
    (10, 1, 7),
    (4, 4, 9),
    (-2, -5, -3),
]

for t in tests:
    print(t, "->", median_of_three(*t))
````

## 11. Worked example: phone battery action

For a phone battery reminder, we can combine numeric conditions and text conditions.

Rules:

- battery percentage must be from `0` to `100`
- charging status must be `yes` or `no`
- if battery is `20` or below and not charging -> `Charge now`
- if battery is `20` or below and charging -> `Charging`
- if battery is `80` or above and charging -> `Unplug soon`
- otherwise -> `OK`

This example is useful because it shows input validation and `and`.


````python
def battery_action(battery, charging):
    charging = charging.lower()

    if battery < 0 or battery > 100:
        return "ERROR"
    elif charging not in ("yes", "no"):
        return "ERROR"
    elif battery <= 20 and charging == "no":
        return "Charge now"
    elif battery <= 20 and charging == "yes":
        return "Charging"
    elif battery >= 80 and charging == "yes":
        return "Unplug soon"
    else:
        return "OK"

examples = [
    (15, "no"),
    (15, "yes"),
    (85, "yes"),
    (60, "no"),
    (120, "no"),
]

for ex in examples:
    print(ex, "->", battery_action(*ex))
````

## 12. Worked example: movie ticket price

Suppose a cinema uses these ticket rules:

- age below `0` is invalid -> `ERROR`
- age below `7` -> free ticket, price `0`
- age `60` or above -> senior ticket, price `80`
- student with status `yes` -> student ticket, price `100`
- otherwise -> regular ticket, price `150`

Order matters. We should check invalid input and special discounts before the regular price.


````python
def movie_ticket_price(age, student):
    student = student.lower()

    if age < 0:
        return "ERROR"
    elif student not in ("yes", "no"):
        return "ERROR"
    elif age < 7:
        return 0
    elif age >= 60:
        return 80
    elif student == "yes":
        return 100
    else:
        return 150

examples = [
    (5, "no"),
    (16, "yes"),
    (35, "no"),
    (65, "no"),
    (-1, "yes"),
]

for ex in examples:
    print(ex, "->", movie_ticket_price(*ex))
````

## 13. Worked example: delivery fee

A food delivery app calculates delivery fee from distance and order amount.

Rules:

- distance and order amount cannot be negative
- if order amount is at least `300` and distance is at most `5` km -> free delivery
- otherwise, distance at most `3` km -> `20` baht
- otherwise, distance at most `10` km -> `40` baht
- otherwise -> `80` baht

This is a common `if` / `elif` problem because the first true rule is selected.


````python
def delivery_fee(distance_km, order_amount):
    if distance_km < 0 or order_amount < 0:
        return "ERROR"
    elif order_amount >= 300 and distance_km <= 5:
        return 0
    elif distance_km <= 3:
        return 20
    elif distance_km <= 10:
        return 40
    else:
        return 80

examples = [
    (2.5, 120),
    (4.0, 350),
    (7.5, 250),
    (12.0, 500),
    (-1.0, 100),
]

for ex in examples:
    print(ex, "->", delivery_fee(*ex))
````

### Why condition order matters

In an `if` / `elif` chain, Python stops at the **first true condition**.

For the delivery fee example, this customer should get free delivery:

```python
distance_km = 4
order_amount = 350
```

If we check only `distance_km <= 10` before the free-delivery rule, Python may return `40` too early.

So we should put special rules first:

1. invalid input
2. free delivery rule
3. normal distance fee rules


## 14. Worked example: Days in Month

The book uses **B.E.** year input, so convert to **C.E.** first:

$$
\text{CE} = \text{BE} - 543
$$

February has 29 days exactly when the C.E. year is a leap year:

- divisible by 4 and not divisible by 100, or
- divisible by 400


````python

def days_in_month(month, year_be):
    year_ce = year_be - 543

    if month in (1, 3, 5, 7, 8, 10, 12):
        return 31
    elif month in (4, 6, 9, 11):
        return 30
    elif month == 2:
        leap = (year_ce % 4 == 0 and year_ce % 100 != 0) or (year_ce % 400 == 0)
        if leap:
            return 29
        else:
            return 28
    else:
        return None

for month, year_be in [(10, 2557), (2, 2557), (4, 2556), (2, 2547)]:
    print(month, year_be, "->", days_in_month(month, year_be))
````

## 16. Worked example: grade cutting

This version validates the score range **first**.

That matters because `if` / `elif` stops at the first true condition.
So a value like `120` must be rejected before checking the grade bands.


````python
def cut_grade(x):
    if x < 0 or x > 100:
        return "ERROR"
    elif x >= 80:
        return "A"
    elif x >= 75:
        return "B+"
    elif x >= 70:
        return "B"
    elif x >= 65:
        return "C+"
    elif x >= 60:
        return "C"
    elif x >= 55:
        return "D+"
    elif x >= 50:
        return "D"
    else:
        return "F"

for x in [87.25, 69.95, 120, -3, 50, 75]:
    print(x, "->", cut_grade(x))
````

## 16. Worked example: parking fee

Rules:

1. at most 15 minutes -> free
2. over 15 minutes but not more than 3 hours -> 10 baht per started hour
3. from hour 4 through hour 6 -> 20 baht per started hour
4. over 6 hours -> flat 200 baht

A convenient method is:

- convert entry and exit times to minutes
- compute total parked minutes
- convert to **started hours** when needed


````python

import math

def parking_fee(h_in, m_in, h_out, m_out):
    start = h_in * 60 + m_in
    end = h_out * 60 + m_out
    parked = end - start

    if parked <= 15:
        return 0

    hours_started = math.ceil(parked / 60)

    if hours_started <= 3:
        return hours_started * 10
    elif hours_started <= 6:
        return 3 * 10 + (hours_started - 3) * 20
    else:
        return 200

examples = [
    (7, 0, 7, 15),
    (7, 0, 7, 16),
    (7, 30, 10, 30),
    (7, 30, 10, 31),
    (7, 30, 13, 31),
]

for ex in examples:
    print(ex, "->", parking_fee(*ex))
````

## 20. Unit 02 Selection Cheat Sheet

| Concept | Python pattern | Use when |
|---|---|---|
| Basic `if` | `if condition:` | Run code only when one condition is true. |
| `if` / `else` | `if condition:` / `else:` | Choose between two possible paths. |
| `if` / `elif` / `else` | `if ...` / `elif ...` / `else ...` | Choose one path from many rules. |
| Equal / not equal | `x == 10`, `x != 10` | Compare whether two values are the same or different. |
| Greater / less than | `x > 10`, `x <= 10` | Check numeric limits. |
| Chained comparison | `0 <= score <= 100` | Check whether a number is inside a range. |
| `and` | `age >= 18 and has_card == "yes"` | Require both conditions to be true. |
| `or` | `cash > 0 or card == "yes"` | Require at least one condition to be true. |
| `not` | `not is_closed` | Reverse a Boolean value. |
| Membership check | `month in (1, 3, 5, 7, 8, 10, 12)` | Check whether a value is inside a group. |
| String cleanup | `text.strip().lower()` | Make user text easier to compare. |
| Started hour | `math.ceil(minutes / 60)` | Convert minutes to charged hours for parking or rental fees. |
| Print 2 decimals | `print(f"{price:.2f}")` | Show money neatly. |
| Input as integer | `x = int(input())` | Read whole numbers. |
| Input as decimal | `x = float(input())` | Read numbers with decimals. |
| Multiple inputs | `a, b = [int(e) for e in input().split()]` | Read many numbers from one line. |

### Decision-order tips

1. Check invalid input first.
2. Put special cases before normal cases.
3. Put more specific rules before more general rules.
4. In an `if` / `elif` chain, Python runs only the first true branch.
5. Use indentation carefully because indentation decides which code belongs to which branch.

### Common beginner mistakes

| Mistake | Correct idea |
|---|---|
| Using `=` in a condition | Use `==` for comparison. |
| Forgetting `:` after `if` or `else` | Every branch header needs `:`. |
| Comparing input text without cleaning it | Use `.strip().lower()` first. |
| Checking general rules too early | Put special discounts or invalid cases first. |
| Writing too many repeated comparisons | Use chained comparisons like `50 <= score < 55`. |
