## 08.0 Why do we need functions?

Before functions, we often write code as one long script.

Example: calculate the average score for two subjects.


````python
math_scores = [80, 90, 75]
english_scores = [70, 85, 95]

math_total = 0
for score in math_scores:
    math_total += score
math_average = math_total / len(math_scores)

english_total = 0
for score in english_scores:
    english_total += score
english_average = english_total / len(english_scores)

print("Math average:", math_average)
print("English average:", english_average)
````


The logic is repeated. A function lets us write the logic once and reuse it many times.


````python
def average_score(scores):
    total = 0
    for score in scores:
        total += score
    return total / len(scores)

print("Math average:", average_score(math_scores))
print("English average:", average_score(english_scores))
````


# 08.1 Function basics

A **function** is a named block of code.

Think of a function like a small machine:

```text
input value(s)  ->  function  ->  output value
```

Example:

```python
def square(x):
    return x * x
```

| Part | Meaning |
|---|---|
| `def` | starts a function definition |
| `square` | function name |
| `x` | parameter, the input variable inside the function |
| `return` | sends the result back |
| `square(5)` | function call, using `5` as an argument |

### Parameter vs argument

| Word | Meaning | Example |
|---|---|---|
| Parameter | variable name inside function definition | `x` in `def square(x):` |
| Argument | actual value sent into the function | `5` in `square(5)` |


````python
def square(x):
    return x * x

print(square(3))
print(square(10))
````


## Function with no parameter

Some functions do not need input. They simply do a fixed action.


````python
def print_report_header():
    print("====================")
    print("Student Report")
    print("====================")

print_report_header()
````


## Function with parameters

Parameters allow the function to work with different input values.


````python
def greet_student(name):
    print("Hello", name)

# Same function, different arguments
greet_student("Alice")
greet_student("Boss")
````


## Function with a return value

`return` gives a value back to the program.

This is useful when you want to store the result in a variable or use it in another calculation.


````python
def add_vat(price):
    return price * 1.07

final_price = add_vat(100)
print("Final price:", final_price)
````


## `print()` vs `return`

This is one of the most important ideas in functions.

| Command | Meaning | Can be stored in a variable? |
|---|---|---|
| `print()` | shows something on the screen | No, it only displays |
| `return` | sends a value back to the caller | Yes |

A function that only prints usually returns `None`.


````python
def show_double(x):
    print(x * 2)


def get_double(x):
    return x * 2

show_result = show_double(5)
get_result = get_double(5)

print("show_result =", show_result)
print("get_result =", get_result)
````


## Daily-life example: split a restaurant bill

Problem:

You and your friends eat together. The bill is 1,200 baht. There are 4 people. Service charge is 10%.

Formula:

```text
bill_with_service = bill + bill * service_rate
amount_per_person = bill_with_service / number_of_people
```


````python
def split_bill(bill, number_of_people, service_rate):
    bill_with_service = bill + bill * service_rate
    amount_per_person = bill_with_service / number_of_people
    return amount_per_person

print(split_bill(1200, 4, 0.10))
````


## Common mistakes in function basics

| Mistake | Example | How to fix |
|---|---|---|
| Calling before defining | `answer = square(5)` before `def square(x):` | Define the function first |
| Forgetting parentheses | `square` instead of `square(5)` | Use parentheses to call the function |
| Using `print` when you need a value | `result = print(x)` | Use `return x` inside function |
| Forgetting `return` | function calculates but gives back `None` | Add `return result` |
| Wrong indentation | code outside function by accident | Keep function body indented |

# 08.2 Return, tuple return, and local variables

## `return` stops the function

When Python reaches `return`, the function ends immediately.


````python
def grade_status(score):
    if score < 0 or score > 100:
        return "Invalid score"
    if score >= 50:
        return "Pass"
    return "Fail"

print(grade_status(80))
print(grade_status(40))
print(grade_status(120))
````


## Returning more than one value

A function technically returns **one object**, but that object can be a tuple.

This allows us to return multiple related values.


````python
def min_max(numbers):
    smallest = min(numbers)
    largest = max(numbers)
    return smallest, largest

low, high = min_max([7, 2, 9, 4])
print("low =", low)
print("high =", high)
````


## Local variables

A variable created inside a function is a **local variable**. It only exists inside the function.

This helps avoid accidentally changing variables outside the function.


````python
def calculate_total(price, quantity):
    total = price * quantity
    return total

answer = calculate_total(50, 3)
print(answer)

# The variable `total` only exists inside calculate_total().
# Uncommenting this line would cause an error:
# print(total)
````


## Default parameters

A default parameter gives a value automatically when the caller does not provide one.

Example: VAT is usually 7%, so we can set `vat_rate=0.07`.


````python
def add_vat(price, vat_rate=0.07):
    return price * (1 + vat_rate)

print(add_vat(100))       # uses default 7%
print(add_vat(100, 0.10)) # uses custom 10%
````

# 08.3 Functions with containers

Functions often receive or return containers such as:

- list
- tuple
- set
- dictionary

These containers are useful when one value is not enough.



## Function with a list

Example: calculate total price from a shopping cart.


````python
def cart_total(prices):
    total = 0
    for price in prices:
        total += price
    return total

cart = [120, 35, 60, 200]
print(cart_total(cart))
````


## Mutating a list vs returning a new list

A list can be changed inside a function. This is called **mutation**.

For beginners, returning a new list is often easier to understand because the original list stays safe.


````python
def add_item_mutate(items, new_item):
    items.append(new_item)


def add_item_new_list(items, new_item):
    new_items = items.copy()
    new_items.append(new_item)
    return new_items

shopping = ["milk", "bread"]
add_item_mutate(shopping, "egg")
print("After mutate:", shopping)

shopping = ["milk", "bread"]
updated = add_item_new_list(shopping, "egg")
print("Original:", shopping)
print("Updated:", updated)
````


## Function with a dictionary

Example: store prices by item name.


````python
def get_price(price_dict, item_name):
    if item_name in price_dict:
        return price_dict[item_name]
    return "Item not found"

prices = {
    "milk": 45,
    "bread": 35,
    "egg": 60
}

print(get_price(prices, "milk"))
print(get_price(prices, "coffee"))
````


## Function with a set

Set is useful for unique values.

Example: count how many unique food items a student bought this week.


````python
def count_unique_items(items):
    unique_items = set(items)
    return len(unique_items)

items = ["milk", "bread", "milk", "egg", "bread"]
print(count_unique_items(items))
````

# 08.4 Refactoring with functions

**Refactoring** means improving code structure without changing the final result.

Functions help us refactor a big task into smaller steps.

Example: calculate a simple shopping receipt.

Instead of writing everything in one long block, we can split it into helper functions:

1. calculate subtotal,
2. calculate discount,
3. calculate VAT,
4. calculate final total.


````python
def calculate_subtotal(prices):
    return sum(prices)


def calculate_discount(subtotal, discount_percent):
    return subtotal * discount_percent / 100


def calculate_vat(amount, vat_rate=0.07):
    return amount * vat_rate


def calculate_final_total(prices, discount_percent):
    subtotal = calculate_subtotal(prices)
    discount = calculate_discount(subtotal, discount_percent)
    after_discount = subtotal - discount
    vat = calculate_vat(after_discount)
    final_total = after_discount + vat
    return final_total

prices = [120, 80, 50]
print(calculate_final_total(prices, 10))
````


## Why helper functions are useful

| Benefit | Explanation |
|---|---|
| Easier to read | Each function has one clear job |
| Easier to test | You can test one small function at a time |
| Easier to fix | If discount logic is wrong, fix only `calculate_discount()` |
| Easier to reuse | The same helper can be used in another program |

# 08.5 Recursion

A **recursive function** is a function that calls itself.

Recursion is useful when a problem can be described as:

```text
answer to current problem = current step + answer to smaller problem
```

Every recursive function needs two parts:

| Part | Meaning |
|---|---|
| Base case | the stopping condition |
| Recursive case | the part where the function calls itself |

Without a base case, recursion can run forever until Python raises an error.



## Recursion example 1: countdown

Problem: print numbers from `n` down to `1`.

Recursive thinking:

```text
countdown(n):
    if n == 0: stop
    otherwise: print n, then countdown(n - 1)
```


````python
def countdown(n):
    if n == 0:        # base case
        return
    print(n)
    countdown(n - 1)  # recursive case

countdown(5)
````


## Recursion example 2: sum from 1 to n

Mathematical idea:

```text
sum_to_n(5) = 5 + sum_to_n(4)
sum_to_n(4) = 4 + sum_to_n(3)
sum_to_n(3) = 3 + sum_to_n(2)
sum_to_n(2) = 2 + sum_to_n(1)
sum_to_n(1) = 1
```


````python
def sum_to_n(n):
    if n == 1:
        return 1
    return n + sum_to_n(n - 1)

print(sum_to_n(5))
````


## Recursion example 3: factorial

Factorial means:

```text
5! = 5 × 4 × 3 × 2 × 1
```

Recursive idea:

```text
factorial(5) = 5 × factorial(4)
factorial(4) = 4 × factorial(3)
factorial(3) = 3 × factorial(2)
factorial(2) = 2 × factorial(1)
factorial(1) = 1
```


````python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
````


## Loop vs recursion

For many beginner daily-life tasks, a normal loop is easier than recursion.

| Task | Usually easier with |
|---|---|
| sum spending in a list | loop or `sum()` |
| repeat something 10 times | `for` loop |
| keep asking until user enters stop | `while` loop |
| factorial / tree-like problem | recursion can be useful |

Recursion is important to understand, but do not force recursion when a loop is simpler.

# Unit 08 Cheat Sheet

## Function syntax

| Goal | Pattern | Example |
|---|---|---|
| Define a function | `def name():` | `def hello():` |
| Function with input | `def name(parameter):` | `def square(x):` |
| Call a function | `name(argument)` | `square(5)` |
| Return a value | `return value` | `return x * x` |
| No return value | use `print()` or update data | `print("Hello")` |
| Default parameter | `def f(x=default):` | `def add_vat(price, vat=0.07):` |
| Return two values | `return a, b` | `return min(x), max(x)` |
| Unpack returned values | `a, b = f()` | `low, high = min_max(nums)` |

## `print()` vs `return`

| Use | When to use |
|---|---|
| `print()` | when you only want to display output |
| `return` | when the result must be reused later |

## Containers in functions

| Container | Good for | Example function |
|---|---|---|
| List | ordered collection, may contain duplicates | `sum(prices)` |
| Tuple | fixed group of values | `return low, high` |
| Set | unique values | `set(names)` |
| Dictionary | key-value lookup | `prices["milk"]` |

## Recursion checklist

| Question | Why it matters |
|---|---|
| Do I have a base case? | prevents infinite recursion |
| Does the input become smaller? | moves toward the base case |
| Is recursion clearer than a loop? | if not, use a loop |
| Can Python handle the depth? | very deep recursion can cause errors |

## Common errors

| Error | Common cause | Fix |
|---|---|---|
| `NameError` | function not defined before call | define function first |
| `TypeError` | wrong number/type of arguments | check parameters |
| `None` result | forgot `return` | add `return` |
| infinite recursion | missing base case | add stopping condition |
