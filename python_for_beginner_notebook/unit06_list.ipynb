# 06.1 List

A **list** is an ordered container.  
It can store many values in one variable.

Example:

```python
scores = [80, 75, 92]
```

The values have positions called **indexes**.

Python indexes start at `0`.

```text
scores = [80, 75, 92]
          0   1   2
```

So:

```python
scores[0] == 80
scores[1] == 75
scores[2] == 92
```


## 1. Creating lists

Common ways to create a list:

```python
x = []
x = list()
x = [0] * 10
x = list("abcde")
x = list(range(1, 10, 2))
```

Important meaning:

- `[]` means empty list
- `[0] * 10` repeats the value `0` ten times
- `list("abcde")` turns a string into a list of characters
- `list(range(...))` turns a range into an actual list


````python
empty_1 = []
empty_2 = list()

zeros = [0] * 10
chars = list("abcde")
odd_numbers = list(range(1, 10, 2))

print("empty_1     =", empty_1)
print("empty_2     =", empty_2)
print("zeros       =", zeros)
print("chars       =", chars)
print("odd_numbers =", odd_numbers)
````

## 2. List length and basic operations

Useful built-in functions:

- `len(x)` gives the number of items
- `sum(x)` gives the total, if the list contains numbers
- `max(x)` gives the largest value
- `min(x)` gives the smallest value

Useful operators:

- `+` joins two lists
- `*` repeats a list


````python
x = [1, 2, 3]
y = [4, 5]

print("len(x) =", len(x))
print("sum(x) =", sum(x))
print("max(x) =", max(x))
print("min(x) =", min(x))

print("x + y =", x + y)
print("x * 3 =", x * 3)
````

## 3. `append`, `insert`, and `pop`

These methods change the original list.

```python
x.append(e)
```

adds `e` to the end.

```python
x.insert(i, e)
```

inserts `e` before index `i`.

```python
x.pop(i)
```

removes the item at index `i` and returns the removed value.


````python
x = [10, 20, 30]

x.append(40)
print("after append:", x)

x.insert(1, 15)
print("after insert:", x)

removed = x.pop(2)
print("removed item:", removed)
print("after pop:", x)
````

## 4. Indexing

For a list:

```python
x = [11, 12, 13, 14, 15]
```

Positive indexes go from left to right:

```text
 index:  0   1   2   3   4
 value: 11  12  13  14  15
```

Negative indexes go from right to left:

```text
 index: -5  -4  -3  -2  -1
 value: 11  12  13  14  15
```

The valid index range is:

```python
-len(x) <= index <= len(x)-1
```

For this example, valid indexes are from `-5` to `4`.


````python
x = [11, 12, 13, 14, 15]

print("x[0]  =", x[0])
print("x[2]  =", x[2])
print("x[-1] =", x[-1])
print("x[-5] =", x[-5])

try:
    print(x[5])
except IndexError as e:
    print("Index error example:", e)
````

## 5. Slicing

A **slice** takes a sub-list.

```python
x[start:stop:step]
```

Important rule:

- `start` is included
- `stop` is not included
- slicing always returns a list
- out-of-range slicing is safe and does not crash


````python
x = [11, 12, 13, 14, 15]

print("x[1:4]      =", x[1:4])
print("x[:3]       =", x[:3])
print("x[2:]       =", x[2:])
print("x[::2]      =", x[::2])
print("x[1::2]     =", x[1::2])
print("x[::-1]     =", x[::-1])
print("x[100:200]  =", x[100:200])
print("x[-500:-2] =", x[-500:-2])
````

## 6. Lists are mutable

A string is not easy to change character by character, but a list is mutable.

Mutable means:

You can replace one item:

You can also replace a slice:

## 7. `append` vs `+=`

Beginners often confuse these two.

```python
x.append(y)
```

puts `y` as **one new item** at the end.

```python
x += y
```

extends `x` using the items inside `y`.

So if `y` is a list, the results are different.


````python
x = [1, 2, 3]
x.append([4, 5])
print("append list:", x)

x = [1, 2, 3]
x += [4, 5]
print("+= list    :", x)

x = [1, 2, 3]
x.append(4)
print("append number:", x)

try:
    x = [1, 2, 3]
    x += 4
except TypeError as e:
    print("+= number error:", e)
````

## 8. Copying lists: alias vs real copy

This is one of the most important list topics.

```python
b = a
```

does **not** copy the list.  
It makes `b` refer to the same list as `a`.

Use one of these when you want a new list:

```python
b = list(a)
b = a[:]
b = [e for e in a]
```


````python
a = [1, 2, 3]
b = a          # alias: same list
c = list(a)    # real copy
d = a[:]       # real copy

a[0] = 999

print("a =", a)
print("b =", b, "  # changed because b is the same list as a")
print("c =", c, "  # unchanged copy")
print("d =", d, "  # unchanged copy")
````

## 9. `x = y`, `x = y[:]`, and `x[:] = y`

These look similar but mean different things.

```python
x = y
```

`x` and `y` become the same list.

```python
x = y[:]
```

`x` becomes a new copied list.

```python
x[:] = y
```

`x` stays the same list object, but its contents are replaced.

The last form is useful when another variable already points to the same list and you want all references to see the updated contents.


````python
x = [1, 2, 3]
alias = x

new_values = [7, 8, 9]
x[:] = new_values

print("x     =", x)
print("alias =", alias)
print("alias changed too because the original list contents were replaced")
````

## 10. Searching and counting

Common tools:

```python
x.count(e)
```

counts how many times `e` appears.

```python
e in x
```

checks whether `e` exists in the list.

```python
x.index(e)
```

returns the first index of `e`.

Be careful: `x.index(e)` crashes if `e` does not exist.  
Always check with `in` first.


````python
x = [5, 7, 5, 9, 5]

print("count 5:", x.count(5))
print("7 in x:", 7 in x)
print("8 in x:", 8 in x)

target = 9
if target in x:
    print("first index of", target, "is", x.index(target))
else:
    print(target, "not found")

target = 100
if target in x:
    print(x.index(target))
else:
    print(target, "not found, so we do not call index()")
````

## 11. Sorting: `sort()` vs `sorted()`

This is another common beginner mistake.

```python

```

changes `x` directly and returns `None`.


## 13. Common list processing patterns

### Pattern A: Read `n` values into a list

```python
n = int(input())
data = []
for k in range(n):
    data.append(float(input()))
```

### Pattern B: Process each value

```python
for e in x:
    ...
```

### Pattern C: Modify values inside a list

Use index-based looping:

```python
for i in range(len(x)):
    x[i] = ...
```

Do **not** expect this to modify the original list:

```python
for e in x:
    e = ...
```

That only changes the temporary variable `e`.


````python
scores = [20, 30, 40, 50]

# Wrong idea: this does not change the original list
for e in scores:
    e = e + 10

print("after wrong style:", scores)

# Correct style: use indexes when modifying list values
for i in range(len(scores)):
    scores[i] = scores[i] + 10

print("after correct style:", scores)
````

## 16. Adjacent pairs and all pairs

### Adjacent pairs

Use this when you compare each item with the next item.

```python
for i in range(len(x)-1):
    print(x[i], x[i+1])
```

### All pairs

Use nested loops when every pair matters.

```python
for i in range(len(x)):
    for j in range(i+1, len(x)):
        print(x[i], x[j])
```


````python
x = [5, 1, 4, 2]

print("Adjacent pairs:")
for i in range(len(x)-1):
    print(x[i], x[i+1])

print("\nAll pairs:")
for i in range(len(x)):
    for j in range(i+1, len(x)):
        print(x[i], x[j])
````

## 17. Bubble sort idea

Python already has `sort()` and `sorted()`, but Unit 06 also uses bubble sort to practice list processing.

Bubble sort repeatedly swaps neighboring values if they are in the wrong order.


````python
d = [5, 1, 4, 2]

for k in range(len(d)-1):
    for i in range(len(d)-1):
        if d[i] > d[i+1]:
            d[i], d[i+1] = d[i+1], d[i]

print(d)
````


## 19. Worked example: Josephus-style list simulation

This example shows why lists are useful for simulations.

Imagine we have `m` people numbered from `0` to `m-1` in a circle.  
Starting at index `0`, repeatedly move forward `d` positions and remove that person.  
Continue until one person remains.

The key list tools are:

- `list(range(m))` to create the people
- `% len(q)` to wrap around the circle
- `q.pop(k)` to remove a person


````python
m = 6
d = 2

q = list(range(m))
k = 0

while len(q) > 1:
    k = (k + d) % len(q)
    removed = q.pop(k)
    print("removed:", removed, "remaining:", q)

print("winner:", q[0])
````

````python
def dot_product_or_error(v1, v2):
    if len(v1) != len(v2):
        return "Error"
    dot = 0
    for i in range(len(v1)):
        dot += v1[i] * v2[i]
    return dot

def sort_numbers_csv(numbers):
    numbers = list(numbers)
    numbers.sort()
    return ",".join([str(e) for e in numbers])

def most_frequent_in_order(numbers):
    counts = []
    for e in numbers:
        counts.append(numbers.count(e))
    max_count = max(counts)
    answer = []
    for i in range(len(numbers)):
        if counts[i] == max_count and numbers[i] not in answer:
            answer.append(numbers[i])
    return answer

def extract_headlines_sorted(lines):
    headlines = []
    for line in lines:
        start = line.find("<headline>")
        if start >= 0:
            start = start + len("<headline>")
            end = line.find("</headline>", start)
            if end >= 0:
                headlines.append(line[start:end])
    headlines.sort()
    return headlines

def sort_names_by_length(names):
    temp = []
    for name in names:
        temp.append([len(name), name])
    temp.sort()
    return [name for length, name in temp]

def smallest_good_survivor_d(n):
    for d in range(1, 2*n):
        q = ["G"] * n + ["B"] * n
        k = 0
        while len(q) > 1:
            k = (k + d) % len(q)
            q.pop(k)
        if q[0] == "G":
            return d
    return "Not found"

assert dot_product_or_error([1, 2, 3], [4, 5, 6]) == 32
assert dot_product_or_error([1, 2], [1, 2, 3]) == "Error"
assert sort_numbers_csv([5, 1, 9, 2]) == "1,2,5,9"
assert most_frequent_in_order([5, 2, 5, 3, 2]) == [5, 2]
assert extract_headlines_sorted(["a <headline>Z</headline>", "<headline>A</headline>"]) == ["A", "Z"]
assert sort_names_by_length(["xyz", "xy", "abc"]) == ["xy", "abc", "xyz"]
assert smallest_good_survivor_d(1) == 1
assert smallest_good_survivor_d(3) == 2
assert smallest_good_survivor_d(6) == 3

print("All 06.1 checks passed.")
````

# 06.2 Nested List

A **nested list** is a list that contains another list.

Example:

```python
x = [1, [2, 3], [4, 5, 6]]
```

The item at index `1` is itself a list:

```python
x[1] == [2, 3]
```

To access `3`, use two indexes:

```python
x[1][1]
```

Read this as:

> go to `x[1]`, then go to index `1` inside that list.


````python
x = []

x.append(1)
x.append([2, 3])
x.append([4, 5, 6])
x.append([[7, 8]])
x.append([])

print("x =", x)
print("len(x) =", len(x))
print("x[1][1] =", x[1][1])
print("x[2][1] =", x[2][1])
print("x[3][0][1] =", x[3][0][1])
print("len(x[4]) =", len(x[4]))
````

## 1. Nested lists as records

A nested list can store structured daily-life data.

Example: each inner list stores one grocery item:

```python
[item_name, quantity, price_per_unit]
```

So a shopping list might look like:

```python
shopping_items = [
    ['apple', 4, 0.5],
    ['bread', 2, 1.5]
]
```


````python
shopping_items = [
    ['apple', 4, 0.5],
    ['bread', 2, 1.5],
    ['milk', 1, 2.0]
]

for item in shopping_items:
    name, qty, price = item
    print(name, '-> quantity =', qty, ', unit price =', price)
````

## 3. Creating a zero matrix correctly

To create a `5 x 3` matrix of zeros:

```python
A = []
for k in range(5):
    A.append([0] * 3)
```

Each row must be a different list.


````python
A = []

for k in range(5):
    A.append([0] * 3)

A[0][0] = 99

print(A)
print("Only row 0 changed, because each row is a different list.")
````

## 4. Dangerous shortcut: `[[0] * c] * r`

This creates repeated references to the **same row**.

It looks correct at first, but changing one row changes all rows.


````python
bad = [[0] * 3] * 4
bad[0][1] = 9

print(bad)
print("All rows changed at column 1. This is usually not what we want.")
````

## 5. Matrix addition

To add two matrices, loop over rows and columns.

If:

```python
A = [[1, 2], [1, 1]]
B = [[0, 1], [4, 3]]
```

Then:

```python
C = [[1, 3], [5, 4]]
```


````python
A = [[1, 2], [1, 1]]
B = [[0, 1], [4, 3]]

C = []

for i in range(len(A)):
    C.append([0] * len(A[i]))
    for j in range(len(A[i])):
        C[i][j] = A[i][j] + B[i][j]

print(C)
````

## 6. Ragged nested lists

Nested lists do not need to have equal-length rows.

Example: social followers.

Each inner list stores:

```python
[username, list_of_followers]
```

The list of followers can have different lengths.


````python
f = [
    ["noon", ["pat", "koi"]],
    ["wii", ["noon", "koi"]],
    ["pat", ["koi", "noon", "wii"]],
    ["koi", []]
]

max_username = ""
max_followers = []

for username, followers in f:
    if len(followers) > len(max_followers):
        max_username = username
        max_followers = followers

print(max_username, "has max number of followers:", ", ".join(max_followers))
````

## 7. Sorting nested lists

Python sorts nested lists by comparing inside values from left to right.

Example:

```python
x = [[3, 2], [3, 1], [9], [2, 5], [3]]
x.sort()
```

Python first compares item `0` of each inner list.  
If tied, it compares item `1`, then item `2`, and so on.


````python
x = [[3, 2], [3, 1], [9], [2, 5], [3]]
x.sort()
print(x)
````

## 8. Temporary keys for custom sorting

A useful beginner trick is:

1. create a temporary nested list,
2. put the sorting key first,
3. sort it,
4. extract the original information.

Example: sort users by number of followers, smallest first.

# 06.3 List Comprehension

A **list comprehension** is a compact way to create a list.

It often replaces this pattern:

```python
b = []
for e in a:
    b.append(...)
```

with this pattern:

```python
b = [... for e in a]
```

It is shorter and often faster.


## 1. Map pattern

The **map** pattern transforms every item.

Example: keep only the last digit of every number.

Loop version:

```python
b = []
for e in a:
    b.append(e % 10)
```

List comprehension version:

```python
b = [e % 10 for e in a]
```


````python
a = [12, 35, 49, 100]

b_loop = []
for e in a:
    b_loop.append(e % 10)

b_comp = [e % 10 for e in a]

print("loop:", b_loop)
print("comp:", b_comp)
````

## 2. Filter pattern

The **filter** pattern keeps only some items.

Example: keep strings whose length is greater than 5.

Loop version:

```python
b = []
for e in a:
    if len(e) > 5:
        b.append(e)
```

List comprehension version:

```python
b = [e for e in a if len(e) > 5]
```


````python
words = ["cat", "banana", "school", "AI", "programming"]

long_words = [e for e in words if len(e) > 5]

print(long_words)
````

## 3. Map + filter pattern

This pattern both selects and transforms.

Example: from heights in inches, keep values greater than `10`, then convert them to centimeters.

```python
b = [2.54 * e for e in a if e > 10]
```


````python
heights_in_inches = [5, 10, 12, 20]

heights_cm = [2.54 * e for e in heights_in_inches if e > 10]

print(heights_cm)
````

## 4. Nested loop comprehension

A list comprehension can contain more than one `for`.

Example:

Find pairs `[a, b]` where:

```python
a + b == a * b
```

for `0 <= a <= 20` and `a <= b <= 20`.


````python
pairs = [[a, b] for a in range(0, 21)
              for b in range(a, 21)
              if a + b == a * b]

print(pairs)
````

## 5. Copying a list with comprehension

This creates a new list with the same values:

```python
x = [e for e in a]
```

It is similar to:

```python
x = list(a)
x = a[:]
```

But remember:

```python
x = a
```

does not copy. It aliases the same list.


````python
a = [1, 2, 3]

x = [e for e in a]
y = a

a[0] = 999

print("a =", a)
print("x =", x, "# copy")
print("y =", y, "# alias")
````
