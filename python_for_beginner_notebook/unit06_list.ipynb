# 06.1 List

A **list** stores many values in one variable.

Example:

```python
shopping_items = ["milk", "bread", "egg"]
```

A list is **ordered**, so each item has an index.

```text
shopping_items = ["milk", "bread", "egg"]
                   0       1        2
```

Python indexes start at `0`.

## 1. Creating lists

| Code | Meaning |
|---|---|
| `x = []` | create an empty list |
| `x = list()` | create an empty list |
| `x = ["water"] * 3` | repeat `"water"` 3 times |
| `x = list("cat")` | convert text to a list of characters |
| `x = list(range(1, 6))` | create `[1, 2, 3, 4, 5]` |

````python
empty_cart = []
empty_cart_2 = list()
water_pack = ["water"] * 3
letters = list("cat")
numbers = list(range(1, 6))

print("empty_cart   =", empty_cart)
print("empty_cart_2 =", empty_cart_2)
print("water_pack   =", water_pack)
print("letters      =", letters)
print("numbers      =", numbers)
````

## 2. List length and simple calculations

When a list contains numbers, these functions are very useful.

| Function | Meaning | Daily-life example |
|---|---|---|
| `len(x)` | number of items | number of spending days |
| `sum(x)` | total value | total spending |
| `max(x)` | largest value | highest spending |
| `min(x)` | smallest value | lowest spending |

````python
daily_spending = [120, 80, 200, 50, 150]

print("number of days =", len(daily_spending))
print("total spending =", sum(daily_spending))
print("highest day    =", max(daily_spending))
print("lowest day     =", min(daily_spending))
print("average        =", sum(daily_spending) / len(daily_spending))
````

## 3. Indexing

Indexing selects one item.

```python
x[index]
```

Example:

```text
snacks = ["chips", "milk", "cake", "juice"]
           0        1       2       3
          -4       -3      -2      -1
```

````python
snacks = ["chips", "milk", "cake", "juice"]

print("first item  =", snacks[0])
print("third item  =", snacks[2])
print("last item   =", snacks[-1])
print("second last =", snacks[-2])
````

## 4. Slicing

Slicing selects a sub-list.

```python
x[start:stop:step]
```

Important rules:

- `start` is included
- `stop` is not included
- slicing returns a list
- out-of-range slicing is safe

````python
spending = [100, 120, 80, 90, 150, 300, 250]
# day:       Mon  Tue Wed Thu  Fri  Sat  Sun

print("Mon to Fri      =", spending[0:5])
print("Weekend         =", spending[5:7])
print("From Wed onward =", spending[2:])
print("Every other day =", spending[::2])
print("Reverse order   =", spending[::-1])
````

## 5. Updating list values

Lists are **mutable**.

Mutable means the object can be changed after it is created.

````python
todo = ["wake up", "study", "exercise", "sleep"]

todo[1] = "study Python"
print("after one update:", todo)

todo[2:4] = ["walk", "review notes"]
print("after slice update:", todo)
````

## 6. `append`, `insert`, `pop`, and `remove`

| Method | Meaning | Example |
|---|---|---|
| `x.append(e)` | add `e` to the end | add a shopping item |
| `x.insert(i, e)` | add `e` before index `i` | insert urgent item |
| `x.pop(i)` | remove item at index `i` and return it | remove by position |
| `x.remove(e)` | remove first matching value `e` | remove by name |

````python
groceries = ["milk", "bread", "egg"]

groceries.append("banana")
print("after append:", groceries)

groceries.insert(1, "water")
print("after insert:", groceries)

removed_item = groceries.pop(2)
print("removed item:", removed_item)
print("after pop:", groceries)

groceries.remove("water")
print("after remove:", groceries)
````

## 7. `append` vs `+=`

This is a common beginner mistake.

```python
x.append(y)
```

adds `y` as **one item**.

```python
x += y
```

extends `x` using every item inside `y`.

````python
cart = ["milk", "bread"]
cart.append(["egg", "banana"])
print("append result:", cart)

cart = ["milk", "bread"]
cart += ["egg", "banana"]
print("+= result    :", cart)
````

## 8. Copying lists: alias vs real copy

```python
b = a
```

does **not** copy the list. It makes `b` refer to the same list.

Use one of these for a real copy:

```python
b = list(a)
b = a[:]
b = [e for e in a]
```

````python
original = ["milk", "bread", "egg"]

alias_cart = original
copy_cart = list(original)

original.append("banana")

print("original =", original)
print("alias    =", alias_cart, "# changed too")
print("copy     =", copy_cart, "# unchanged")
````

## 9. Searching and counting

| Code | Meaning |
|---|---|
| `e in x` | check whether `e` exists |
| `x.count(e)` | count how many times `e` appears |
| `x.index(e)` | find the first index of `e` |

Important: `x.index(e)` causes an error if `e` does not exist.  
Always check with `in` first.

````python
orders = ["coffee", "tea", "coffee", "cake", "coffee"]

print("coffee count =", orders.count("coffee"))
print("tea in orders?", "tea" in orders)
print("juice in orders?", "juice" in orders)

target = "cake"
if target in orders:
    print(target, "first appears at index", orders.index(target))
else:
    print(target, "not found")
````

## 10. Sorting: `sort()` vs `sorted()`

```python
x.sort()
```

changes the original list directly and returns `None`.

```python
sorted(x)
```

returns a new sorted list and keeps the original list unchanged.

````python
prices = [50, 20, 100, 80]

sorted_prices = sorted(prices)
print("original after sorted() =", prices)
print("new sorted list         =", sorted_prices)

result = prices.sort()
print("original after sort()   =", prices)
print("result of sort()        =", result)
````

## 11. `split` and `join`

`split` turns text into a list.

```python
text.split()
```

`join` turns a list of strings into one string.

```python
", ".join(items)
```

````python
message = "milk bread egg"
items = message.split()
print("items =", items)

line = ", ".join(items)
print("line  =", line)

numbers_text = "10 20 30"
numbers = [int(e) for e in numbers_text.split()]
print("numbers =", numbers)
````

## 12. Processing a list with a loop

Use a loop when you want to process every item.

````python
spending = [100, 120, 80, 90]

total = 0
for amount in spending:
    total = total + amount

print("total spending =", total)
````

## 13. Changing values inside a list

If you want to change values inside the same list, use indexes.

This does **not** change the original list:

```python
for e in scores:
    e = e + 5
```

This changes the original list:

```python
for i in range(len(scores)):
    scores[i] = scores[i] + 5
```

````python
scores = [70, 80, 65]

for e in scores:
    e = e + 5
print("wrong style:", scores)

for i in range(len(scores)):
    scores[i] = scores[i] + 5
print("correct style:", scores)
````

## 15. Search with `break`

Use `break` when you want to stop a loop early.

Daily-life example: find the first day when spending is over budget.

````python
spending = [120, 150, 90, 250, 100]
budget = 200

for i, amount in enumerate(spending):
    if amount > budget:
        print("first over-budget day is day", i + 1)
        break
else:
    print("no over-budget day")
````

## 16. Adjacent pairs

Use adjacent pairs when comparing each item with the next item.

Pattern:

```python
for i in range(len(x) - 1):
    current = x[i]
    next_value = x[i + 1]
```

````python
steps = [5000, 6200, 5800, 7000]

for i in range(len(steps) - 1):
    change = steps[i + 1] - steps[i]
    print("Day", i + 1, "to day", i + 2, "change =", change)
````

Daily-life example: sort expenses from lowest to highest.

````python
expenses = [250, 80, 120, 40, 300]
expenses_sorted = sorted(expenses)

print("sorted expenses =", expenses_sorted)
print("cheapest        =", expenses_sorted[0])
print("most expensive  =", expenses_sorted[-1])
````

## 06.1 Auto-checks

````python
def update_cart(cart):
    cart = list(cart)
    cart.append("banana")
    cart.insert(0, "water")
    cart.remove("bread")
    return cart

def spending_summary(spending):
    return sum(spending), sum(spending) / len(spending), max(spending), min(spending)

def first_over_budget_day(spending, budget):
    for i, amount in enumerate(spending):
        if amount > budget:
            return i + 1
    return None

def add_bonus(scores, bonus):
    scores = list(scores)
    for i in range(len(scores)):
        scores[i] += bonus
    return scores

assert update_cart(["milk", "bread", "egg"]) == ["water", "milk", "egg", "banana"]
assert spending_summary([120, 80, 200, 50, 150]) == (600, 120.0, 200, 50)
assert first_over_budget_day([120, 150, 90, 250, 100], 200) == 4
assert add_bonus([70, 80, 65], 5) == [75, 85, 70]

print("06.1 checks passed")
````

# 06.2 Nested List

A **nested list** is a list that contains other lists.

It is useful for table-like data.

Example: weekly spending by category.

```python
weekly_spending = [
    [100, 50, 20],   # Monday: food, transport, snack
    [80, 60, 30],    # Tuesday
    [120, 40, 25]    # Wednesday
]
```

Each inner list is one row.

## 1. Creating a nested list

Daily-life example: expenses for 3 days.

| Day | Food | Transport | Snack |
|---|---:|---:|---:|
| Monday | 100 | 50 | 20 |
| Tuesday | 80 | 60 | 30 |
| Wednesday | 120 | 40 | 25 |

````python
expenses = [
    [100, 50, 20],
    [80, 60, 30],
    [120, 40, 25]
]

print(expenses)
````

## 2. Accessing nested-list values

Use two indexes:

```python
x[row][column]
```

Example:

```python
expenses[0][1]
```

means row `0`, column `1`.

````python
expenses = [
    [100, 50, 20],
    [80, 60, 30],
    [120, 40, 25]
]

print("Monday transport =", expenses[0][1])
print("Wednesday snack  =", expenses[2][2])
````

## 3. Looping through rows

Daily-life example: total spending for each day.

````python
expenses = [
    [100, 50, 20],
    [80, 60, 30],
    [120, 40, 25]
]

days = ["Mon", "Tue", "Wed"]

for i, row in enumerate(expenses):
    print(days[i], "total =", sum(row))
````

## 4. Row totals and grand total

Common table calculation:

1. calculate each row total,
2. calculate the grand total.

````python
expenses = [
    [100, 50, 20],
    [80, 60, 30],
    [120, 40, 25]
]

row_totals = []
for row in expenses:
    row_totals.append(sum(row))

grand_total = sum(row_totals)

print("row totals  =", row_totals)
print("grand total =", grand_total)
````

## 5. Column totals

Column totals are useful when each column has a meaning.

Example:

- column 0 = food
- column 1 = transport
- column 2 = snack

````python
expenses = [
    [100, 50, 20],
    [80, 60, 30],
    [120, 40, 25]
]

column_totals = [0, 0, 0]

for row in expenses:
    for j in range(len(row)):
        column_totals[j] += row[j]

print("food total      =", column_totals[0])
print("transport total =", column_totals[1])
print("snack total     =", column_totals[2])
````

## 6. Validating row length

When using table-like data, each row should usually have the same number of columns.

Daily-life example: each day should have exactly 3 expense categories.

````python
expenses = [
    [100, 50, 20],
    [80, 60],        # missing snack value
    [120, 40, 25]
]

valid = True
for row in expenses:
    if len(row) != 3:
        valid = False
        break

if valid:
    print("Data is valid")
else:
    print("Data is invalid")
````

## 7. Flattening a nested list

Flattening means turning a nested list into one normal list.

````python
grocery_by_bag = [
    ["milk", "egg"],
    ["rice"],
    ["banana", "water"]
]

all_items = []
for bag in grocery_by_bag:
    for item in bag:
        all_items.append(item)

print(all_items)
````

## 8. Table data with names

Sometimes each row contains text and numbers.

Example: café orders.

```python
orders = [
    ["coffee", 2, 50],
    ["cake", 1, 80]
]
```

Each row means:

```text
[item name, quantity, price per item]
```

````python
orders = [
    ["coffee", 2, 50],
    ["cake", 1, 80],
    ["tea", 3, 40]
]

total = 0
for item, quantity, price in orders:
    line_total = quantity * price
    total += line_total
    print(item, "=", line_total)

print("Total bill =", total)
````

## 06.2 Auto-checks

````python
def daily_totals(expenses):
    result = []
    for row in expenses:
        result.append(sum(row))
    return result

def category_totals(expenses):
    if len(expenses) == 0:
        return []
    result = [0] * len(expenses[0])
    for row in expenses:
        for j in range(len(row)):
            result[j] += row[j]
    return result

def is_valid_table(data, expected_columns):
    for row in data:
        if len(row) != expected_columns:
            return False
    return True

def cafe_bill(orders):
    total = 0
    for item, quantity, price in orders:
        total += quantity * price
    return total

def flatten(nested):
    result = []
    for row in nested:
        for item in row:
            result.append(item)
    return result

assert daily_totals([[100, 50, 20], [80, 60, 30], [120, 40, 25]]) == [170, 170, 185]
assert category_totals([[100, 50, 20], [80, 60, 30], [120, 40, 25]]) == [300, 150, 75]
assert is_valid_table([[100, 50, 20], [80, 60], [120, 40, 25]], 3) == False
assert cafe_bill([["coffee", 2, 50], ["cake", 1, 80], ["tea", 3, 40]]) == 300
assert flatten([["milk", "egg"], ["rice"], ["banana", "water"]]) == ["milk", "egg", "rice", "banana", "water"]

print("06.2 checks passed")
````

# 06.3 List Comprehension

A **list comprehension** is a shorter way to create a list.

Normal loop:

```python
result = []
for e in x:
    result.append(e * 2)
```

List comprehension:

```python
result = [e * 2 for e in x]
```

Use list comprehension when the logic is simple and easy to read.

## 1. Map pattern

The **map pattern** transforms every item.

Example: add 5 bonus points to every score.

````python
scores = [70, 80, 65]

bonus_scores = [score + 5 for score in scores]

print(bonus_scores)
````

## 2. Filter pattern

The **filter pattern** keeps only some items.

Example: keep only expenses greater than 100.

````python
expenses = [50, 120, 80, 200, 90]

large_expenses = [amount for amount in expenses if amount > 100]

print(large_expenses)
````

## 3. Map + filter pattern

This pattern selects and transforms at the same time.

Example: apply a 10% discount only to prices at least 100.

````python
prices = [50, 120, 80, 200]

discounted = [price * 0.9 for price in prices if price >= 100]

print(discounted)
````

## 4. Convert input text to numbers

This is one of the most common beginner patterns.

```python
numbers = [int(e) for e in text.split()]
```

Meaning:

1. split text into words,
2. convert each word to integer,
3. store all integers in a list.

````python
text = "10 20 30"

numbers = [int(e) for e in text.split()]

print(numbers)
````

## 5. Convert list values to strings for `join`

`join` works with strings.

If the list contains numbers, convert them first.

````python
scores = [80, 90, 75]

text = ", ".join([str(score) for score in scores])

print(text)
````

## 6. Counting with `sum`

You can count matching items by creating `1` for each match.

Example: count how many scores passed.

````python
scores = [45, 60, 80, 30, 75]

pass_count = sum([1 for score in scores if score >= 50])

print(pass_count)
````

## 7. Conditional expression inside list comprehension

This pattern chooses one value or another.

```python
value_if_true if condition else value_if_false
```

Example: label each score as `Pass` or `Fail`.

````python
scores = [45, 60, 80, 30, 75]

labels = ["Pass" if score >= 50 else "Fail" for score in scores]

print(labels)
````

## 8. Flatten a nested list with comprehension

Normal nested loop:

```python
all_items = []
for bag in grocery_by_bag:
    for item in bag:
        all_items.append(item)
```

List comprehension:

```python
all_items = [item for bag in grocery_by_bag for item in bag]
```

````python
grocery_by_bag = [
    ["milk", "egg"],
    ["rice"],
    ["banana", "water"]
]

all_items = [item for bag in grocery_by_bag for item in bag]

print(all_items)
````

## 9. Safe nested-list creation

To create a table of zeros, use this:

```python
table = [[0] * 3 for row in range(4)]
```

Avoid this:

```python
table = [[0] * 3] * 4
```

because all rows refer to the same inner list.

````python
good_table = [[0] * 3 for row in range(4)]
good_table[0][0] = 9

bad_table = [[0] * 3] * 4
bad_table[0][0] = 9

print("good_table =", good_table)
print("bad_table  =", bad_table)
````

## 10. When not to use list comprehension

Do **not** use list comprehension only to print things.

Avoid this:

```python
[print(item) for item in items]
```

Use a normal loop instead:

```python
for item in items:
    print(item)
```

List comprehension should create a useful list.

````python
items = ["milk", "bread", "egg"]

for item in items:
    print(item)
````

# Unit 06 Cheat Sheet

## 1. List basics

| Goal | Python pattern | Example |
|---|---|---|
| Create empty list | `x = []` | `cart = []` |
| Create list with values | `x = [a, b, c]` | `scores = [70, 80, 90]` |
| Get length | `len(x)` | `len(scores)` |
| Get total | `sum(x)` | `sum(spending)` |
| Get max / min | `max(x)`, `min(x)` | `max(scores)` |
| First item | `x[0]` | `cart[0]` |
| Last item | `x[-1]` | `cart[-1]` |
| Slice | `x[start:stop]` | `spending[0:5]` |
| Reverse | `x[::-1]` | `cart[::-1]` |

## 2. Changing lists

| Goal | Python pattern | Example |
|---|---|---|
| Replace item | `x[i] = value` | `scores[0] = 100` |
| Add to end | `x.append(value)` | `cart.append("milk")` |
| Insert item | `x.insert(i, value)` | `cart.insert(0, "water")` |
| Remove by index | `x.pop(i)` | `cart.pop(1)` |
| Remove by value | `x.remove(value)` | `cart.remove("milk")` |
| Sort original | `x.sort()` | `scores.sort()` |
| New sorted list | `sorted(x)` | `sorted(scores)` |
| Real copy | `y = list(x)` or `y = x[:]` | `backup = cart[:]` |
| Alias, not copy | `y = x` | both names point to same list |

## 3. Searching and counting

| Goal | Python pattern | Example |
|---|---|---|
| Check existence | `value in x` | `"milk" in cart` |
| Count value | `x.count(value)` | `orders.count("coffee")` |
| Find first index | `x.index(value)` | `cart.index("milk")` |
| Safe index search | `if value in x: x.index(value)` | avoid error |

## 4. Loop patterns

| Goal | Python pattern |
|---|---|
| Loop through values | `for e in x:` |
| Loop through indexes | `for i in range(len(x)):` |
| Need index and value | `for i, e in enumerate(x):` |
| Stop early | `break` |
| Search with not found case | `for ...: ... break` then `else:` |
| Compare neighbors | `for i in range(len(x) - 1):` |

## 5. Nested list patterns

| Goal | Python pattern | Example |
|---|---|---|
| Create nested list | `x = [[...], [...]]` | expenses table |
| Access row | `x[i]` | `expenses[0]` |
| Access cell | `x[i][j]` | `expenses[0][1]` |
| Loop rows | `for row in x:` | process each day |
| Loop cells | `for row in x: for e in row:` | flatten data |
| Row total | `sum(row)` | daily total |
| Validate row length | `len(row) == expected` | check table data |

## 6. List comprehension patterns

| Goal | Python pattern | Example |
|---|---|---|
| Transform every item | `[expression for e in x]` | `[e + 5 for e in scores]` |
| Filter items | `[e for e in x if condition]` | `[e for e in scores if e >= 50]` |
| Transform and filter | `[expression for e in x if condition]` | `[p * 0.9 for p in prices if p >= 100]` |
| If/else expression | `[A if condition else B for e in x]` | pass/fail labels |
| Convert input text | `[int(e) for e in text.split()]` | numbers from text |
| Join numbers as text | `", ".join([str(e) for e in x])` | output list nicely |
| Flatten nested list | `[e for row in x for e in row]` | all grocery items |
| Safe table of zeros | `[[0] * c for row in range(r)]` | rows are independent |

## 7. Common beginner mistakes

| Mistake | Problem | Better way |
|---|---|---|
| `x[len(x)]` | index out of range | last item is `x[-1]` |
| `y = x` for copy | creates alias | use `y = x[:]` or `list(x)` |
| `x.append([a, b])` when you want separate items | adds one nested list | use `x += [a, b]` |
| `x.sort()` assigned to variable | returns `None` | use `sorted(x)` for new list |
| modifying `e` in `for e in x` | does not update list | use indexes |
| `[[0] * c] * r` | shared inner rows | use `[[0] * c for row in range(r)]` |
