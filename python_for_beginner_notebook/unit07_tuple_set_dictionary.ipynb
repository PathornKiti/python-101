# 07.0 Big Picture: Choosing the Right Container

Before learning each container, compare the four common containers.

| Question | Best container | Why |
|---|---|---|
| I need to keep ordered values and edit them later. | `list` | Lists are ordered and mutable. |
| I need to keep ordered values together, but not edit them directly. | `tuple` | Tuples are ordered and immutable. |
| I need only unique values. | `set` | Sets automatically remove duplicates. |
| I need to look up a value using a key. | `dict` | Dictionaries map keys to values. |

Simple mental model:

```text
list  -> sequence of editable items
tuple -> fixed record / fixed group
set   -> unique collection
dict  -> key-value table
```

````python
# Quick comparison example
shopping_list = ['milk', 'bread', 'milk']              # list keeps duplicates and order
location = ('Floor 3', 'Room 305')                     # tuple keeps related values together
unique_items = {'milk', 'bread', 'milk'}               # set removes duplicate 'milk'
menu_price = {'milk': 45, 'bread': 35}                 # dict maps item -> price

print('list :', shopping_list)
print('tuple:', location)
print('set  :', unique_items)
print('dict :', menu_price)
````

# 07.1 Tuple

A **tuple** is an ordered container, similar to a list.

The most important difference:

```text
list  -> can be changed directly
tuple -> cannot be changed directly
```

A tuple is useful when several values belong together as **one record**.

Examples:

```python
point = (3, 4)
student_record = ('May', 85)
address_part = ('Bangkok', 'Thailand')
```

Use a tuple when:

- order matters,
- values should stay together,
- you do not want accidental editing,
- you want to unpack values into variables.

## 07.1.1 Creating tuples

Common ways to create tuples:

```python
a = (1, 2, 3)
b = tuple([1, 2, 3])
c = tuple('abc')
empty_tuple = ()
```

Important beginner warning:

| Code | Meaning |
|---|---|
| `(5,)` | tuple with one item |
| `(5)` | just integer `5` inside parentheses |

The comma is what makes a one-item tuple.

````python
a = (1, 2, 3)
b = tuple([1, 2, 3])
c = tuple('abc')
empty_tuple = ()

one_item_tuple = (5,)
not_tuple = (5)

print(a)
print(b)
print(c)
print(empty_tuple)
print(type(one_item_tuple))
print(type(not_tuple))
````

## 07.1.2 Tuple indexing and slicing

Tuples are ordered, so we can use indexes like lists.

```text
items = ('rice', 'egg', 'milk', 'bread')
          0       1      2       3
         -4      -3     -2      -1
```

Useful patterns:

| Pattern | Meaning |
|---|---|
| `items[0]` | first item |
| `items[-1]` | last item |
| `items[1:3]` | items from index 1 up to before index 3 |
| `items[:2]` | first two items |
| `items[2:]` | from index 2 to the end |

````python
items = ('rice', 'egg', 'milk', 'bread')

print('First item:', items[0])
print('Last item:', items[-1])
print('Middle slice:', items[1:3])
print('First two:', items[:2])
print('From index 2:', items[2:])
````

## 07.1.3 Tuple cannot be edited directly

A list can be edited directly:

```python
scores = [70, 80, 90]
scores[0] = 75
```

A tuple cannot be edited directly:

```python
scores = (70, 80, 90)
scores[0] = 75    # error
```

If you want a changed tuple, create a **new tuple** from old parts.

Pattern:

```python
new_tuple = old_tuple[:index] + (new_value,) + old_tuple[index+1:]
```

Why `(new_value,)`? Because the replacement piece must also be a tuple.

````python
scores = (70, 80, 90, 85)
index = 1
new_score = 88

updated_scores = scores[:index] + (new_score,) + scores[index+1:]

print('Original:', scores)
print('Updated :', updated_scores)
````

## 07.1.4 Tuple `+` and `*`

Tuples can be joined or repeated.

| Operator | Meaning | Example |
|---|---|---|
| `+` | join tuples | `(1, 2) + (3, 4)` |
| `*` | repeat tuple | `('Hi',) * 3` |

The result is a new tuple.

````python
breakfast = ('rice', 'egg')
drinks = ('water', 'coffee')

meal = breakfast + drinks
repeat_reminder = ('Drink water',) * 3

print(meal)
print(repeat_reminder)
````

## 07.1.5 Tuple packing and unpacking

Tuple packing means putting several values together.

```python
student = 'May', 20, 'Engineering'
```

Tuple unpacking means taking values out into separate variables.

```python
name, age, faculty = student
```

The number of variables must match the number of tuple items.

````python
student = 'May', 20, 'Engineering'
print('Packed tuple:', student)

name, age, faculty = student
print('Name   :', name)
print('Age    :', age)
print('Faculty:', faculty)
````

## 07.1.6 Swapping variables with tuple unpacking

A common Python pattern:

```python
a, b = b, a
```

This is useful when two values need to switch places.

````python
left_seat = 'Tom'
right_seat = 'Pam'

print('Before:', left_seat, right_seat)

left_seat, right_seat = right_seat, left_seat

print('After :', left_seat, right_seat)
````

## 07.1.7 List of tuples for records

A very common pattern is a **list of tuples**.

Example:

```python
students = [('Tom', 85), ('Pam', 92), ('Jim', 78)]
```

Meaning:

- the outer `list` stores many records,
- each inner `tuple` stores one fixed record: `(name, score)`.

This is useful when each record has a fixed structure.

````python
students = [('Tom', 85), ('Pam', 92), ('Jim', 78)]

for name, score in students:
    print(name, 'got', score)
````

## 07.1.8 Tuple of lists: important detail

A tuple itself cannot be edited directly.
But if a tuple contains a list, the list inside can still be edited.

Example:

```python
data = ([1, 2], [3, 4])
```

This is not allowed:

```python
data[0] = [9, 9]
```

But this is allowed:

```python
data[0][1] = 99
```

Why? The tuple still points to the same list object, but the list content changes.

````python
weekly_tasks = (['homework', 'review'], ['exercise'])

print('Before:', weekly_tasks)

weekly_tasks[0].append('practice Python')

print('After :', weekly_tasks)
````

## 07.1.9 Common tuple mistakes

| Mistake | Why it is wrong | Fix |
|---|---|---|
| `x = (5)` | This is an integer, not a tuple. | Use `x = (5,)`. |
| `t[0] = 99` | Tuple items cannot be assigned directly. | Create a new tuple. |
| `t.append(9)` | Tuple has no `append`. | Use a list if you need append. |
| `a, b = (1, 2, 3)` | Number of variables does not match. | Use `a, b, c = ...`. |

# 07.2 Set

A **set** stores unique values.

Important ideas:

- duplicates are automatically removed,
- order is not guaranteed,
- set items have no index,
- membership checking with `in` is convenient,
- set operations help compare groups.

Daily-life examples:

```python
unique_students = {'Tom', 'Pam', 'Tom'}
required_items = {'notebook', 'pen', 'calculator'}
```

In the first example, duplicate `'Tom'` is stored only once.

## 07.2.1 Creating sets

Common ways to create sets:

```python
s = {1, 2, 3}
s = set()
s = set([1, 2, 2, 3])
s = set('banana')
```

Important beginner warning:

| Code | Meaning |
|---|---|
| `{}` | empty dictionary |
| `set()` | empty set |

So do **not** use `{}` when you want an empty set.

````python
a = {1, 2, 3}
b = set()
c = set([1, 2, 2, 3])
d = set('banana')

print(a)
print(b)
print(c)
print(d)
print(type({}))
print(type(set()))
````

## 07.2.2 Set removes duplicates

This is one of the most common uses of a set.

Example problem:

> A student writes a shopping basket, but some items appear many times. Find the unique items.

```python
basket = ['milk', 'bread', 'milk', 'egg', 'bread']
unique_items = set(basket)
```

If you want a clean ordered display, use `sorted()`.

````python
basket = ['milk', 'bread', 'milk', 'egg', 'bread']

unique_items = set(basket)
ordered_unique_items = sorted(unique_items)

print('Unique set:', unique_items)
print('Sorted display:', ordered_unique_items)
````

## 07.2.3 Membership with `in`

Use `in` when you want to check whether a value exists.

```python
if name in registered_students:
    print('Found')
```

This is useful for:

- attendance checking,
- checking whether an item is on a shopping list,
- checking whether a skill is already learned,
- checking whether a coupon code is valid.

````python
valid_coupon_codes = {'SAVE10', 'STUDENT20', 'FREESHIP'}
code_input = 'STUDENT20'

if code_input in valid_coupon_codes:
    print('Coupon accepted')
else:
    print('Invalid coupon')
````

## 07.2.4 Adding and removing set items

Useful methods:

| Method | Meaning | Beginner note |
|---|---|---|
| `s.add(value)` | add one value | duplicates are still stored once |
| `s.update(values)` | add many values | values can come from a list, tuple, or set |
| `s.discard(value)` | remove value if it exists | no error if missing |
| `s.remove(value)` | remove value | error if missing |

For beginners, `discard()` is safer than `remove()`.

````python
attendees = {'Tom', 'Pam'}

attendees.add('Jim')
attendees.update(['May', 'Ann'])
attendees.discard('Pam')
attendees.discard('Unknown')   # no error

print(attendees)
````

## 07.2.5 Set operations

Set operations compare groups.

Assume:

```python
A = {'Tom', 'Pam', 'Jim'}
B = {'Pam', 'Jim', 'May'}
```

| Operation | Python | Meaning | Daily-life interpretation |
|---|---|---|---|
| Union | `A | B` | in A or B | everyone who joined at least one club |
| Intersection | `A & B` | in both A and B | students in both clubs |
| Difference | `A - B` | in A but not B | students only in A |
| Symmetric difference | `A ^ B` | in exactly one side | students who joined only one club |

````python
robot_club = {'Tom', 'Pam', 'Jim'}
ai_club = {'Pam', 'Jim', 'May'}

print('At least one club:', robot_club | ai_club)
print('Both clubs       :', robot_club & ai_club)
print('Robot only       :', robot_club - ai_club)
print('Exactly one club :', robot_club ^ ai_club)
````

## 07.2.6 Subset and superset

Use subset and superset when you want to check whether one group is fully inside another group.

Example:

```python
required = {'python', 'math'}
student_skills = {'python', 'math', 'excel'}
```

Question:

> Does the student have all required skills?

Python:

```python
required.issubset(student_skills)
```

````python
required = {'python', 'math'}
student_skills = {'python', 'math', 'excel'}

print(required.issubset(student_skills))
print(student_skills.issuperset(required))
````

## 07.2.7 Sets have no index

A set has no fixed order.

This means this is wrong:

```python
s[0]
```

Use a list when position matters.
Use a set when uniqueness or membership matters.

````python
# Example: use list when order matters
morning_routine = ['wake up', 'brush teeth', 'breakfast']
print('First step:', morning_routine[0])

# Example: use set when uniqueness matters
completed_tasks = {'homework', 'exercise', 'homework'}
print('Completed unique tasks:', completed_tasks)
````

## 07.2.8 Common set mistakes

| Mistake | Why it is wrong | Fix |
|---|---|---|
| `s = {}` | This creates a dictionary, not a set. | Use `s = set()`. |
| `s[0]` | Set has no index. | Convert to list or use a list from the start. |
| Expecting order | Set order is not guaranteed. | Use `sorted(s)` for display. |
| `s.add([1, 2])` | Lists are mutable and cannot be set items. | Use a tuple: `s.add((1, 2))`. |

# 07.3 Dictionary

A **dictionary** stores key-value pairs.

```python
menu = {'coffee': 55, 'tea': 45}
```

Meaning:

```text
key   -> 'coffee'
value -> 55
```

Use a dictionary when you want to ask:

> Given this key, what is the value?

Daily-life examples:

| Situation | Key | Value |
|---|---|---|
| Cafe menu | drink name | price |
| Phone contact | person name | phone number |
| Student grades | student ID | grade |
| Inventory | item name | quantity |

## 07.3.1 Creating dictionaries

Common ways to create dictionaries:

```python
d = {}
d = dict()
menu = {'coffee': 55, 'tea': 45}
```

Each pair is written as:

```python
key: value
```

Keys are unique. If the same key appears twice, the later value replaces the earlier value.

````python
menu = {'coffee': 55, 'tea': 45, 'cake': 80}
empty_dict = {}

print(menu)
print(empty_dict)
print(menu['coffee'])
````

## 07.3.2 Accessing values by key

Use square brackets when you are sure the key exists:

```python
d[key]
```

If the key does not exist, Python gives an error.

Safer beginner patterns:

```python
if key in d:
    print(d[key])
else:
    print('Not Found')
```

or:

```python
d.get(key, default_value)
```

````python
contacts = {'Mom': '081-111-1111', 'Dad': '082-222-2222'}
name = 'Mom'

if name in contacts:
    print(contacts[name])
else:
    print('Not Found')

print(contacts.get('Teacher', 'Not Found'))
````

## 07.3.3 Adding, updating, and deleting data

Use the same pattern for adding and updating:

```python
d[key] = value
```

If the key is new, Python adds it.
If the key already exists, Python updates it.

To delete:

```python
d.pop(key)
```

Check the key first if you are not sure it exists.

````python
inventory = {'pen': 10, 'notebook': 5}

inventory['eraser'] = 3       # add
inventory['pen'] = 12         # update

if 'notebook' in inventory:
    inventory.pop('notebook') # delete

print(inventory)
````

## 07.3.4 Looping through a dictionary

Useful dictionary views:

| Code | Meaning |
|---|---|
| `d.keys()` | all keys |
| `d.values()` | all values |
| `d.items()` | all key-value pairs |

Most of the time, `d.items()` is easiest when you need both key and value.

````python
menu = {'coffee': 55, 'tea': 45, 'cake': 80}

print('Keys:')
for item in menu.keys():
    print(item)

print('Values:')
for price in menu.values():
    print(price)

print('Items:')
for item, price in menu.items():
    print(item, price)
````

## 07.3.5 Sorting dictionary output

A dictionary is mainly for lookup, not sorted display.
But you can sort when you print.

Sort by key:

```python
for key in sorted(d.keys()):
    print(key, d[key])
```

For beginner-friendly sorting by value, create temporary tuples.

````python
scores = {'Tom': 85, 'Pam': 92, 'Jim': 78}

print('Sorted by name:')
for name in sorted(scores.keys()):
    print(name, scores[name])

print('Sorted by score from high to low:')
temp = []
for name, score in scores.items():
    temp.append((-score, name))

temp.sort()

for negative_score, name in temp:
    print(name, -negative_score)
````

## 07.3.6 Counting with a dictionary

A very common dictionary task is counting frequency.

Example:

> Count how many times each item appears in a basket.

Beginner pattern:

```python
count = {}
for item in basket:
    if item not in count:
        count[item] = 1
    else:
        count[item] += 1
```

Why check first?
Because the key must exist before we can add `1` to its value.

````python
basket = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

item_count = {}
for item in basket:
    if item not in item_count:
        item_count[item] = 1
    else:
        item_count[item] += 1

print(item_count)
````

## 07.3.7 Dictionary with list values

Sometimes one key has many values.

Example:

- key = customer name
- value = list of items bought

Use a list as the value when:

- order matters,
- duplicates are meaningful,
- you want to remember history.

````python
records = [
    ('Tom', 'coffee'),
    ('Pam', 'tea'),
    ('Tom', 'cake'),
    ('Tom', 'coffee')
]

customer_history = {}

for name, item in records:
    if name not in customer_history:
        customer_history[name] = []
    customer_history[name].append(item)

print(customer_history)
````

## 07.3.8 Dictionary with set values

Sometimes one key has many values, but duplicates should be removed.

Example:

- key = club name
- value = set of student names

Use a set as the value when:

- duplicates should be removed,
- order does not matter,
- you mostly care about membership.

````python
records = [
    ('Tom', 'Music'),
    ('Pam', 'Sports'),
    ('Jim', 'Music'),
    ('Tom', 'Music')
]

club_to_students = {}

for name, club in records:
    if club not in club_to_students:
        club_to_students[club] = set()
    club_to_students[club].add(name)

print(club_to_students)
````

## 07.3.9 Inverting a dictionary of sets

Sometimes you have this direction:

```python
student_to_courses = {
    '001': {'python', 'math'},
    '002': {'python', 'english'}
}
```

But you want this direction:

```python
course_to_students = {
    'python': {'001', '002'},
    'math': {'001'},
    'english': {'002'}
}
```

This is called **inverting** the dictionary because the relationship is flipped.

````python
student_to_courses = {
    '001': {'python', 'math'},
    '002': {'python', 'english'},
    '003': {'math'}
}

course_to_students = {}

for sid, courses in student_to_courses.items():
    for course in courses:
        if course not in course_to_students:
            course_to_students[course] = set()
        course_to_students[course].add(sid)

print(course_to_students)
````

## 07.3.10 Tuple keys in dictionaries

A dictionary key must be immutable.

Allowed key examples:

- number,
- string,
- tuple of immutable values.

Not allowed key examples:

- list,
- set,
- dictionary.

Useful daily-life example: menu price by `(item, size)`.

```python
price[('coffee', 'small')]
price[('coffee', 'large')]
```

````python
price = {
    ('coffee', 'small'): 45,
    ('coffee', 'large'): 60,
    ('tea', 'small'): 40,
    ('tea', 'large'): 55
}

order = ('coffee', 'large')

print(price[order])
print(price.get(('cake', 'small'), 'Not Found'))
````

## 07.3.11 Common dictionary mistakes

| Mistake | Why it is wrong | Fix |
|---|---|---|
| `d['missing']` | Error if key does not exist. | Use `if key in d` or `d.get(key, default)`. |
| `count[item] += 1` for a new item | Key does not exist yet. | Initialize first. |
| `d[[1, 2]] = 'value'` | List cannot be a key. | Use tuple `(1, 2)`. |
| `{'A': 1, 'A': 2}` | Keys are unique; first value is overwritten. | Use different keys or store list/set as value. |

# 07.4 Mixed Practice: Choosing the Right Container

In real problems, you often combine containers.

Examples:

| Structure | Meaning |
|---|---|
| `list` of `tuple` | many fixed records, e.g. `[('Tom', 85), ('Pam', 92)]` |
| `dict` with `list` values | one key has many ordered values, duplicates allowed |
| `dict` with `set` values | one key has many unique values |
| `dict` with `tuple` keys | lookup by two or more pieces of information |

The skill is to choose the structure that makes the task simple.

## Mixed Example: Shopping basket summary

Problem:

You have a basket with repeated items.

```python
basket = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
```

Create:

- `unique_items`: set of unique items,
- `ordered_unique_items`: sorted list of unique items,
- `item_count`: dictionary counting each item.

Container choices:

| Result | Container | Reason |
|---|---|---|
| `unique_items` | `set` | remove duplicates |
| `ordered_unique_items` | `list` | sorted display |
| `item_count` | `dict` | item name -> count |

````python
basket = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

unique_items = set(basket)
ordered_unique_items = sorted(unique_items)

item_count = {}
for item in basket:
    if item not in item_count:
        item_count[item] = 1
    else:
        item_count[item] += 1

print('Unique set:', unique_items)
print('Sorted unique list:', ordered_unique_items)
print('Item count:', item_count)
````

````python
# Reference solution code: Mixed Practice

# M1
study_log = [
    ('Mon', 'Python'),
    ('Tue', 'Math'),
    ('Wed', 'Python'),
    ('Thu', 'English'),
    ('Fri', 'Math')
]

subjects_studied = set()
day_to_subject = {}
subject_count = {}

for day, subject in study_log:
    subjects_studied.add(subject)
    day_to_subject[day] = subject

    if subject not in subject_count:
        subject_count[subject] = 1
    else:
        subject_count[subject] += 1

# M2
price = {
    ('coffee', 'small'): 45,
    ('coffee', 'large'): 60,
    ('tea', 'small'): 40,
    ('tea', 'large'): 55
}
orders = [('coffee', 'large'), ('tea', 'small'), ('coffee', 'small')]

total_price = 0
for order in orders:
    total_price += price[order]

# M3
student_skills = {
    'Tom': {'python', 'math'},
    'Pam': {'excel', 'math'},
    'Jim': {'python', 'excel', 'math'}
}
required_skills = {'python', 'math'}

eligible_students = []
for name, skills in student_skills.items():
    if required_skills.issubset(skills):
        eligible_students.append(name)

eligible_students.sort()

print('Mixed practice solutions created.')
````

# Unit 07 Cheat Sheet

## Container decision table

| Need | Use | Example |
|---|---|---|
| Ordered editable values | `list` | `['milk', 'bread']` |
| Ordered fixed record | `tuple` | `('Tom', 85)` |
| Unique values | `set` | `{'Tom', 'Pam'}` |
| Key-value lookup | `dict` | `{'coffee': 55}` |
| Many records with fixed fields | list of tuples | `[('Tom', 85), ('Pam', 92)]` |
| One key with many ordered values | dict with list values | `{'Tom': ['coffee', 'cake']}` |
| One key with many unique values | dict with set values | `{'Music': {'Tom', 'Pam'}}` |
| Lookup using multiple fields | dict with tuple keys | `{('coffee', 'large'): 60}` |

## Tuple cheat sheet

| Task | Code pattern | Meaning |
|---|---|---|
| Create tuple | `t = (1, 2, 3)` | stores ordered values |
| One-item tuple | `t = (1,)` | comma is required |
| First item | `t[0]` | index starts at 0 |
| Last item | `t[-1]` | negative index from end |
| Slice | `t[1:3]` | index 1 up to before 3 |
| Join tuples | `a + b` | returns new tuple |
| Repeat tuple | `t * 3` | returns new tuple |
| Unpack | `name, score = record` | split tuple into variables |
| Swap | `a, b = b, a` | exchange values |
| Replace item | `t[:i] + (new,) + t[i+1:]` | create a new tuple |

## Set cheat sheet

| Task | Code pattern | Meaning |
|---|---|---|
| Create set | `s = {1, 2, 3}` | unique values |
| Empty set | `s = set()` | `{}` is dictionary |
| Remove duplicates | `set(items)` | keeps unique values only |
| Sorted display | `sorted(s)` | returns sorted list |
| Add one item | `s.add(x)` | add value |
| Add many items | `s.update(items)` | add multiple values |
| Remove safely | `s.discard(x)` | no error if missing |
| Union | `A | B` | in A or B |
| Intersection | `A & B` | in both A and B |
| Difference | `A - B` | in A but not B |
| Symmetric difference | `A ^ B` | in exactly one side |
| Subset check | `small.issubset(big)` | all small items are in big |

## Dictionary cheat sheet

| Task | Code pattern | Meaning |
|---|---|---|
| Create dictionary | `d = {'a': 1}` | key-value pairs |
| Empty dictionary | `d = {}` | no key-value pairs yet |
| Access value | `d[key]` | error if missing |
| Safe access | `d.get(key, default)` | returns default if missing |
| Check key | `key in d` | True if key exists |
| Add/update | `d[key] = value` | add new or update old |
| Delete | `d.pop(key)` | remove key-value pair |
| Loop keys | `for k in d:` | loop through keys |
| Loop values | `for v in d.values():` | loop through values |
| Loop pairs | `for k, v in d.items():` | loop through key-value pairs |
| Count items | `count[x] += 1` after init | frequency counting |
| Group into lists | `d[key].append(value)` | keep ordered history |
| Group into sets | `d[key].add(value)` | keep unique group members |

> Note: in the set cheat sheet, union is normally written as `A | B`. If your keyboard or notebook font displays the vertical bar strangely, it means the same as `A.union(B)`.
