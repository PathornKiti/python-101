# 07.1 Tuple

A **tuple** is similar to a list because it stores many values in order.

The key difference:

```text
list  -> can be edited
tuple -> cannot be edited directly
```

Example:

```python
point = (3, 4)
```

This tuple can represent a point with `x = 3` and `y = 4`.

Use a tuple when:

- order matters,
- the group of values should stay together,
- the values should not be changed accidentally.


## 1. Creating tuples

Common ways to create tuples:

```python
t = (1, 2, 3)
t = tuple([1, 2, 3])
t = tuple('abc')
empty_tuple = ()
```

Important beginner warning:

```python
one_item = (1,)
not_tuple = (1)
```

`(1,)` is a tuple with one item.  
`(1)` is just the number `1` inside parentheses.


````python
a = (1, 2, 3)
b = tuple([1, 2, 3])
c = tuple('abc')
empty_tuple = ()

one_item = (1,)
not_tuple = (1)

print(a)
print(b)
print(c)
print(empty_tuple)
print(type(one_item))
print(type(not_tuple))
````

## 2. Tuple indexing and slicing

Tuple indexing works like list indexing.

```text
letters = ('A', 'B', 'C', 'D')
            0    1    2    3
```

Useful patterns:

```python
letters[0]      # first item
letters[-1]     # last item
letters[1:3]    # from index 1 up to before index 3
```


````python
letters = ('A', 'B', 'C', 'D')

print(letters[0])
print(letters[-1])
print(letters[1:3])
````

## 3. Tuple cannot be edited directly

A list can be edited:

```python
x = [10, 20, 30]
x[0] = 99
```

A tuple cannot be edited directly:

```python
t = (10, 20, 30)
t[0] = 99      # error
```

If we want a changed tuple, we create a **new tuple** from old pieces.


````python
numbers = (10, 20, 30, 40)

# Replace the item at index 1 with 99 by creating a new tuple.
updated_numbers = numbers[:1] + (99,) + numbers[2:]

print(numbers)
print(updated_numbers)
````

## 4. Tuple `+` and `*`

Tuple operators are similar to list operators.

| Operator | Meaning |
|---|---|
| `+` | join tuples |
| `*` | repeat tuple |

The result is a **new tuple**.


````python
a = (1, 2)
b = (3, 4)

joined = a + b
repeated = a * 3

print(joined)
print(repeated)
````

## 5. Tuple packing and unpacking

Python can pack several values into one tuple:

```python
person = 'Tom', 20, 'Engineering'
```

Python can also unpack a tuple into separate variables:

```python
name, age, faculty = person
```

The number of variables must match the number of tuple items.


````python
person = 'Tom', 20, 'Engineering'
print(person)

name, age, faculty = person

print(name)
print(age)
print(faculty)
````

## 6. Swapping variables

Tuple unpacking makes swapping variables very easy.

```python
a, b = b, a
```

This means:

1. pack the right side values,
2. unpack them into the left side variables.


````python
a = 5
b = 9

print("Before:", a, b)

a, b = b, a

print("After:", a, b)
````

## 7. List of tuples

A common pattern is a **list of tuples**.

Example: each tuple stores one student's name and score.

```python
students = [('Tom', 85), ('Pam', 92), ('Jim', 78)]
```

The outer container is a list because we may add or remove students.  
The inner container is a tuple because each record keeps related values together.


````python
students = [('Tom', 85), ('Pam', 92), ('Jim', 78)]

for name, score in students:
    print(name, score)
````

## 8. Tuple of lists

A tuple cannot be edited directly, but it can contain lists.

```python
data = ([1, 2], [3, 4])
```

This is not allowed:

```python
data[0] = [9, 9]
```

But this is allowed:

```python
data[0][1] = 99
```

Why? Because the tuple still contains the same list object, but the list inside is edited.


````python
data = ([1, 2], [3, 4])

print("Before:", data)

data[0][1] = 99

print("After:", data)
````

## 9. Common tuple mistakes

### Mistake 1: forgetting the comma in a one-item tuple

```python
x = (5)     # int
x = (5,)    # tuple
```

### Mistake 2: trying to append to a tuple

```python
t.append(9)     # error
```

Tuples do not have `append`, `insert`, `pop`, or item assignment.

### Mistake 3: unpacking into the wrong number of variables

```python
a, b = (1, 2, 3)    # error
```

The left side must have the same number of variables as the tuple items.

# 07.2 Set

A **set** stores unique values.

Important ideas:

- duplicates are automatically removed,
- membership checking with `in` is convenient,
- a set has no fixed order,
- a set has no index,
- set operations are useful for comparing groups.

Example:

```python
s = {1, 2, 2, 3}
```

The duplicate `2` is removed.


## 1. Creating sets

Common ways to create sets:

```python
s = {1, 2, 3}
s = set()
s = set([1, 2, 2, 3])
s = set('banana')
```

Important beginner warning:

```python
x = {}
```

This creates an empty **dictionary**, not an empty set.

To create an empty set, use:

```python
x = set()
```


````python
a = {1, 2, 3}
b = set()
c = set([1, 2, 2, 3])
d = set('banana')

print(a)
print(b)
print(c)
print(d)
print(type({}))
print(type(set()))
````

## 2. Set removes duplicates

This is one of the most common uses of `set`.

If we have a list with repeated values:

```python
numbers = [3, 1, 2, 3, 2, 1]
```

we can remove duplicates with:

```python
unique_numbers = set(numbers)
```

If we want ordered display, use:

```python
sorted(unique_numbers)
```


````python
numbers = [3, 1, 2, 3, 2, 1]

unique_numbers = set(numbers)
ordered_unique_numbers = sorted(unique_numbers)

print(unique_numbers)
print(ordered_unique_numbers)
````

## 3. Membership with `in`

Use `in` to check whether a value exists in a set.

```python
if value in registered_students:
    print('Found')
```

This is useful when you only care whether a value exists.


````python
registered_students = {'Tom', 'Pam', 'Jim'}

name = 'Pam'

if name in registered_students:
    print(name, "is registered")
else:
    print(name, "is not registered")
````

## 4. Adding and removing set items

Useful set methods:

```python
s.add(value)
s.update(other_values)
s.discard(value)
```

`discard` is beginner-friendly because it does not crash when the value is not found.


````python
students = {'Tom', 'Pam'}

students.add('Jim')
students.update(['May', 'Ann'])
students.discard('Pam')
students.discard('Unknown')  # no error

print(students)
````

## 5. Set operations

Set operations help compare groups.

Assume:

```python
A = {1, 2, 3}
B = {3, 4, 5}
```

| Operation | Meaning | Result |
|---|---|---|
| `A.union(B)` or `A | B` | in A or B | `{1, 2, 3, 4, 5}` |
| `A.intersection(B)` or `A & B` | in both A and B | `{3}` |
| `A.difference(B)` or `A - B` | in A but not B | `{1, 2}` |
| `A.symmetric_difference(B)` or `A ^ B` | in only one side | `{1, 2, 4, 5}` |


````python
A = {1, 2, 3}
B = {3, 4, 5}

print("union:", A | B)
print("intersection:", A & B)
print("A - B:", A - B)
print("B - A:", B - A)
print("symmetric difference:", A ^ B)
````

## 6. Subset and superset

Use these when you want to ask whether one group is fully inside another group.

```python
small.issubset(big)
big.issuperset(small)
```


````python
required = {'python', 'math'}
student_skills = {'python', 'math', 'excel'}

print(required.issubset(student_skills))
print(student_skills.issuperset(required))
````

## 7. Sets have no index

A set has no guaranteed order.

This means this is wrong:

```python
s[0]
```

Use a list when you need position.  
Use a set when you need unique values or membership checking.


## 8. Common set mistakes

### Mistake 1: using `{}` for an empty set

```python
s = {}      # dictionary
s = set()   # empty set
```

### Mistake 2: expecting set order

Sets are not for ordered data.

### Mistake 3: indexing a set

```python
s[0]   # error
```

### Mistake 4: adding a list into a set

```python
s.add([1, 2])   # error
```

A list is editable, so it cannot be stored as a set item.

# 07.3 Dictionary

A **dictionary** stores data as key-value pairs.

```python
grades = {'5931111121': 'A', '5932222221': 'B+'}
```

Here:

- key = student ID
- value = grade

Use a dictionary when you want to ask:

```text
Given this key, what is the value?
```

Example:

```python
grades['5931111121']
```


## 1. Creating dictionaries

Common ways to create dictionaries:

```python
d = {}
d = dict()
d = {'JAN': 1, 'FEB': 2, 'MAR': 3}
```

Each pair is written as:

```python
key: value
```


````python
months = {'JAN': 1, 'FEB': 2, 'MAR': 3}
empty_dict = {}

print(months)
print(empty_dict)
print(months['JAN'])
````

## 2. Accessing values by key

Use square brackets with the key:

```python
d[key]
```

If the key does not exist, Python gives an error.

Safer beginner patterns:

```python
if key in d:
    print(d[key])
else:
    print('Not Found')
```

or:

```python
d.get(key, default_value)
```


````python
grades = {'001': 'A', '002': 'B+', '003': 'C'}
sid = '002'

if sid in grades:
    print(grades[sid])
else:
    print('Not Found')

print(grades.get('999', 'Not Found'))
````

## 3. Adding, updating, and deleting dictionary data

Add or update:

```python
d[key] = value
```

Delete:

```python
d.pop(key)
```

Check before deleting if you are not sure the key exists.


````python
grades = {'001': 'A', '002': 'B+'}

grades['003'] = 'C'      # add
grades['002'] = 'A-'     # update

if '001' in grades:
    grades.pop('001')    # delete

print(grades)
````

## 4. Looping through a dictionary

Useful dictionary views:

```python
d.keys()
d.values()
d.items()
```

Most of the time, `d.items()` is easiest when you need both key and value.


````python
scores = {'Tom': 85, 'Pam': 92, 'Jim': 78}

print("Keys:")
for name in scores.keys():
    print(name)

print("Values:")
for score in scores.values():
    print(score)

print("Items:")
for name, score in scores.items():
    print(name, score)
````

## 5. Sorting dictionary output

A dictionary is for lookup, not mainly for sorted display.

To display by sorted keys:

```python
for key in sorted(d.keys()):
    print(key, d[key])
```

To display by sorted values, create temporary tuples and sort them.


````python
scores = {'Tom': 85, 'Pam': 92, 'Jim': 78}

print("Sorted by name:")
for name in sorted(scores.keys()):
    print(name, scores[name])

print("Sorted by score from high to low:")
temp = []
for name, score in scores.items():
    temp.append((-score, name))

temp.sort()

for negative_score, name in temp:
    print(name, -negative_score)
````

## 6. Counting with a dictionary

A common dictionary task is counting frequency.

Example: count characters in a string.

Beginner pattern:

```python
count = {}
for ch in text:
    if ch not in count:
        count[ch] = 1
    else:
        count[ch] += 1
```

Why do we check first?  
Because the key must exist before we can add `1` to its value.


````python
text = 'banana'

count = {}
for ch in text:
    if ch not in count:
        count[ch] = 1
    else:
        count[ch] += 1

print(count)
````

## 7. Dictionary with list values

Sometimes one key has many values.

Example:

- key = customer name
- value = list of phone models bought in order

```python
history = {
    'Tom': ['iPhone 13', 'iPhone 15'],
    'Pam': ['Galaxy S22']
}
```

Use a list as the value when order matters and duplicates may be meaningful.


````python
history = {}

records = [
    ('Tom', 'iPhone 13'),
    ('Pam', 'Galaxy S22'),
    ('Tom', 'iPhone 15'),
    ('Jim', 'OPPO Reno')
]

for name, model in records:
    if name not in history:
        history[name] = []
    history[name].append(model)

print(history)
````

## 8. Dictionary with set values

Use a set as the value when each key has many values but duplicates should be removed.

Example:

- key = faculty
- value = set of student names

```python
faculty_to_names = {
    'Engineering': {'Tom', 'Jim'},
    'Arts': {'Pam', 'Tom'}
}
```


````python
records = [
    ('Tom', 'Engineering'),
    ('Pam', 'Arts'),
    ('Jim', 'Engineering'),
    ('Tom', 'Engineering')
]

faculty_to_names = {}

for name, faculty in records:
    if faculty not in faculty_to_names:
        faculty_to_names[faculty] = set()
    faculty_to_names[faculty].add(name)

print(faculty_to_names)
````

## 9. Inverting a dictionary of sets

Sometimes we have:

```python
student_to_courses = {
    '001': {'c001', 'c002'},
    '002': {'c002', 'c003'}
}
```

But we want:

```python
course_to_students = {
    'c001': {'001'},
    'c002': {'001', '002'},
    'c003': {'002'}
}
```

This is called **inverting** the dictionary because the relationship is flipped.


````python
student_to_courses = {
    '001': {'c001', 'c002'},
    '002': {'c002', 'c003'},
    '003': {'c001'}
}

course_to_students = {}

for sid, courses in student_to_courses.items():
    for course in courses:
        if course not in course_to_students:
            course_to_students[course] = set()
        course_to_students[course].add(sid)

print(course_to_students)
````

## 10. Tuple keys in dictionaries

A dictionary key must be immutable.

Allowed keys include:

- numbers,
- strings,
- tuples of immutable values.

Not allowed keys include:

- lists,
- dictionaries,
- sets.

Example of a useful tuple key:

```python
height[(x, y)]
```

This means we use a coordinate pair as one key.


````python
height = {
    (0, 0): 10,
    (0, 1): 12,
    (1, 0): 15
}

point = (0, 1)

print(height[point])
print(height.get((9, 9), 'Not Found'))
````

## 11. Choosing between list, tuple, set, and dictionary

Use this quick decision guide.

| Situation | Good container |
|---|---|
| Keep phone models used by one customer from old to new | `list` |
| Store one coordinate `(x, y)` | `tuple` |
| Check whether a student is registered | `set` |
| Find grade from student ID | `dict` |
| Store many student records as `(name, score)` | `list` of `tuple` |
| Store many students per faculty without duplicates | `dict` with `set` values |
| Store many phone purchases per customer in order | `dict` with `list` values |


## 12. Common dictionary mistakes

### Mistake 1: accessing a missing key

```python
d['missing']   # error
```

Use `if key in d` or `d.get(key, default)`.

### Mistake 2: counting without initializing

```python
count[ch] += 1   # error if ch is new
```

Check first.

### Mistake 3: using a list as a dictionary key

```python
d[[1, 2]] = 'value'   # error
```

Use a tuple instead:

```python
d[(1, 2)] = 'value'
```

### Mistake 4: forgetting that keys are unique

```python
d = {'A': 1, 'A': 2}
```

The last value replaces the earlier value.

## Mixed Example: Student registry query

Problem idea:

You receive records of `(name, faculty)`.  
Then you receive a list of faculties to ask about.  
Return all student names in those faculties, without duplicates, sorted alphabetically.

Good data structure:

```python
faculty_to_names = {
    faculty: set_of_names
}
```

Why?

- Records are tuple pairs.
- Many records are stored in a list.
- A faculty maps to many names, so dictionary is useful.
- Names should not duplicate, so set is useful.
- Final answer should be ordered, so convert to sorted list.


````python
records = [
    ('Tom', 'Engineering'),
    ('Pam', 'Arts'),
    ('Jim', 'Engineering'),
    ('May', 'Science'),
    ('Ann', 'Engineering'),
    ('Tom', 'Engineering')
]

ask_faculties = ['Engineering', 'Science']

faculty_to_names = {}

for name, faculty in records:
    if faculty not in faculty_to_names:
        faculty_to_names[faculty] = set()
    faculty_to_names[faculty].add(name)

answer = set()
for faculty in ask_faculties:
    if faculty in faculty_to_names:
        answer = answer | faculty_to_names[faculty]

answer = sorted(answer)

print(answer)
````

````python
# Reference solution code: Mixed Practice

# M1
basket = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

unique_items = set(basket)
ordered_unique_items = sorted(unique_items)

item_count = {}
for item in basket:
    if item not in item_count:
        item_count[item] = 1
    else:
        item_count[item] += 1

# M2
student_skills = {
    'Tom': {'python', 'math'},
    'Pam': {'excel', 'math'},
    'Jim': {'python', 'excel', 'math'}
}
required_skills = {'python', 'math'}

eligible_students = []
for name, skills in student_skills.items():
    if required_skills.issubset(skills):
        eligible_students.append(name)
eligible_students.sort()

# M3
records = [
    ('apple', 'small', 2),
    ('apple', 'large', 3),
    ('bread', 'regular', 4)
]

price_lookup = {}
for x, y, value in records:
    price_lookup[(x, y)] = value

lookup_result = price_lookup.get((0, 1), 'Not Found')

print(unique_items)
print(item_count)
print(ordered_unique_items)
print(eligible_students)
print(price_lookup)
print(lookup_result)
````
