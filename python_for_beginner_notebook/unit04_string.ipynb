## 1. What is a string?

A **string** is text.

Examples:
- a name: `"Maya"`
- a phone number written as text: `"0812345678"`
- an email: `"maya@example.com"`
- a message: `"See you tomorrow!"`

Important idea:

| Value | Type | Meaning |
|---|---|---|
| `25` | integer | a number for calculation |
| `"25"` | string | text that looks like a number |

Use `str(x)` when you want to turn a number into text.


````python
age = 16
name = "Maya"

print(type(age))
print(type(name))
print("My name is " + name)
print("I am " + str(age) + " years old")
````

## 2. Quotes in strings

Python allows single quotes and double quotes.

Use the type of quote that makes your text easier to write.


````python
s1 = 'Hello'
s2 = "Hello"
s3 = "I'm learning Python"
s4 = 'She said "Hi"'

print(s1)
print(s2)
print(s3)
print(s4)
````

## 3. Length and indexing

`len(s)` counts how many characters are in a string.

If `s = "Python"`:

| Expression | Result | Meaning |
|---|---:|---|
| `len(s)` | `6` | total characters |
| `s[0]` | `'P'` | first character |
| `s[1]` | `'y'` | second character |
| `s[-1]` | `'n'` | last character |
| `s[-2]` | `'o'` | second from the end |

Indexing starts from `0`, not `1`.


````python
s = "Python"

print("length:", len(s))
print("first character:", s[0])
print("second character:", s[1])
print("last character:", s[-1])
print("second last character:", s[-2])
````

### Daily-life example: student ID

Suppose a student ID is `"STU2026007"`.

We may want to read parts of it:
- first 3 characters = prefix,
- middle 4 characters = year,
- last 3 characters = running number.


````python
student_id = "STU2026007"

prefix = student_id[:3]
year = student_id[3:7]
number = student_id[-3:]

print("prefix:", prefix)
print("year:", year)
print("number:", number)
````

## 4. Slicing

Slicing means taking part of a string.

General pattern:

```python
s[start:stop]
```

Python takes characters from `start` up to **before** `stop`.

Useful patterns:

| Pattern | Meaning |
|---|---|
| `s[:4]` | first 4 characters |
| `s[4:]` | from index 4 to the end |
| `s[-4:]` | last 4 characters |
| `s[::-1]` | reversed string |


````python
phone = "0812345678"

print("first 3 digits:", phone[:3])
print("last 4 digits:", phone[-4:])
print("hidden phone:", "*" * 6 + phone[-4:])
print("reversed:", phone[::-1])
````

## 5. Cleaning user input

User input often contains extra spaces or inconsistent letter cases.

Useful methods:

| Method | Example | Meaning |
|---|---|---|
| `strip()` | `"  hi  ".strip()` | remove outer spaces |
| `lower()` | `"YES".lower()` | convert to lowercase |
| `upper()` | `"yes".upper()` | convert to uppercase |

These methods return a **new string**. They do not change the old string unless you assign the result back.


````python
answer = "  YeS  "

clean_answer = answer.strip().lower()

print("original:", repr(answer))
print("cleaned :", repr(clean_answer))

if clean_answer == "yes":
    print("User agreed")
else:
    print("User did not agree")
````

## 6. Searching inside a string

Two common ways:

```python
"word" in text
```

and

```python
text.find("word")
```

| Method | Result |
|---|---|
| `"free" in message` | `True` or `False` |
| `message.find("free")` | index position, or `-1` if not found |


````python
message = "Free delivery today!"
clean_message = message.lower()

print("free" in clean_message)
print(clean_message.find("delivery"))
print(clean_message.find("pizza"))
````

## 7. Looping through characters

Use a `for` loop when you want to inspect each character.

Common daily-life uses:
- count digits in a note,
- count vowels in a word,
- check whether a password contains a number,
- build a cleaned version of text.


````python
note = "Order A12 costs 250 baht"

digit_count = 0
for ch in note:
    if "0" <= ch <= "9":
        digit_count += 1

print("number of digit characters:", digit_count)
````

## 8. Strings are immutable

Strings cannot be changed character by character.

This does **not** work:

```python
s = "cat"
s[0] = "b"
```

Instead, create a new string:

```python
s = "b" + s[1:]
```


````python
word = "cat"
new_word = "b" + word[1:]

print(word)
print(new_word)
````

## 9. Worked example A — clean a coupon code

### Problem
A user types a coupon code, but they may add spaces or use lowercase letters.

Return the cleaned coupon code.

### Example
`"  save50  "` → `"SAVE50"`


````python
def clean_coupon(code):
    return code.strip().upper()

print(clean_coupon("  save50  "))
print(clean_coupon(" NewUser10 "))
````

## 10. Worked example B — extract username from email

### Problem
Given an email address, return the part before `@`.

### Example
`"maya@example.com"` → `"maya"`


````python
def get_email_username(email):
    email = email.strip()
    at_pos = email.find("@")
    return email[:at_pos]

print(get_email_username("maya@example.com"))
print(get_email_username(" student01@school.ac.th "))
````

## 11. Worked example C — mask a phone number

### Problem
Show only the last 4 digits of a phone number.
All earlier digits should become `*`.

### Example
`"0812345678"` → `"******5678"`


````python
def mask_phone(phone):
    phone = phone.strip()
    hidden_part = "*" * (len(phone) - 4)
    visible_part = phone[-4:]
    return hidden_part + visible_part

print(mask_phone("0812345678"))
print(mask_phone("0998887777"))
````

## 12. Worked example D — count vowels in a message

### Problem
Count how many English vowels appear in a message.

Vowels are `a`, `e`, `i`, `o`, `u`.

### Example
`"I love Python"` → `4`


````python
def count_vowels(text):
    count = 0
    for ch in text.lower():
        if ch in "aeiou":
            count += 1
    return count

print(count_vowels("I love Python"))
print(count_vowels("Good morning"))
````

## 13. Worked example E — check message keyword

### Problem
A shop wants to know whether a customer message contains the word `refund`.
The search should ignore uppercase/lowercase differences.

### Example
`"I need a REFUND"` → `True`


````python
def contains_refund(text):
    return "refund" in text.lower()

print(contains_refund("I need a REFUND"))
print(contains_refund("Thank you for your help"))
````

## 14. Worked example F — convert date format

### Problem
Convert a date from `dd/mm/yyyy` to `yyyy-mm-dd`.

### Example
`"24/05/2026"` → `"2026-05-24"`


````python
def convert_date(date_text):
    date_text = date_text.strip()
    dd = date_text[:2]
    mm = date_text[3:5]
    yyyy = date_text[6:]
    return yyyy + "-" + mm + "-" + dd

print(convert_date("24/05/2026"))
print(convert_date("01/12/2025"))
````

## Solution 9 — Create simple username


````python
def solution_9_create_username(first_name, birth_year):
    name_part = first_name.strip().lower()[:3]
    year_part = str(birth_year)[-2:]
    return name_part + year_part

print(solution_9_create_username("Maya", 2008))
````

# Unit 04 String Cheat Sheet

| Goal | Python pattern | Example | Result / note |
|---|---|---|---|
| Create text | `s = "hello"` | `name = "Maya"` | Stores text |
| Count characters | `len(s)` | `len("Python")` | `6` |
| First character | `s[0]` | `"Python"[0]` | `'P'` |
| Last character | `s[-1]` | `"Python"[-1]` | `'n'` |
| First `n` characters | `s[:n]` | `"Python"[:3]` | `'Pyt'` |
| Last `n` characters | `s[-n:]` | `"0812345678"[-4:]` | `'5678'` |
| Reverse string | `s[::-1]` | `"abc"[::-1]` | `'cba'` |
| Remove outer spaces | `s.strip()` | `"  hi  ".strip()` | `'hi'` |
| Lowercase | `s.lower()` | `"YES".lower()` | `'yes'` |
| Uppercase | `s.upper()` | `"save".upper()` | `'SAVE'` |
| Check contains | `x in s` | `"free" in msg.lower()` | `True` / `False` |
| Find position | `s.find(x)` | `"abc".find("b")` | `1` |
| Not found | `s.find(x)` | `"abc".find("z")` | `-1` |
| Loop characters | `for ch in s:` | count digits | Reads one character at a time |
| Build new text | `result += ch` | cleaning text | Strings are immutable |
| Repeat text | `"*" * n` | `"*" * 4` | `'****'` |
| Join text | `a + b` | `"Hi " + "Maya"` | `'Hi Maya'` |
| Convert to string | `str(x)` | `"age " + str(16)` | avoids type error |

## Final checklist

Before moving on, students should be comfortable with:

- using indexing and slicing,
- cleaning input before checking it,
- using `lower()` for case-insensitive search,
- using `find()` before slicing around a special character,
- looping through text to count characters,
- creating a new string instead of changing one character directly.
