## Project Scenario

You are the programmer for a small space agency.

The final goal is to print a simple mission report.


# Part 1: Basic Mission Variables

This part uses:

- string
- integer
- float
- boolean
- `print()`
- `type()`


```python
# Basic mission information

ship_name = "Python Explorer"
captain_name = "Luna"
crew_count = 5
fuel_level = 78.5
oxygen_level = 91.2
mission_active = True

print("Ship:", ship_name)
print("Captain:", captain_name)
print("Crew count:", crew_count)
print("Fuel level:", fuel_level)
print("Oxygen level:", oxygen_level)
print("Mission active:", mission_active)

print("\nData types:")
print(type(ship_name))
print(type(crew_count))
print(type(fuel_level))
print(type(mission_active))
```

# Part 2: Mission Status with Selection

This part uses:

- `if`
- `elif`
- `else`
- comparison operators
- logical operators


```python
# Mission decision based on fuel and oxygen

if fuel_level >= 70 and oxygen_level >= 80:
    status = "Ready for deep space travel"
elif fuel_level >= 50 and oxygen_level >= 60:
    status = "Ready, but monitor resources carefully"
else:
    status = "Not ready. Refill resources first"

print("Mission status:", status)
```

# Part 3: String Cleaning

Mission messages may have extra spaces or inconsistent capital letters.

This part uses:

- `.strip()`
- `.lower()`
- `.upper()`
- `.replace()`
- `in`


```python
# Raw mission message

raw_message = "   warning: asteroid near sector-7   "

clean_message = raw_message.strip()
alert_message = clean_message.upper()
safe_message = clean_message.replace("asteroid", "space rock")

print("Raw message:", raw_message)
print("Clean message:", clean_message)
print("Alert message:", alert_message)
print("Safer message:", safe_message)

if "warning" in clean_message.lower():
    print("This message contains a warning.")
else:
    print("This message is normal.")
```

# Part 4: Lists and Loops

The spaceship will visit planets in order.

This part uses:

- list
- `for` loop
- `range()`
- accumulator variable


```python
# List of planned planets

planets = ["Mars", "Jupiter", "Saturn", "Neptune"]

print("Mission route:")

for i in range(len(planets)):
    print("Stop", i + 1, ":", planets[i])

# Count total stops manually
total_stops = 0

for planet in planets:
    total_stops = total_stops + 1

print("\nTotal planned stops:", total_stops)
```

# Part 5: Tuple, Set, and Dictionary

This part uses:

- tuple for coordinates
- set for unique planets
- dictionary for mission resources


```python
# Tuple: fixed coordinate for the spaceship starting point

start_coordinate = (0, 0, 0)

# Set: unique visited planets
visited_planets = {"Mars", "Jupiter", "Mars", "Saturn"}

# Dictionary: mission resources
resources = {
    "fuel": 78.5,
    "oxygen": 91.2,
    "food": 64.0
}

print("Start coordinate:", start_coordinate)
print("Visited planets:", visited_planets)
print("Resources:", resources)

print("\nFuel:", resources["fuel"])
print("Oxygen:", resources["oxygen"])
print("Food:", resources["food"])
```

# Part 6: File Handling

The program will save a mission log into a text file and read it back.

This part uses:

- `open()`
- write mode `"w"`
- read mode `"r"`
- `.write()`
- `.read()`


```python
# Write mission log to a file

log_text = ""
log_text = log_text + "Mission Log\n"
log_text = log_text + "Ship: " + ship_name + "\n"
log_text = log_text + "Captain: " + captain_name + "\n"
log_text = log_text + "Status: " + status + "\n"

with open("mission_log.txt", "w") as file:
    file.write(log_text)

print("Mission log saved.")
```

```python
# Read mission log from file

with open("mission_log.txt", "r") as file:
    saved_log = file.read()

print(saved_log)
```

# Part 7: Functions

Functions help us reuse code.

This part uses:

- `def`
- parameters
- return value


```python
def check_resource(resource_name, amount):
    if amount >= 70:
        result = resource_name + " is good"
    elif amount >= 40:
        result = resource_name + " is medium"
    else:
        result = resource_name + " is low"

    return result


print(check_resource("Fuel", resources["fuel"]))
print(check_resource("Oxygen", resources["oxygen"]))
print(check_resource("Food", resources["food"]))
```

# Part 8: Recursion

Recursion means a function calls itself.

Here, the countdown function calls itself until it reaches zero.


```python
def countdown(number):
    if number == 0:
        print("Launch!")
    else:
        print(number)
        countdown(number - 1)


countdown(5)
```

# Part 9: NumPy Mission Data

The spaceship records fuel levels for 7 mission days.

This part uses:

- `import numpy as np`
- NumPy array
- `.mean()`
- `.max()`
- `.min()`


```python
import numpy as np

fuel_by_day = np.array([78.5, 74.0, 69.5, 61.0, 55.5, 48.0, 43.5])
oxygen_by_day = np.array([91.2, 89.0, 86.5, 82.0, 78.5, 74.0, 70.5])

print("Fuel by day:", fuel_by_day)
print("Oxygen by day:", oxygen_by_day)

print("\nAverage fuel:", fuel_by_day.mean())
print("Highest fuel:", fuel_by_day.max())
print("Lowest fuel:", fuel_by_day.min())

print("\nAverage oxygen:", oxygen_by_day.mean())
print("Highest oxygen:", oxygen_by_day.max())
print("Lowest oxygen:", oxygen_by_day.min())
```

# Part 10: OOP Class

Now we organize the mission using classes.

This part uses:

- class
- `__init__`
- attributes
- methods
- object creation


```python
class Spaceship:
    def __init__(self, name, captain, fuel, oxygen):
        self.name = name
        self.captain = captain
        self.fuel = fuel
        self.oxygen = oxygen

    def show_info(self):
        print("Ship:", self.name)
        print("Captain:", self.captain)
        print("Fuel:", self.fuel)
        print("Oxygen:", self.oxygen)

    def check_status(self):
        if self.fuel >= 70 and self.oxygen >= 80:
            return "Excellent condition"
        elif self.fuel >= 50 and self.oxygen >= 60:
            return "Acceptable condition"
        else:
            return "Critical condition"


my_ship = Spaceship("Python Explorer", "Luna", 78.5, 91.2)

my_ship.show_info()
print("Status:", my_ship.check_status())
```

# Part 11: OOP with Inheritance

Inheritance lets one class reuse another class.

Here, `ResearchSpaceship` is a special type of `Spaceship`.


```python
class ResearchSpaceship(Spaceship):
    def __init__(self, name, captain, fuel, oxygen, research_topic):
        super().__init__(name, captain, fuel, oxygen)
        self.research_topic = research_topic

    def show_research_topic(self):
        print("Research topic:", self.research_topic)


research_ship = ResearchSpaceship(
    "Python Explorer",
    "Luna",
    78.5,
    91.2,
    "Searching for water on Mars"
)

research_ship.show_info()
research_ship.show_research_topic()
print("Status:", research_ship.check_status())
```

# Final Capstone Challenge

Now combine everything into one final mission report.

Your report should include:

1. Ship name
2. Captain name
3. Mission route
4. Unique visited planets
5. Resource status
6. Average fuel and oxygen
7. Final mission decision


```python
# Final Mission Report

print("=" * 40)
print("FINAL SPACE MISSION REPORT")
print("=" * 40)

print("Ship:", my_ship.name)
print("Captain:", my_ship.captain)

print("\nMission route:")
for planet in planets:
    print("-", planet)

print("\nUnique visited planets:")
for planet in visited_planets:
    print("-", planet)

print("\nResource status:")
for resource_name in resources:
    print("-", check_resource(resource_name.title(), resources[resource_name]))

average_fuel = fuel_by_day.mean()
average_oxygen = oxygen_by_day.mean()

print("\nAverage fuel:", average_fuel)
print("Average oxygen:", average_oxygen)

if average_fuel >= 60 and average_oxygen >= 75:
    final_decision = "Mission successful"
elif average_fuel >= 45 and average_oxygen >= 60:
    final_decision = "Mission completed, but resources were risky"
else:
    final_decision = "Mission failed because resources were too low"

print("\nFinal decision:", final_decision)

print("=" * 40)
```
