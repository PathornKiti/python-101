# 10.1 What is OOP?

**OOP** stands for **Object-Oriented Programming**.

The key idea is simple:

> Put related data and related actions together.

Without OOP, we may store data in many variables or dictionaries.
With OOP, we can create a meaningful object.

Example idea:

```python
wallet_owner = "Ann"
wallet_balance = 500
```

This works, but the data is separate.

With OOP, we can create:

```python
wallet = Wallet("Ann", 500)
```

Now the wallet can store its own data and perform its own actions.

## Class, object, attribute, and method

| Word | Meaning | Daily-life analogy |
|---|---|---|
| Class | A blueprint for creating objects | A form template |
| Object | One real thing created from a class | One completed form |
| Attribute | Data stored inside an object | Name, price, balance |
| Method | Function inside a class | Actions like pay, add, update |

Example:

```python
class Wallet:
    ...

my_wallet = Wallet("Ann", 500)
```

Here:

- `Wallet` is the class.
- `my_wallet` is an object.
- `owner` and `balance` are attributes.
- `add_money()` and `spend()` are methods.

## Procedural thinking vs OOP thinking

### Procedural style

You focus on functions and separate data.

```python
balance = 500
balance = balance + 100
balance = balance - 80
```

### OOP style

You focus on objects that know their own data and behavior.

```python
wallet = Wallet("Ann", 500)
wallet.add_money(100)
wallet.spend(80)
```

Both styles are useful.  
OOP becomes helpful when your program has many related pieces of data and actions.

# 10.3 Constructor: `__init__`

The method `__init__` runs automatically when we create an object.

It is often called a **constructor**.

The job of `__init__` is to prepare the object's attributes.

Pattern:

```python
class ClassName:
    def __init__(self, parameter1, parameter2):
        self.attribute1 = parameter1
        self.attribute2 = parameter2
```

Important idea:

- parameters are input values passed into the object
- attributes are values stored inside the object

````python
class Wallet:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

w1 = Wallet("Ann", 500)
w2 = Wallet("Ben", 1200)

print(w1.owner, w1.balance)
print(w2.owner, w2.balance)
````

## Why do we need `self`?

`self` means **the current object**.

When we write:

```python
w1.balance
```

we mean the `balance` attribute of object `w1`.

Inside the class, we write:

```python
self.balance
```

because the method must know **which object's balance** it is working with.

### Common beginner mistake

Incorrect:

```python
class Wallet:

```

This does not store data inside the object.

Correct:

```python
class Wallet:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
```

# 10.4 Methods

A **method** is a function written inside a class.

A method usually has `self` as the first parameter.

There are two common types of methods:

1. **Return methods**: calculate and return information.
2. **Update methods**: change the object's internal data.

````python
class Wallet:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def can_buy(self, price):
        return self.balance >= price

    def add_money(self, amount):
        self.balance += amount

    def spend(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return True
        return False

wallet = Wallet("Ann", 500)

print(wallet.can_buy(300))   # True
print(wallet.spend(300))     # True
print(wallet.balance)        # 200
print(wallet.spend(900))     # False
print(wallet.balance)        # still 200
wallet.add_money(100)
print(wallet.balance)        # 300
````

## Method call style

These two calls mean the same thing:

```python
wallet.spend(100)
```

and

```python
Wallet.spend(wallet, 100)
```

Usually, we use the first style because it is easier to read.

````python
wallet = Wallet("Ben", 1000)

print(wallet.spend(200))
print(Wallet.spend(wallet, 200))
print(wallet.balance)
````

# 10.5 Encapsulation: protect object behavior

**Encapsulation** means keeping related data and behavior inside the same object.

It also means we should update object data through methods, so we can control the rules.

Example rule:

> A wallet balance should not become negative.

Instead of changing `wallet.balance` directly, we use a method like `spend()`.

````python
wallet = Wallet("Cara", 100)

# Good style: use a method that checks the rule.
print(wallet.spend(150))
print(wallet.balance)

# Bad style: direct update can break the rule.
wallet.balance = -999
print(wallet.balance)
````

The code above shows why methods are useful.

In real projects, we often design methods to protect the object's data from invalid updates.

# 10.6 `__str__`: make an object readable

When we use `print(object)`, Python needs to know how to convert the object to text.

The special method `__str__` tells Python what string to show.

Without `__str__`, Python shows a memory-address-style result.

With `__str__`, the output becomes readable.

````python
class Wallet:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return self.owner + " has " + str(self.balance) + " baht"

wallet = Wallet("Ann", 500)
print(wallet)
print(str(wallet))
````

# 10.7 `__lt__`: compare and sort objects

`__lt__` means **less than**.

Python uses `__lt__` when we write:

```python
a < b
```

Python also uses it when sorting a list of objects.

Example rule:

> A wallet with less money should come first.

If two wallets have the same balance, sort by owner name.

````python
class Wallet:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return self.owner + " has " + str(self.balance) + " baht"

    def __lt__(self, other):
        if self.balance != other.balance:
            return self.balance < other.balance
        return self.owner < other.owner

wallets = [
    Wallet("Ben", 300),
    Wallet("Ann", 500),
    Wallet("Cara", 300),
]

wallets.sort()

for wallet in wallets:
    print(wallet)
````

## Why sorting works

When Python sorts the list, it compares two objects many times.

Each comparison calls `__lt__`.

For example:

```python
wallets[0] < wallets[1]
```

is similar to:

```python
wallets[0].__lt__(wallets[1])
```

# 10.8 Objects can store lists, sets, and dictionaries

An attribute can be any Python object.

A class can store:

- numbers
- strings
- lists
- tuples
- sets
- dictionaries
- other objects

Example: a `TodoList` object can store a list of tasks.

````python
class TodoList:
    def __init__(self, owner):
        self.owner = owner
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def complete_task(self, task):
        if task in self.tasks:
            self.tasks.remove(task)
            return True
        return False

    def pending_count(self):
        return len(self.tasks)

    def __str__(self):
        return self.owner + " has " + str(len(self.tasks)) + " pending task(s)"

my_list = TodoList("Ann")
my_list.add_task("finish homework")
my_list.add_task("buy milk")

print(my_list)
print(my_list.tasks)
print(my_list.complete_task("buy milk"))
print(my_list.tasks)
````

# 10.9 Composition: an object can contain another object

**Composition** means building a larger object from smaller objects.

Daily-life example:

- A `Product` object stores product information.
- A `ShoppingCart` object stores many products.

So a shopping cart **contains** product objects.

````python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return self.name + " (" + str(self.price) + " baht)"

class ShoppingCart:
    def __init__(self, owner):
        self.owner = owner
        self.items = []   # each item is [product, quantity]

    def add_product(self, product, quantity):
        for item in self.items:
            if item[0].name == product.name:
                item[1] += quantity
                return
        self.items.append([product, quantity])

    def total_price(self):
        total = 0
        for product, quantity in self.items:
            total += product.price * quantity
        return total

    def __str__(self):
        return self.owner + "'s cart total = " + str(self.total_price()) + " baht"

milk = Product("Milk", 45)
bread = Product("Bread", 30)

cart = ShoppingCart("Ann")
cart.add_product(milk, 2)
cart.add_product(bread, 3)
cart.add_product(milk, 1)

print(cart)
for product, quantity in cart.items:
    print(product, "x", quantity)
````

# 10.10 Abstraction

**Abstraction** means using something without needing to know every internal detail.

Example:

```python
cart.total_price()
```

The user does not need to know that the method loops through `self.items` and multiplies price by quantity.
The user only needs to know:

> Call `total_price()` to get the cart total.

This makes code easier to use.

# 10.11 Inheritance

**Inheritance** means one class can reuse and extend another class.

Example:

- `Member` is a general class.
- `StudentMember` is a more specific class.

A student member is still a member, but it may have a different discount rule.

Inheritance is useful when classes share many things but have small differences.

````python
class Member:
    def __init__(self, name):
        self.name = name

    def get_discount_rate(self):
        return 0.05

    def final_price(self, price):
        return price * (1 - self.get_discount_rate())

class StudentMember(Member):
    def get_discount_rate(self):
        return 0.10

normal = Member("Ann")
student = StudentMember("Ben")

print(normal.final_price(1000))
print(student.final_price(1000))
````

# 10.12 Polymorphism

**Polymorphism** means different object types can respond to the same method name.

In the previous example, both objects can call:

```python
final_price(1000)
```

But the result is different because `get_discount_rate()` is different.

This allows Python to work with many object types in a flexible way.

````python
members = [
    Member("Ann"),
    StudentMember("Ben"),
    Member("Cara"),
    StudentMember("Dan"),
]

for member in members:
    print(member.name, member.final_price(1000))
````

## Exercise 1 — Product Class

Create a class named `Product`.

Each product has:

- `name`
- `price`
- `quantity`

Required methods:

| Method | Requirement |
|---|---|
| `__init__(name, price, quantity)` | Store all attributes |
| `__str__()` | Return text like `Milk: 3 item(s), 45 baht each` |
| `total_value()` | Return `price * quantity` |
| `restock(n)` | Add `n` to quantity |
| `sell(n)` | If enough stock, reduce quantity and return `True`; otherwise return `False` |
| `__lt__(other)` | Sort by total value, then name |

Daily-life idea: this is useful for a small shop inventory.

````python
# Exercise 1 skeleton

class Product:
    def __init__(self, name, price, quantity):
        pass

    def __str__(self):
        pass

    def total_value(self):
        pass

    def restock(self, n):
        pass

    def sell(self, n):
        pass

    def __lt__(self, other):
        pass
````

<details>
<summary>Solution idea</summary>

Store product data using:

```python
self.name
self.price
self.quantity
```

For selling, check the stock before reducing quantity.

For sorting, compare `total_value()` first. If equal, compare product name.

</details>

````python
# Exercise 1 reference solution

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return self.name + ": " + str(self.quantity) + " item(s), " + str(self.price) + " baht each"

    def total_value(self):
        return self.price * self.quantity

    def restock(self, n):
        self.quantity += n

    def sell(self, n):
        if n <= self.quantity:
            self.quantity -= n
            return True
        return False

    def __lt__(self, other):
        if self.total_value() != other.total_value():
            return self.total_value() < other.total_value()
        return self.name < other.name
````

## Exercise 4 — Student Score Class

Create a class named `StudentScore`.

Each student score object has:

- `name`
- `scores`, which starts as an empty list

Required methods:

| Method | Requirement |
|---|---|
| `__init__(name)` | Store name and create empty score list |
| `add_score(score)` | Add score only if it is between 0 and 100; return `True` if added, otherwise `False` |
| `average()` | Return average score; if no score, return `0` |
| `status()` | Return `Pass` if average is at least 50; otherwise `Not pass` |
| `__str__()` | Return text like `Ann: 75.0 (Pass)` |
| `__lt__(other)` | Sort by average, then name |

Daily-life idea: a class can combine student data with grade logic.

````python
# Exercise 4 skeleton

class StudentScore:
    def __init__(self, name):
        pass

    def add_score(self, score):
        pass

    def average(self):
        pass

    def status(self):
        pass

    def __str__(self):
        pass

    def __lt__(self, other):
        pass
````

````python
# Exercise 4 reference solution

class StudentScore:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, score):
        if 0 <= score <= 100:
            self.scores.append(score)
            return True
        return False

    def average(self):
        if len(self.scores) == 0:
            return 0
        return sum(self.scores) / len(self.scores)

    def status(self):
        if self.average() >= 50:
            return "Pass"
        return "Not pass"

    def __str__(self):
        return self.name + ": " + str(self.average()) + " (" + self.status() + ")"

    def __lt__(self, other):
        if self.average() != other.average():
            return self.average() < other.average()
        return self.name < other.name
````

Create two classes:

- `items`, which starts as an empty list

[product, quantity]

Required methods for `ShoppingCart`:

| Method | Requirement |
|---|---|
| `add_product(product, quantity)` | Add product quantity; if product already exists, increase quantity |

| `total_price()` | Return total cart price |
| `__str__()` | Return text like `Ann's cart total = 120 baht` |
| `__lt__(other)` | Sort by total price, then owner name |

Daily-life idea: this is an example of **composition**, because a cart contains product objects.

````python
# Exercise 5 skeleton

class Product:
    def __init__(self, name, price):
        pass

    def __str__(self):
        pass

class ShoppingCart:
    def __init__(self, owner):
        pass

    def add_product(self, product, quantity):
        pass

    def remove_product(self, product_name):
        pass

    def total_price(self):
        pass

    def __str__(self):
        pass

    def __lt__(self, other):
        pass
````

````python
# Exercise 5 reference solution

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return self.name + " (" + str(self.price) + " baht)"

class ShoppingCart:
    def __init__(self, owner):
        self.owner = owner
        self.items = []

    def add_product(self, product, quantity):
        for item in self.items:
            if item[0].name == product.name:
                item[1] += quantity
                return
        self.items.append([product, quantity])

    def remove_product(self, product_name):
        for item in self.items:
            if item[0].name == product_name:
                self.items.remove(item)
                return True
        return False

    def total_price(self):
        total = 0
        for product, quantity in self.items:
            total += product.price * quantity
        return total

    def __str__(self):
        return self.owner + "'s cart total = " + str(self.total_price()) + " baht"

    def __lt__(self, other):
        if self.total_price() != other.total_price():
            return self.total_price() < other.total_price()
        return self.owner < other.owner
````

````python
# Exercise 5 auto-checks

milk = Product("Milk", 45)
bread = Product("Bread", 30)
egg = Product("Egg", 10)

cart1 = ShoppingCart("Ann")
cart1.add_product(milk, 2)
cart1.add_product(bread, 1)
cart1.add_product(milk, 1)

assert str(milk) == "Milk (45 baht)"
assert len(cart1.items) == 2
assert cart1.items[0][1] == 3
assert cart1.total_price() == 165
assert str(cart1) == "Ann's cart total = 165 baht"
assert cart1.remove_product("Bread") == True
assert cart1.total_price() == 135
assert cart1.remove_product("Coffee") == False

cart2 = ShoppingCart("Ben")
cart2.add_product(egg, 20)
assert [cart.owner for cart in sorted([cart2, cart1])] == ["Ann", "Ben"]

print("Exercise 5 passed")
````

````python
# Exercise 6 reference solution

class Member:
    def __init__(self, name):
        self.name = name

    def get_discount_rate(self):
        return 0.05

    def final_price(self, price):
        return price * (1 - self.get_discount_rate())

class StudentMember(Member):
    def get_discount_rate(self):
        return 0.10
````

## OOP concepts

| Concept | Meaning | Example |

| Class | Blueprint for creating objects | `class Wallet:` |
| Object | One real item created from a class | `w = Wallet("Ann", 500)` |

| Polymorphism | Different classes use the same method name differently | `final_price()` |

## Common class syntax

| Purpose | Pattern |
|---|---|
| Create class | `class ClassName:` |
| Constructor | `def __init__(self, x):` |
| Create attribute | `self.x = x` |
| Create method | `def method_name(self):` |
| Create object | `obj = ClassName(value)` |
| Access attribute | `obj.x` |
| Call method | `obj.method_name()` |
| Store object in list | `items = [obj1, obj2]` |
| Sort objects | define `__lt__`, then use `sorted(items)` |

## Special methods

| Method | Used for | Example |
|---|---|---|
| `__init__` | Object creation | `Wallet("Ann", 500)` |
| `__str__` | Readable printing | `print(wallet)` |
| `__lt__` | Less-than comparison and sorting | `wallet1 < wallet2` |
| `__eq__` | Equality comparison | `a == b` |

## Method design patterns

| Method type | What it does | Example |
|---|---|---|
| Getter-style method | Returns information without changing object | `total_price()` |
| Update method | Changes object data | `add_money()` |
| Validation method | Checks whether action is allowed | `can_buy()` |
| Action method | Does something and returns success/failure | `spend()` |

## Beginner mistakes and fixes

| Mistake | Why it is wrong | Fix |
|---|---|---|
| Forget `self` | Python does not know the current object | Use `def method(self):` |
| Forget `self.` | Data is not stored in the object | Use `self.name = name` |
| Change data directly | Can break rules | Use methods like `spend()` |
| Call method without parentheses | Method does not run | Use `obj.method()` |
| Compare objects without `__lt__` | Python does not know sorting rule | Define `__lt__` |
| Use same object accidentally | Two names point to one object | Create a new object with constructor |

## When should we use a class?

Use a class when:

- one thing has many related attributes
- the thing also has actions
- you want to reuse the same structure many times
- you want cleaner and more organized code

| Student profile | `Student` |
| Wallet | `Wallet` |
| Product inventory | `Product` |
