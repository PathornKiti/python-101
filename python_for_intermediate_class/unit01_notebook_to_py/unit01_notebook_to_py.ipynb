# Unit 01: From Notebook to `.py` Files

In beginner Python, notebooks are a friendly place to learn because you can mix code, output, and Markdown explanations. In intermediate Python, you also need to learn how to move important code into `.py` files so it can become reusable project code.


## Learning goals

By the end of this lesson, you should be able to:

- explain when a notebook is useful;
- explain when a `.py` file is better;
- turn notebook cells into functions;
- write a `main()` function;
- use `if __name__ == "__main__":`;
- import script functions back into a notebook.


## Why notebooks are useful

Use a notebook when you are exploring, learning, or explaining. Notebooks are good for:

1. trying code one small step at a time;
2. showing charts, tables, and outputs beside the code;
3. writing Markdown explanations for students or readers;
4. experimenting before you know the final structure of the program.


## Why `.py` files are important

Move code to a `.py` file when the code becomes stable, repeated, or important. Python files are better for:

- **reuse**: import the same function in many places;
- **testing**: test functions with tools such as `pytest`;
- **automation**: run scripts from the terminal or a scheduler;
- **Git review**: text files are easier to compare than notebook JSON;
- **team projects**: organize code into modules and packages;
- **clean notebooks**: keep notebooks focused on explanation and results.


## Notebook-style code before conversion

Notebook code often starts as a direct sequence of cells. This is okay for exploration, but it can become hard to reuse if the variables are hidden across many cells.


```python
student_name = "Maya"
scores = [88, 92, 79, 95]

average = sum(scores) / len(scores)

if average >= 90:
    grade = "A"
elif average >= 80:
    grade = "B"
elif average >= 70:
    grade = "C"
elif average >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Student: {student_name}")
print(f"Average: {average:.2f}")
print(f"Grade: {grade}")
```

## Step 1: Find reusable pieces

Ask: What part of this code might I need again? In this example, average calculation, letter-grade assignment, and report building are reusable pieces. Each reusable idea can become a function.


```python
def calculate_average(scores):
    if not scores:
        raise ValueError("scores must contain at least one number")
    return sum(scores) / len(scores)


def assign_letter_grade(average):
    if average >= 90:
        return "A"
    if average >= 80:
        return "B"
    if average >= 70:
        return "C"
    if average >= 60:
        return "D"
    return "F"
```

## Step 2: Move the functions into a `.py` file

The file `notebook_to_py_example.py` contains the script version of this lesson. Notice these best practices:

- functions have clear names;
- function inputs are parameters;
- functions return values instead of depending only on `print`;
- docstrings explain what functions do;
- `main()` contains the script workflow;
- `if __name__ == "__main__": main()` runs the workflow only when the file is executed directly.


## Step 3: Import script functions back into the notebook

After moving logic into a `.py` file, the notebook can stay clean. The notebook can import functions, run examples, and explain results.


```python
from notebook_to_py_example import build_student_report

print(build_student_report("Maya", [88, 92, 79, 95]))
```

## Best-practice checklist

When converting notebook code to `.py`, use this checklist:

- Give every important operation a clear function name.
- Keep functions small and focused on one job.
- Use parameters instead of hard-coded notebook variables.
- Return values so functions can be reused and tested.
- Use `print()` mostly in `main()` or for final user output.
- Add docstrings for functions that students or teammates will use.
- Add type hints when they make the code easier to understand.
- Avoid depending on notebook execution order.
- Run the `.py` file from the terminal to confirm it works outside Jupyter.
- Import the `.py` file into the notebook to keep learning materials clean.
