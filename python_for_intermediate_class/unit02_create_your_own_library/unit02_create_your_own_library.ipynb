# Unit 02: Create Your Own Library with a Data Model

In this notebook, you will learn how to import your own reusable Python code from a package. The package for this lesson is named `student_learning_library`.


## Learning Goals

By the end of this notebook, you should be able to:

- import a function from your own library;
- import a class from your own library;
- import a data model from your own library;
- use the same library code from a notebook and from a `.py` script.


## Folder Structure

The reusable library code is kept in a package folder:

```text
student_learning_library/
  __init__.py
  learning_tools.py
```

The notebook and `demo_imports.py` file should import from the package instead of copying the library code.


````python
from student_learning_library import StudyLibrary, StudySession, estimate_study_points
````

## 1. Import and Use a Function

A function is good for one reusable action. Here, `estimate_study_points()` calculates study points from minutes and completion status.


````python
points = estimate_study_points(minutes=45, completed=True)
points
````

## 2. Import and Use a Data Model

A data model stores related information in a clear structure. This unit uses a `@dataclass` named `StudySession`.


````python
session = StudySession(topic="Library imports", minutes=45, completed=True)
session
````

## 3. Import and Use a Class

A class can store data and provide behavior. `StudyLibrary` stores many sessions and creates a report.


````python
library = StudyLibrary(student_name="Maya")
library.add_session(session)
library.add_session(StudySession(topic="Data models", minutes=30, completed=False))

print(library.build_report())
````

## 4. Why This Matters

When your reusable logic is in a library, you can use it in many places:

- notebooks for learning and explanation;
- scripts for running a workflow;
- tests for checking correctness;
- apps for building user-facing projects.
