## 1. String literals and quotes

A string is a sequence of characters.  
Strings can contain letters, digits, spaces, and symbols.

Common ways to write strings in Python:
- single quotes: `'hello'`
- double quotes: `"hello"`
- triple single quotes: `'''hello'''`
- triple double quotes: `"""hello"""`

Important:
- `12` is an integer
- `'12'` is a string
- use `str(x)` to convert a value to a string


````python
a = 'I am "Python".'
b = "I'm Python."
c = '''I'm "Python".'''
d = """I'm "Python"."""

print(a)
print(b)
print(c)
print(d)
print(type('12'))
print(type(12))
print(str(12) + ' apples')
````

## 2. Length and indexing

If `s = "ABCDEFG"` then:
- `len(s)` is `7`
- `s[0]` is `'A'`
- `s[1]` is `'B'`
- `s[-1]` is `'G'`
- `s[-2]` is `'F'`

Valid indices:
- from left: `0` to `len(s)-1`
- from right: `-1` to `-len(s)`

Out-of-range indexing causes an error.


````python
s = "ABCDEFG"

print("len(s) =", len(s))
print("s[0] =", s[0])
print("s[-1] =", s[-1])
print("s[2] =", s[2])
print("s[-3] =", s[-3])
````

### A quick note on invalid indexing

Indexing must stay within the allowed range.  
Slicing is forgiving, but direct indexing is not.


````python
s = "01234"

print(s[0], s[4])

try:
    print(s[5])
except Exception as e:
    print(type(e).__name__, ":", e)

try:
    print(s[-6])
except Exception as e:
    print(type(e).__name__, ":", e)
````

## 3. Slicing

Slicing lets us take part of a string.

General form:
```python
s[start:stop:step]
```

Useful patterns:
- `s[:]` gives the whole string
- `s[2:5]` gives characters at index `2, 3, 4`
- `s[:4]` gives the first 4 characters
- `s[3:]` gives everything from index `3` onward
- `s[::2]` gives every second character
- `s[::-1]` reverses the string

Unlike direct indexing, slicing does **not** fail when the bounds are too large.


````python
s = "ABCDEFG"

print("s[:]    ->", s[:])
print("s[2:5]  ->", s[2:5])
print("s[:4]   ->", s[:4])
print("s[3:]   ->", s[3:])
print("s[::2]  ->", s[::2])
print("s[1::2] ->", s[1::2])
print("s[::-1] ->", s[::-1])
````

### Safe slicing examples

These are valid even when the bounds are far outside the string.


````python
s = "01234"

print(s[2:50000])
print(s[4999:50000])
print(s[-500:-2])
print(s[-3:-500:-1])
print(s[-500:-300])
````

## 4. Iterating through strings

A very common pattern is:

```python
for c in s:
    ...
```

This processes one character at a time from left to right.


````python
s = "Python"

for c in s:
    print(c)
````

### Iterate with index too

When you need both the position and the character:
- use `range(len(s))`, or
- use `enumerate(s)`


````python
s = "Python"

print("Using range(len(s))")
for i in range(len(s)):
    print(i, s[i])

print("\nUsing enumerate(s)")
for i, c in enumerate(s):
    print(i, c)
````

### Iterate from right to left

Two common approaches:
- `for c in s[::-1]`
- loop with negative indices


````python
s = "Python"

print("Reverse using slicing")
for c in s[::-1]:
    print(c, end=" ")
print()

print("Reverse using negative indices")
for k in range(-1, -(len(s)+1), -1):
    print(s[k], end=" ")
print()
````

## 5. Common string methods

For this unit, the key methods are:

- `s.upper()`
- `s.lower()`
- `s.strip()`
- `s.find(t)`
- `s.find(t, start)`

These methods return a new string or a new value.  
They do not change the original string unless you assign the result back.


````python
s = "  Python 3.6  "

print("original     :", repr(s))
print("upper()      :", repr(s.upper()))
print("lower()      :", repr(s.lower()))
print("strip()      :", repr(s.strip()))
print("find('th')   :", s.find("th"))
print("find('x')    :", s.find("x"))
print("find('o', 4) :", s.find("o", 4))

s = s.strip().upper()
print("reassigned   :", repr(s))
````

### `find` and `in`

These two checks are closely related:

```python
if t in s:
    ...
```

and

```python
if s.find(t) >= 0:
    ...
```

Both test whether `t` appears in `s`.


````python
s = "engineering"

print(s.find("ng"))
print("ng" in s)
print("xy" in s)
print(s.find("xy"))
````

## 7. Strings are immutable

You cannot change one character in place.

This is invalid:

```python
s[2] = 'a'
```

Instead, build a new string:
```python
s = s[:2] + 'a' + s[3:]
```


````python
s = "Python"
print("before:", s)

s = s[:2] + 'A' + s[3:]
print("after :", s)
````

## 9. Common string-processing patterns

This section collects the patterns students should know well.


### 9.1 Count digits in a string


````python
s = "room 24B, floor 7"

c = 0
for e in s:
    if '0' <= e <= '9':
        c += 1

print(c)
````

### 9.2 Process adjacent pairs

When you need to look at `s[k]` and `s[k+1]`, loop only to `len(s)-2`:

```python
for k in range(len(s)-1):
    ...
```


````python
s = "a1b23c456"

c = 0
for k in range(len(s)-1):
    if '0' <= s[k] <= '9' and '0' <= s[k+1] <= '9':
        c += 1

print(c)
````

### 9.3 Count occurrences of a substring

Two standard ways:
1. repeated `find`
2. sliding window with slicing

These count **overlapping** occurrences too.


````python
def count_occurrences_find(s, t):
    c = 0
    k = s.find(t)
    while k >= 0:
        c += 1
        k = s.find(t, k + 1)
    return c

def count_occurrences_slice(s, t):
    c = 0
    for i in range(len(s) - len(t) + 1):
        if s[i:i+len(t)] == t:
            c += 1
    return c

s = "aaaaaa"
t = "aaa"

print(count_occurrences_find(s, t))
print(count_occurrences_slice(s, t))
````

### 9.4 Build a result string gradually

Sometimes the easiest way is to accumulate output in a string.


````python
result = ""
for n in range(2, 30):
    for k in range(2, n):
        if n % k == 0:
            break
    else:
        result += str(n) + ", "

print(result[:-2])
````

### 9.5 Extract text between markers

A very useful `find` pattern:
- find the start marker
- jump to the end of the start marker
- find the closing marker from there
- slice the inside text


````python
t = "<title>Python 101</title>"

a = t.find("<title>")
if a >= 0:
    j = a + len("<title>")
    b = t.find("</title>", j)
    print(t[j:b])
else:
    print("No title")
````

### 9.6 Convert month number to month abbreviation using one string

This is a nice Unit 04 trick.


````python
months = "JANFEBMARAPRMAYJUNJULAUGSEPOCTNOVDEC"

for m in [1, 2, 8, 12]:
    print(m, "->", months[(m-1)*3:m*3])
````

### 9.7 Convert one code to another code using two compact strings

Another classic chapter trick:
- keep all country codes in one string
- keep all currency codes in another string
- locate the country code
- convert the block index


````python
countries = "usthjpcnsg"
currency_codes = "USDTHBJPYCNYSGD"

for c in ["us", "th", "jp", "cn", "sg"]:
    k = countries.find(c)
    if k >= 0:
        k //= 2
        print(c, "->", currency_codes[3*k:3*(k+1)])
````

## 10. Common mistakes

These are the mistakes students make all the time in Unit 04.


### Mistake A: unmatched or badly nested quotes

Examples that fail:
```python
s = "String
s = String'
s = "string'
s = "You shouted "Hey!" very loudly yesterday"
```

Fix them by matching quotes properly or escaping inner quotes.


### Mistake B: forgetting `.strip()` when reading text

If a user types `ok ` with a trailing space, then:

```python
s = input()
if s == 'ok':
    ...
```

will fail unexpectedly.

Prefer:
```python
s = input().strip()
```


### Mistake C: indexing beyond the string range

This can fail when the string is shorter than expected.


````python
s = "abc"

try:
    t = ""
    for i in range(10):
        t = s[i] + t
except Exception as e:
    print(type(e).__name__, ":", e)
````

### Mistake D: using a non-integer index

String indices must be integers.


````python
s = "abcdef"
n = 5

try:
    print(s[n/2])
except Exception as e:
    print(type(e).__name__, ":", e)

print("Correct:", s[n//2])
````

### Mistake E: trying to modify a character inside a string

This fails because strings are immutable.


````python
t = "python"

try:
    t[0], t[-1] = t[-1], t[0]
except Exception as e:
    print(type(e).__name__, ":", e)

t = t[-1] + t[1:-1] + t[0]
print("Correct:", t)
````

### Mistake F: forgetting to escape backslashes in file paths

For example:
```python
s = "c:\\temp\\data\\input.txt"
```

or use forward slashes instead.


### Mistake G: adding a string to a non-string directly

This fails:
```python
print("average = " + avg)
```

Fix it with:
```python
print("average = " + str(avg))
```

or simply:
```python
print("average =", avg)
```


````python
avg = 3.75
print("average =", avg)
print("average = " + str(avg))
````

## 11. Worked example: check whether any digit repeats

Problem idea:
- read a string
- ignore non-digits
- decide whether any digit appears more than once

This is one of the main worked examples in the chapter.


````python
def has_repeated_digit(text):
    for k, e in enumerate(text.strip()):
        if '0' <= e <= '9' and e in text[:k]:
            return True
    return False

print(has_repeated_digit("..125.9.0."))
print(has_repeated_digit("..125.9.2."))
````

### Why this works

At position `k`, the slice `text[:k]` contains everything to the left of the current character.

So:
- if the current character is not a digit, skip it
- if it is a digit and already appears in `text[:k]`, then a repeated digit exists


## 12. Worked example: date conversion

Convert a date from `mm/dd/yyyy` to `dd MMM yyyy`.

Example:
- `08/27/2014` becomes `27 AUG 2014`


````python
def convert_date_format(date_text):
    mm, dd, yyyy = date_text.strip().split("/")
    months = "JANFEBMARAPRMAYJUNJULAUGSEPOCTNOVDEC"
    m = int(mm)
    month_name = months[(m-1)*3:m*3]
    return f"{dd} {month_name} {yyyy}"

print(convert_date_format("08/27/2014"))
print(convert_date_format("02/01/2018"))
````

## 13. Worked example: count uppercase letters

Read one line and count uppercase English letters only.


````python
def count_uppercase_letters(s):
    c = 0
    for e in s:
        if 'A' <= e <= 'Z':
            c += 1
    return c

print(count_uppercase_letters("HelLo WoRLd"))
print(count_uppercase_letters("PYTHON123"))
print(count_uppercase_letters("yes!"))
````

## 14. Worked example: count vowel groups

A vowel group is a consecutive block of vowels.

Examples:
- `beautiful` has `eau`, `i`, `u` so the answer is `3`
- `group` has `ou` so the answer is `1`


````python
def count_vowel_groups(word):
    c = 0
    for i, ch in enumerate(word):
        if ch in "aeiou" and (i == 0 or word[i-1] not in "aeiou"):
            c += 1
    return c

for w in ["beautiful", "vowel", "group", "rhythm"]:
    print(w, "->", count_vowel_groups(w))
````

## Solution 1: duplicate every character


````python
def solution_1_duplicate_every_char(s):
    t = ''
    for e in s.strip():
        t += e * 2
    return t

print(solution_1_duplicate_every_char("pypy"))
````

## Solution 2: duplicate unless already adjacent


````python
def solution_2_duplicate_unless_adjacent(s):
    s = s.strip()
    t = ''
    s = ' ' + s + ' '
    for i in range(1, len(s)-1):
        t += s[i]
        if s[i-1] != s[i] != s[i+1]:
            t += s[i]
    return t

print(solution_2_duplicate_unless_adjacent("pythonnnaa"))
````

## Solution 3: palindrome check


````python
def solution_3_palindrome_flag(s):
    s = s.strip()
    if s == s[::-1]:
        return 'Y'
    return 'N'

print(solution_3_palindrome_flag("level"))
print(solution_3_palindrome_flag("python"))
````

## Solution 4: zero-pad a number


````python
def solution_4_zero_pad(d, n):
    t = "0" * n + str(d)
    t = t[-max(n, len(str(d))):]
    return t

print(solution_4_zero_pad(37, 5))
print(solution_4_zero_pad(12345, 3))
````

## Solution 5: one-digit hexadecimal to decimal


````python
def solution_5_hex_to_decimal(h):
    h = h.strip()
    d = '0123456789ABCDEF'.find(h)
    if d >= 0:
        return d
    return 'invalid hex digit'

for h in ["0", "9", "A", "F", "G"]:
    print(h, "->", solution_5_hex_to_decimal(h))
````

## Solution 6: count odd digits


````python
def solution_6_count_odd_digits(t):
    t = t.strip()
    c = 0
    for e in t:
        if e in '13579':
            c += 1
    return c

print(solution_6_count_odd_digits("a1357b80"))
````

## Solution 7: count adjacent vowel pairs


````python
def solution_7_count_adjacent_vowel_pairs(t):
    t = t.strip()
    c = 0
    for k in range(len(t)-1):
        if t[k] in 'aeiou' and t[k+1] in 'aeiou':
            c += 1
    return c

for w in ["beauty", "cooperate", "sky"]:
    print(w, "->", solution_7_count_adjacent_vowel_pairs(w))
````

## Solution 8: binary string to decimal


````python
def solution_8_binary_to_decimal(b):
    b = b.strip()[::-1]
    d = 0
    for i in range(len(b)):
        d += int(b[i]) * 2**i
    return d

print(solution_8_binary_to_decimal("01101"))
print(solution_8_binary_to_decimal("1010"))
````

## Solution 9: repeated-digit check


````python
def solution_9_has_repeated_digit(t):
    t = t.strip()
    for k, e in enumerate(t):
        if '0' <= e <= '9' and e in t[:k]:
            return True
    return False

print(solution_9_has_repeated_digit("..125.9.0."))
print(solution_9_has_repeated_digit("..125.9.2."))
````
