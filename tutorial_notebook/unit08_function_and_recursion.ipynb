# 08.1 Function Basics

A **function** is a named block of code.

Think of a function like a small machine:

```text
input values  ->  function  ->  output value
```

Example:

```python
def square(x):
    return x * x
```

Here:

| Part | Meaning |
|---|---|
| `def` | starts a function definition |
| `square` | function name |
| `x` | parameter, a value sent into the function |
| `return` | sends a result back to the caller |

Important rule:

```text
Define the function before calling it.
```


````python
def square(x):
    return x * x

print(square(3))
print(square(10))
````

## Function with no parameter

A function does not always need input.

Example: a function that always prints the same message.


````python
def print_header():
    print("====================")
    print("Student Report")
    print("====================")

print_header()
````

## Function with parameters

A **parameter** is a variable that receives a value when the function is called.

In this example, `name` is a parameter.


````python
def greet(name):
    print("Hello", name)

greet("Alice")
greet("Boss")
````

## Function with a return value

`return` sends a value back to the place where the function was called.

The caller can store the returned value in a variable.


````python
def add_tax(price):
    return price * 1.07

final_price = add_tax(100)
print(final_price)
````

## `print()` vs `return`

This is very important.

```text
print()  -> shows something on the screen
return   -> gives a value back to the program
```

A function can print something without returning a useful value.
A function can also return a value without printing anything.


````python
def show_double(x):
    print(x * 2)

def get_double(x):
    return x * 2

show_result = show_double(5)
get_result = get_double(5)

print("show_result =", show_result)
print("get_result =", get_result)
````

In the output above, `show_result` is `None` because `show_double()` does not return a value.

`None` means “no value was returned”.

# 08.2 Return, Tuple Return, and Local Variables

## `return` stops the function immediately

When Python reaches `return`, the function stops right there.

Any code after that `return` inside the same path will not run.


````python
def grade_status(score):
    if score >= 50:
        return "Pass"
    return "Fail"

print(grade_status(80))
print(grade_status(40))
````

## Returning more than one value

A function returns **one object**.

But that one object can be a tuple.

Example:


````python
def min_max(numbers):
    smallest = min(numbers)
    largest = max(numbers)
    return (smallest, largest)

result = min_max([7, 2, 9, 4])
print(result)

low, high = min_max([7, 2, 9, 4])
print("low =", low)
print("high =", high)
````

## Local variables

A variable created inside a function is a **local variable**.

It lives only inside that function.


````python
def calculate_total(prices):
    total = 0       # local variable
    for price in prices:
        total += price
    return total

print(calculate_total([10, 20, 30]))

# The variable total only exists inside calculate_total().
# Trying to print(total) here would cause an error.
````

Bad idea:


## Worked example: line intersection

We want a function that receives two lines:

```text
y = m*x + b
y = n*x + c
```

The function should return:

- `(x, y)` if the lines intersect at one point,
- `"parallel"` if the lines are parallel,
- `"same"` if they are the same line.

Reasoning:

```text
m*x + b = n*x + c
(m - n)*x = c - b
x = (c - b) / (m - n)
```

But if `m == n`, the lines have the same slope.
Then they are either parallel or the same line.


````python
def line_intersection(m, b, n, c):
    if m == n:
        if b == c:
            return "same"
        return "parallel"

    x = (c - b) / (m - n)
    y = m * x + b
    return (x, y)

print(line_intersection(2, 1, -1, 7))
print(line_intersection(2, 1, 2, 5))
print(line_intersection(2, 1, 2, 1))
````

# 08.3 Functions with List, Set, and Dictionary

Functions can receive lists, sets, and dictionaries as parameters.

This is very useful because real data often comes as a collection.

But there is one important warning.

```text
If a function changes a list, set, or dict directly, the original object outside the function also changes.
```

These containers are **mutable**, which means they can be edited.


## Example: changing a list inside a function


````python
def double_in_place(numbers):
    for i in range(len(numbers)):
        numbers[i] = numbers[i] * 2

values = [1, 2, 3]
double_in_place(values)
print(values)
````

The original list changed.

This can be useful, but it can also cause bugs if we did not expect it.


## Example: returning a new list instead


````python
def doubled_copy(numbers):
    result = []
    for value in numbers:
        result.append(value * 2)
    return result

values = [1, 2, 3]
new_values = doubled_copy(values)

print("original:", values)
print("new:", new_values)
````

## Functions with dictionaries

Dictionaries are useful for lookup problems.

Example: count how many times each word appears.


````python
def word_count(words):
    counts = {}
    for word in words:
        if word not in counts:
            counts[word] = 0
        counts[word] += 1
    return counts

print(word_count(["cat", "dog", "cat", "bird", "dog", "cat"]))
````

## Functions with sets

Sets are useful when we care about uniqueness or overlap.


````python
def common_students(class_a, class_b):
    return set(class_a) & set(class_b)

python_students = ["A", "B", "C"]
math_students = ["B", "C", "D"]

print(common_students(python_students, math_students))
````

## Worked example: course lookup with dictionary of sets

Suppose we store course enrollments like this:

```python
course_to_students = {
    "python": {"A", "B", "C"},
    "math": {"B", "C", "D"},
    "english": {"A", "D"}
}
```

We can create a function to count:

- students who take both courses,
- students who take exactly one of the two courses,
- students who take at least one of the two courses.


````python
def course_overlap(course_to_students, course1, course2):
    students1 = course_to_students.get(course1, set())
    students2 = course_to_students.get(course2, set())

    both = students1 & students2
    exactly_one = students1 ^ students2
    at_least_one = students1 | students2

    return (len(both), len(exactly_one), len(at_least_one))

course_to_students = {
    "python": {"A", "B", "C"},
    "math": {"B", "C", "D"},
    "english": {"A", "D"}
}

print(course_overlap(course_to_students, "python", "math"))
````

# 08.4 Function Refactoring

**Refactoring** means improving the structure of code without changing the final result.

Functions are useful for refactoring because they let us give a name to repeated logic.

Example problem:

We want to work with calendar dates.

Useful helper functions:

| Function | Job |
|---|---|
| `is_leap_year(y)` | check whether year `y` is a leap year |
| `days_in_year(y)` | return `365` or `366` |
| `day_of_year(d, m, y)` | return the day number within the year |
| `count_days_between(...)` | combine the helper functions to solve the full problem |


## Step 1: Write a small helper function

A leap year in the Gregorian calendar follows this rule:

```text
A year is a leap year if:
- it is divisible by 4, and
- it is not divisible by 100, unless it is also divisible by 400
```


````python
def is_leap_year(y):
    return y % 4 == 0 and (y % 100 != 0 or y % 400 == 0)

print(is_leap_year(2020))
print(is_leap_year(1900))
print(is_leap_year(2000))
````

## Step 2: Reuse the helper function


````python
def days_in_year(y):
    if is_leap_year(y):
        return 366
    return 365

print(days_in_year(2020))
print(days_in_year(2021))
````

## Step 3: Build a larger function from smaller functions


````python
def day_of_year(d, m, y):
    days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    if is_leap_year(y):
        days_in_month[1] = 29

    total = 0
    for i in range(m - 1):
        total += days_in_month[i]

    total += d
    return total

print(day_of_year(1, 1, 2020))
print(day_of_year(31, 12, 2020))
print(day_of_year(31, 12, 2021))
````

## Step 4: Use all helper functions together

The next function counts the number of days from date A to date B, including both dates.

For simplicity, date A and date B are assumed to be in different years, and date A comes before date B.


````python
def count_days_between(d1, m1, y1, d2, m2, y2):
    day_a = day_of_year(d1, m1, y1)
    day_b = day_of_year(d2, m2, y2)

    middle_days = 0
    for year in range(y1 + 1, y2):
        middle_days += days_in_year(year)

    remaining_days_in_year_a = days_in_year(y1) - day_a + 1
    return remaining_days_in_year_a + middle_days + day_b

print(count_days_between(1, 1, 2018, 1, 1, 2020))
print(count_days_between(25, 12, 1999, 9, 3, 2000))
````

# 08.5 Recursion

A **recursive function** is a function that calls itself.

A recursive function needs two parts:

| Part | Meaning |
|---|---|
| Base case | the case where the answer is known immediately |
| Recursive case | the case where the function calls itself with a smaller problem |

Without a base case, recursion may continue forever until Python stops it.


## Example: factorial

Factorial means:

```text
5! = 5 × 4 × 3 × 2 × 1
```

Recursive idea:

```text
factorial(0) = 1
factorial(n) = n × factorial(n-1)
```


````python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(0))
print(factorial(5))
````

## How the recursive calls work

For `factorial(4)`:

```text
factorial(4)
= 4 * factorial(3)
= 4 * 3 * factorial(2)
= 4 * 3 * 2 * factorial(1)
= 4 * 3 * 2 * 1 * factorial(0)
= 4 * 3 * 2 * 1 * 1
= 24
```


## Example: Fibonacci-style recursion

Some recursive functions have more than one recursive call.

Example sequence:

```text
fib(0) = 0
fib(1) = 1
fib(n) = fib(n-1) + fib(n-2)
```


````python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n - 1) + fib(n - 2)

for i in range(8):
    print(i, fib(i))
````

## Common recursion mistakes

### Mistake 1: Missing the base case

```python
def factorial(n):
    return n * factorial(n - 1)
```

This never stops.

### Mistake 2: Putting the base case after `return`

```python
def factorial(n):
    return n * factorial(n - 1)
    if n == 0:
        return 1
```

The `if` line is never reached.

### Mistake 3: Calling recursion with the same input

```python
def f(n):
    if n == 0:
        return 0
    return f(n)
```

The input never gets smaller, so the function never reaches the base case.


## Worked example: sum of a nested list

A nested list can contain integers or more lists.

Example:

```python
[1, [2, 3], [4, [5]]]
```

The total should be:

```text
1 + 2 + 3 + 4 + 5 = 15
```

Recursive idea:

```text
If x is an integer, return x.
If x is a list, sum every element inside it recursively.
```


````python
def sum_nested(x):
    if type(x) != list:
        return x

    total = 0
    for item in x:
        total += sum_nested(item)
    return total

print(sum_nested([1, [2, 3], [4, [5]]]))
print(sum_nested([[], [[]]]))
````
